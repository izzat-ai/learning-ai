- ushbu sahifada api dan ma'lumotlar olib , uni json ko'rinishida , jsonni esa DF qilib o'qish o'rganiladi
- api lar sun'iy intellektdan olindi

```python
import numpy as np
import pandas as pd
```

```python
url = "https://saurav.tech/NewsAPI/top-headlines/category/technology/us.json"
```

```python
import requests
r = requests.get(url)
```

```python
# ma'lumotlar tori yuklanganini tekshirish
r.status_code
```

```python
# json qilib o'qish
data_json = r.json()
data_json
```

```python
# dataframe qilish
df = pd.DataFrame(data_json['articles'])
df.head()
```

```python
df.info()
```

- ushbu DF da quyidagi ma'lumotlar berilgan :
- `source` - yangilik manbasi
- `author` - maqola muallifi
- `title` - yangilik sarlavhasi
- `description` - qisqacha mazmuni
- `url` -maqolaning to'liq linki
- va vahokazo

```python
url2 = "https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&order=market_cap_desc&per_page=10&page=1"
r2 = requests.get(url2)
```

```python
# ma'lumotlar tori yuklanganini tekshirish
r2.status_code
```

```python
# r2 ni json qilish
data_json2 = r2.json()
data_json2
```

```python
# data_json2 ni DF qilish
df2 = pd.DataFrame(data_json2)
df2.head()
```
