```python
import numpy as np
import pandas as pd
```

```python
import sqlite3 as sql
```

```python
db = sql.connect("company.db")
cursor = db.cursor()
```

```python
query = """
CREATE TABLE IF NOT EXISTS employees (
    employee_id INTEGER PRIMARY KEY,
    name TEXT,
    department TEXT,
    salary INT
);

INSERT INTO employees (employee_id, name, department, salary)
VALUES
    (1, 'Ali', 'IT', 800),
    (2, 'Hasan', 'HR', 700),
    (3, 'Vali', 'IT', 900),
    (4, 'Aziza', 'Sales', 750);
"""
cursor.executescript(query)
db.commit()
```

```python
query2 = """
CREATE TABLE IF NOT EXISTS projects (
    project_id INTEGER PRIMARY KEY,
    project_name TEXT,
    budget INT
);

INSERT INTO projects (project_id, project_name, budget)
VALUES
    (1, 'CRM System', 10000),
    (2, 'Website', 8000);
"""
cursor.executescript(query2)
db.commit()
```

```python
query3 = """
CREATE TABLE IF NOT EXISTS orders (
  order_id INT PIRMARY KEY,
  customer TEXT,
  product TEXT,
  amount INT,
  status TEXT
);
INSERT INTO orders (order_id, customer, product, amount, status)
VALUES
(1, 'Ali', 'Laptop', 1200, 'pending'),
(2, 'Hasan', 'Phone', 800, 'pending'),
(3, 'Vali', 'Tablet', 600, 'shipped');
"""
cursor.executescript(query3)
db.commit()
```
