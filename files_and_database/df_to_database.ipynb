```python
import numpy as np
import pandas as pd
```

```python
import sqlite3 as sql
```

```python
# database yaratish
db = sql.connect("shop.db")
cursor = db.cursor()
```

```python
# DataFramelar AI yordamida yaratildi
df_customers = pd.DataFrame({
    'customer_id': [1, 2, 3, 4],
    'name': ['Ali', 'Vali', 'Hasan', 'Husan'],
    'city': ['Tashkent', 'Samarkand', 'Bukhara', 'Tashkent']
})
df_customers
```

```python
# df_customersni databasega saqlash
df_customers.to_sql('customers', db)
```

```python
df_orders = pd.DataFrame({
    'order_id': [101, 102, 103, 104],
    'customer_id': [1, 2, 1, 3],
    'product': ['Laptop', 'Phone', 'Tablet', 'Laptop'],
    'amount': [1200, 800, 600, 1300]
})
df_orders
```

```python
# df_ordersni databasega saqlash
df_orders.to_sql('orders', db)
```

```python
# databasedagi jadvallarni aniqlash
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
cursor.fetchall()
```

```python
df_new_orders = pd.DataFrame({
    'order_id': [105, 106],
    'customer_id': [2, 4],
    'product': ['Phone', 'Laptop'],
    'amount': [900, 1400]
})
df_new_orders
```

```python
# orders jadvaliga qo'shimcha ma'lumotlar qo'shish
df_new_orders.to_sql('orders', db, if_exists='append', index=False)
```

```python
pd.read_sql("SELECT * FROM orders", db, index_col='index')
```
