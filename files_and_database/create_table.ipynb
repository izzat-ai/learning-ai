```python
import numpy as np
import pandas as pd
```

```python
import sqlite3 as sql
```

```python
db = sql.connect("company.db")
cursor = db.cursor()
```

```python
query = """
CREATE TABLE IF NOT EXISTS employees(
  employee_id INT PRIMARY KEY,
  name VARCHAR(50),
  department TEXT,
  salary INT
)
"""
cursor.execute(query)
db.commit()
```

```python
# databasedagi jadvallarni ko'rish
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
cursor.fetchall()
```

```python
query2 = """
INSERT INTO employees VALUES
(0, 'John Doe', 'HR', 11000),
(1, 'Jane Smith', 'IT', 14000),
(2, 'Daminik Toretto', 'AI', 17000)
"""
cursor.execute(query2)
db.commit()
```

```python
pd.read_sql("SELECT * FROM employees", db)
```

```python
query3 = """
CREATE TABLE IF NOT EXISTS projects(
  project_id INT PRIMARY KEY,
  project_name TEXT,
  budget INT
)
"""
cursor.execute(query3)
db.commit()
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
cursor.fetchall()
```

```python
query4 = """
INSERT INTO projects VALUES
(0, 'project A', 11000),
(1, 'project B', 14000),
(2, 'project C', 9000)
"""
cursor.execute(query4)
db.commit()
```

```python
pd.read_sql("SELECT * FROM projects", db)
```

```python
query5 = """
CREATE TABLE IF NOT EXISTS employee_projects(
  employee_id INT,
  project_id INT,
  role TEXT
);

INSERT INTO employee_projects VALUES
(1, 1, 'manager'),
(2, 2, 'hr'),
(3, 3, 'engineer');
"""
cursor.executescript(query5)
db.commit()
```
