```python
import numpy as np
import pandas as pd
```

```python
import sqlite3 as sql
```

```python
db = sql.connect("company.db")
cursor = db.cursor()
```

```python
query = """
CREATE TABLE IF NOT EXISTS employees(
  employee_id INT PRIMARY KEY,
  name VARCHAR(50),
  department TEXT,
  salary INT
)
"""
cursor.execute(query)
db.commit()
```

```python
# databasedagi jadvallarni ko'rish
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
cursor.fetchall()
```

```python
query2 = """
INSERT INTO employees VALUES
(0, 'John Doe', 'HR', 11000),
(1, 'Jane Smith', 'IT', 14000),
(2, 'Daminik Toretto', 'AI', 17000)
"""
cursor.execute(query2)
db.commit()
```
