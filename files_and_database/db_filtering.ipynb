```python
import numpy as np
import pandas as pd
```

```python
import sqlite3 as sql
```

```python
db = sql.connect("store.db")
cursor = db.cursor()
```

```python
# yangi jadval yaratib ma'lumotlar qo'shish
query = """
CREATE TABLE customers (
    customer_id INTEGER PRIMARY KEY,
    name TEXT,
    city TEXT,
    age INTEGER
);

INSERT INTO customers VALUES
(1, 'Ali', 'Tashkent', 25),
(2, 'Vali', 'Samarkand', 32),
(3, 'Aziza', 'Tashkent', 28),
(4, 'Maryam', 'Bukhara', 40),
(5, 'Hasan', 'Samarkand', 22),
(6, 'Husan', 'Tashkent', 35);
"""
cursor.executescript(query)
db.commit()
```

```python
query2 = """
CREATE TABLE orders (
    order_id INTEGER PRIMARY KEY,
    customer_id INTEGER,
    product TEXT,
    amount INTEGER,
    year INTEGER
);

INSERT INTO orders VALUES
(1, 1, 'Phone', 1200, 2023),
(2, 2, 'Laptop', 2200, 2024),
(3, 3, 'Tablet', 800, 2024),
(4, 1, 'Laptop', 2100, 2022),
(5, 4, 'Phone', 1300, 2023),
(6, 5, 'Tablet', 750, 2024),
(7, 6, 'Laptop', 2500, 2024);
"""
cursor.executescript(query2)
db.commit()
```

```python
# Toshkentda yashedigan va yoshi 30 dan kichik bo'lgan ma'lumotlarni olish
pd.read_sql("SELECT name, age FROM customers WHERE city='Tashkent' AND age<30", db)
```

```python
# mahsulot nomi va narxi 2024-yildagi 2000 dan oshiq ma'lumotlarni olish
pd.read_sql("SELECT product, amount FROM orders WHERE year==2024 AND amount>2000", db)
```

Ma'lumotlar bazasidan: Laptop sotib olgan lekin Tashkent shahrida yashamaydigan mijozlarga tegishli buyurtmalarni toping natijada buyurtma ID va mijoz ID chiqsin

```python
pd.read_sql("""
SELECT orders.order_id, orders.customer_id FROM orders
INNER JOIN customers ON orders.customer_id=customers.customer_id
WHERE orders.product='Laptop'
AND customers.city!='Tashkent'
""", db)
```
