```python
import numpy as np
import pandas as pd
```

```python
import sqlite3 as sql
```

```python
db = sql.connect("company.db")
cursor = db.cursor()
```

```python
query = """
CREATE TABLE employees (
    emp_id INTEGER PRIMARY KEY,
    name TEXT,
    department TEXT,
    salary INTEGER,
    city TEXT
);

INSERT INTO employees VALUES
(1, 'Ali', 'IT', 800, 'Tashkent'),
(2, 'Vali', 'IT', 900, 'Tashkent'),
(3, 'Hasan', 'HR', 600, 'Samarkand'),
(4, 'Husan', 'HR', 650, 'Tashkent'),
(5, 'Maryam', 'Finance', 1000, 'Bukhara'),
(6, 'Aziza', 'Finance', 950, 'Tashkent'),
(7, 'Aliya', 'IT', 800, 'Samarkand');
"""
cursor.executescript(query)
db.commit()
```

```python
query2 = """
CREATE TABLE projects (
    project_id INTEGER PRIMARY KEY,
    project_name TEXT,
    department TEXT,
    budget INTEGER
);

INSERT INTO projects VALUES
(1, 'AI System', 'IT', 50000),
(2, 'Website', 'IT', 20000),
(3, 'Recruitment', 'HR', 15000),
(4, 'Audit', 'Finance', 30000),
(5, 'AI System', 'IT', 50000);
"""
cursor.executescript(query2)
db.commit()
```

```python
# ismlar va maoshlarni yuqoridan pastga qarab olish
pd.read_sql("SELECT name, salary FROM employees ORDER BY salary DESC", db)
```

```python
# databasedagi bo'limlarni takrorlanmas shu bilan birga tartibli holda olish
comd = "SELECT DISTINCT employees.department FROM employees INNER JOIN projects ON employees.department = projects.department ORDER BY employees.department"
pd.read_sql(comd, db)
```

```python
# noyob proyekt nomlarini va ularning budjetlarini kamayish tartibida olish
pd.read_sql("SELECT DISTINCT project_name, budget FROM projects ORDER BY budget DESC", db)
```
