```python
# company databasini yaratish
db2 = sql.connect('company.db')

# cursor yaratish
cursor2 = db2.cursor()
```

```python
cursor2.execute("CREATE TABLE employees (emp_id INTEGER PRIMARY KEY, name TEXT, department TEXT, salary INTEGER);")
```

```python
query = """
INSERT INTO employees (emp_id, name, department, salary) VALUES
(101, 'Ali',    'IT',      1200),
(102, 'Vali',   'IT',      1300),
(103, 'Hasan',  'HR',       900),
(104, 'Husan',  'HR',       950),
(105, 'Maryam', 'Finance', 1500),
(106, 'Zaynab', 'Finance', 1600);
"""
# yuqoridagi ma'lumotlarni qo'shish
cursor2.execute(query)
```

```python
# o'zgarishlarni saqlash
db2.commit()
```

```python
query_tab = """
CREATE TABLE projects (
    project_id INTEGER PRIMARY KEY,
    emp_id INTEGER,
    project_name TEXT,
    hours INTEGER,
    FOREIGN KEY (emp_id) REFERENCES employees(emp_id)
);
"""
cursor2.execute(query_tab)
```

```python
# ma'lumotlar qo'shish
query_data = """
INSERT INTO projects (project_id, emp_id, project_name, hours) VALUES
(201, 101, 'AI System',     120),
(202, 102, 'Web App',       100),
(203, 101, 'Data Pipeline',  80),
(204, 103, 'Recruitment',   70),
(205, 105, 'Budget Plan',   90);
"""
cursor2.execute(query_data)
# o'zgarishlarni saqlash
db2.commit()
```
