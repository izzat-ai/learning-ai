On this page we will practice working with files and data warehouses. The DataFrames created are from AI.

```python
import numpy as np
import pandas as pd
```

```python
df_customers = pd.DataFrame({
    "customer_id": [1, 2, 3, 4],
    "name": ["Ali", "Vali", "Hasan", "Olim"],
    "city": ["Toshkent", "Samarqand", "Toshkent", "Buxoro"],
    "age": [23, 30, 27, 35]
})
df_customers
```

```python
# customers jadvalini csv faylga saqlash
df_customers.to_csv('customers.csv', index=False)
```

```python
# saqlangan faylni DF ko'rinishida o'qish
pd.read_csv('customers.csv')
```

```python
df_products = pd.DataFrame({
    "product": ["Laptop", "Phone", "Tablet", "Monitor"],
    "price": [1200, 800, 600, 300],
    "stock": [5, 10, 7, 12]
})
df_products
```

```python
# products jadvalini excel fayliga yozish
df_products.to_excel('products.xlsx', index=False)
```

```python
# excel faylni DF qilib o'qish
df_products2 = pd.read_excel('products.xlsx')
df_products2
```

```python
# narxi 500 dan yuqori bo'lgan mahsulotlarni olish
df_products2[df_products2['price']>500]
```

```python
df_users = pd.DataFrame({
    "user_id": [101, 102, 103],
    "username": ["ali01", "vali02", "hasan03"],
    "is_active": [True, False, True]
})
df_users
```

```python
# users jadvalini json fayliga saqlash
df_users.to_json('users.json', index=False)
```

```python
# json faylidan o'qish
df_users2 = pd.read_json('users.json')
df_users2
```

```python
# faqatgina faol foydalanuvchilarni olish
df_users2[df_users2['is_active'] == True]
```

```python
df_logs = pd.DataFrame({
    "date": ["2025-01-01", "2025-01-02", "2025-01-03"],
    "event": ["login", "logout", "login"],
    "user": ["Ali", "Vali", "Ali"]
})
df_logs
```

```python
# logs jadvalini oddiy matn ko'rinishida saqlash
df_logs.to_csv('logs.txt', index=False)
```

```python
# faylni qayta o'qish va faqat login hodisalarini olish
logs2 = pd.read_csv('logs.txt')
logs2[logs2['event'] == 'login']
```

#####**SQLite**

```python
import sqlite3 as sql
```

```python
db = sql.connect('databases.db')
cursor = db.cursor()
```

```python
df_orders = pd.DataFrame({
    "order_id": [1, 2, 3, 4],
    "customer": ["Ali", "Vali", "Ali", "Olim"],
    "amount": [250, 400, 150, 600]
})
df_orders
```

```python
# orders jadvalini databasega saqlash
df_orders.to_sql('orders', db)
```

```python
# databasedagi jadvalni o'qish
pd.read_sql("SELECT * FROM orders", db, index_col='index')
```

```python
# databasedan faqat Aliga oid ma'lumotlarni olish
pd.read_sql("SELECT * FROM orders WHERE customer=='Ali'", db, index_col='index')
```

```python
df_2023 = pd.DataFrame({
    "month": ["Jan", "Feb", "Mar"],
    "sales": [1200, 1500, 1100]
})

df_2024 = pd.DataFrame({
    "month": ["Jan", "Feb", "Mar"],
    "sales": [1400, 1600, 1300]
})
```

```python
# ikkala DF ni 2 xil faylga saqlash
df_2023.to_csv('first_file.csv', index=False)
df_2024.to_csv('second_file.csv', index=False)
```

```python
df1 = pd.read_csv('first_file.csv')
df2 = pd.read_csv('second_file.csv')

# ikkala DF ni bitta qilish
df1_df2 = pd.concat([df1, df2], ignore_index=True)
df1_df2
```
