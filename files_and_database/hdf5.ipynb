```python
import numpy as np
import pandas as pd
```

```python
df_sales = pd.DataFrame({
    'order_id': [1, 2, 3, 4],
    'year': [2023, 2023, 2024, 2024],
    'product': ['Phone', 'Laptop', 'Phone', 'Laptop'],
    'amount': [500, 1200, 650, 1400]
})
df_sales
```

```python
df_customers = pd.DataFrame({
    'customer_id': [101, 102, 103],
    'region': ['Asia', 'Europe', 'Asia'],
    'age': [25, 34, 29]
})
df_customers
```

- yuqoridagi ikki jadvalni hdf5 fayliga saqlash

```python
hdfobj = pd.HDFStore('sales.h5')
```

```python
# kalitlarni ko'rish
hdfobj.keys()
```

```python
# birinchi DF ni saqlash
hdfobj.put('sales', df_sales)
```

```python
# ikkinchi DF ni saqlash
hdfobj['customers'] = df_customers
```

```python
# kalitlarni ko'rish
hdfobj.keys()
```

```python
# customers ma'lumotlarini olish
hdfobj['customers']
```

```python
hdfobj['sales']
```

```python
phone = df_sales[df_sales['product'] == 'Phone']
phone
```

```python
phone.loc[:, 'amount'] = phone['amount']*1.1
```

```python
phone.drop('amount', inplace=True)
```

```python
phone
```

```python
phone.to_hdf('sales.h5', key='sales', mode='a', data_columns=True)
```
