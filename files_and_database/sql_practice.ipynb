```python
import numpy as np
import pandas as pd
```

```python
# sqliteni chaqirish
import sqlite3 as sql
from pprint import pprint as print
```

```python
# database yaratish
db = sql.connect("company.db")

# cursor yaratish
cursor = db.cursor()
```

```python
# jadval qo'shish
query1 = """
CREATE TABLE employees (
    emp_id INTEGER PRIMARY KEY,
    name TEXT,
    department TEXT,
    salary INTEGER
);
"""
cursor.execute(query1)
```

```python
# employees jadvaliga ma'lumotlar qo'shish
query2 = """
INSERT INTO employees VALUES
(1, 'Ali', 'IT', 1200),
(2, 'Vali', 'IT', 1300),
(3, 'Hasan', 'HR', 900),
(4, 'Husan', 'HR', 950),
(5, 'Maryam', 'Finance', 1500);
"""
cursor.execute(query2)

# o'zgarishlarni saqlash
db.commit()
```

```python
# projects jadvalini yaratish
query3 = """
CREATE TABLE projects (
    project_id INTEGER PRIMARY KEY,
    emp_id INTEGER,
    project_name TEXT,
    hours INTEGER
);
"""
cursor.execute(query3)
```

```python
# yuqoridagi yangi yaratilgan jadvalga ma'lumotlar qo'shish
query4 = """
INSERT INTO projects VALUES
(101, 1, 'AI System', 120),
(102, 1, 'Data Pipeline', 80),
(103, 2, 'Web App', 100),
(104, 3, 'Recruitment', 70);
"""
cursor.execute(query4)

# o'zgarishlarni saqlash
db.commit()
```

```python
# attendance jadvalini yaratish va ma'lumotlarini qo'shish
query5 = """
CREATE TABLE attendance (
    emp_id INTEGER,
    date TEXT,
    status TEXT
);

INSERT INTO attendance VALUES
(1, '2024-01-01', 'Present'),
(1, '2024-01-02', 'Absent'),
(2, '2024-01-01', 'Present'),
(3, '2024-01-01', 'Present'),
(4, '2024-01-01', 'Absent');
"""
cursor.executescript(query5)

# o'zgarishlarni saqlash
db.commit()
```

```python
# barcha jadvallarni aniqlash
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")

# barcha javoblarni olish
cursor.fetchall()
```

```python
employs = pd.read_sql("SELECT * FROM employees", db, index_col="emp_id")
employs
```

```python
# projects jadvalini DF qilib olish
projects = pd.read_sql("SELECT * FROM projects", db)
projects
```

```python
# attendance jadvalini DF qilib olish
att = pd.read_sql("SELECT * FROM attendance", db, index_col='emp_id')
att
```

```python
# databasedan faqatgina IT hodimlarini ismi va maoshini olish
pd.read_sql("SELECT name, salary FROM employees WHERE department='IT'", db)
```

```python
# projects jadvalidan har bir hodim nechta loyiha qiganini aniqlash
count_projects = pd.read_sql("SELECT emp_id, COUNT(project_id) AS loyihalar_soni FROM projects GROUP BY emp_id;", db)
count_projects
```

#####**SQL jadvalini DF qilib o'qish**

```python
# yangi database yaratish
db2 = sql.connect("school.db")

cursor2 = db2.cursor()
```

```python
# school databasiga yangi jadval va uning ma'lumotlarini qo'shish
query6 = """
CREATE TABLE students (
    student_id INTEGER PRIMARY KEY,
    name TEXT,
    class TEXT,
    age INTEGER
);

INSERT INTO students VALUES
(1, 'Ali', '10-A', 16),
(2, 'Vali', '10-A', 17),
(3, 'Hasan', '10-B', 16),
(4, 'Husan', '11-A', 17),
(5, 'Maryam', '11-B', 16);
"""
cursor2.executescript(query6)
db2.commit()
```

```python
# yangi jadval va ma'lumotlarini qo'shish
query7 = """
CREATE TABLE scores (
    student_id INTEGER,
    subject TEXT,
    score INTEGER
);

INSERT INTO scores VALUES
(1, 'Math', 85),
(1, 'Physics', 78),
(2, 'Math', 90),
(3, 'Math', 88),
(4, 'Physics', 92),
(5, 'Math', 95);
"""
cursor2.executescript(query7)
db2.commit()
```

```python
query8 = """
CREATE TABLE teachers (
    teacher_id INTEGER PRIMARY KEY,
    name TEXT,
    subject TEXT
);

INSERT INTO teachers VALUES
(1, 'Karimov', 'Math'),
(2, 'Akmalov', 'Physics');
"""
cursor2.executescript(query8)
db2.commit()
```

```python
# school databasidagi barcha jadvallarni aniqlash
cursor2.execute("SELECT name FROM sqlite_master WHERE type='table';")
cursor2.fetchall()
```

```python
# school databasedan sutdents jadvalini DF qilib o'qish
pd.read_sql("SELECT * FROM students", db2, index_col='student_id')
```

```python
# score jadvalidan faqatgina matematika fanining ma'lumotlarini olish
pd.read_sql("SELECT student_id, subject, score FROM scores WHERE subject='Math'", db2, index_col='student_id')
```

students va scores jadvallarini birlashtirilgan korinishda yuklab oling
