```python
num = int(input("Enter the number: "))
# Define a lambda function to calculate the square of a number
square = lambda num:num**2
print(square(num))
```

```python
# Get two numbers from the user and convert them to integers
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# Define a lambda function to add two numbers
add = lambda num1, num2: num1+num2
print(add(num1, num2))
```

```python
word = input("Enter the word: ")
# Define a lambda function to check if the first letter of the word is uppercase
word_up = lambda word: word[0].isupper()
print(word_up(word))
```

```python
numbers = [11, 22, 33, 44, 55, 66, 77]
# Define a lambda function to filter even numbers from the list
filt_nums = lambda numbers: [i for i in numbers if i%2==0]
```

```python
print(filt_nums(numbers))
```

```python
prices = [100, 250, 400, 150]
# Define a lambda function to increase a number by 10%
func = lambda i: i*1.1
# Use map to apply the function to each item in the 'prices' list and convert the result to a list
map_func = list(map(func, prices))
```

```python
# Print the list of prices after applying the function
print(map_func)
```

```python
students = [
    {'name': 'Ali', 'age': 25},
    {'name': 'Vali', 'age': 35},
    {'name': 'G\'ani', 'age': 20}
]
```

```python
# Sort the list of students by age using a lambda function as the key
sort_func = sorted(students, key=lambda i: i['age'])
```

```python
print(sort_func)
```
