```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
df_exam = pd.DataFrame({
    'Math': [78, 85, 90, 66, 72],
    'Physics': [82, 88, 91, 70, 75],
    'Chemistry': [80, 84, 89, 68, 73]
}, index=['Student_1', 'Student_2', 'Student_3', 'Student_4', 'Student_5'])
df_exam
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df_exam, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.xlabel('Subjects')
plt.ylabel('Students')
plt.show()
```

- *This heatmap visualizes student exam scores across three subjects, using color intensity to represent performance levels:*
- Top Performer (Deep Red): Student_3 achieved the highest scores in all subjects (Math: 90, Physics: 91, Chemistry: 89). Their row is highlighted in deep red, indicating peak performance.
- Average Performance (Light Red/White): Student_1 and Student_2 show consistent mid-to-high range scores, mostly between 78 and 88.
- Lower Scores (Deep Blue): Student_4 recorded the lowest scores, particularly in Math (66). The blue color immediately signals areas that may require academic support.
- *A quick glance shows that Physics tends to have the highest scores across most students, as its column appears warmer compared to Math or Chemistry*

```python
df_weather = pd.DataFrame({
    'Mon': [12, 18, 25],
    'Tue': [14, 20, 27],
    'Wed': [13, 19, 26],
    'Thu': [15, 21, 28],
    'Fri': [16, 22, 29]
}, index=['City_A', 'City_B', 'City_C'])
df_weather
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df_weather, annot=True, cmap='coolwarm', fmt='d', cbar=True)
plt.title('Weather Temperatures')
plt.xlabel('Days')
plt.ylabel('Cities')
plt.show()
```

- *This heatmap visualizes weekly temperature variations across three different cities from Monday to Friday:*
- Warmest City (City_C): Represented by the bottom row in deep red tones. It has the highest temperatures, ranging from 25C to 29C, peaking on Friday.
- Moderate City (City_B): Shown in the middle row with neutral/light tones. Temperatures are stable, gradually increasing from 18C to 22C throughout the week.
- Coolest City (City_A): The top row is in deep blue, indicating the lowest temperatures in the dataset (12C to 16C).
- General Trend: There is a visible warming trend across all cities as the week progresses toward Friday
