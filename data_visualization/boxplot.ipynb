```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_salary = pd.DataFrame({
    'Department': ['IT', 'IT', 'IT', 'HR', 'HR', 'HR', 'Sales', 'Sales', 'Sales'],
    'Salary': [1200, 1500, 1800, 900, 1000, 1100, 1300, 1600, 2000]
})
df_salary
```

```python
plt.figure(figsize=(8, 5))
plt.boxplot(df_salary['Salary'], vert=True, patch_artist=True)
plt.title('Salary Box Plot')
plt.show()
```

- *This Salary Box Plot illustrates the statistical distribution of employee salaries*
- The orange line inside the box shows that the median salary is approximately $1,300. This means half of the employees earn more than this amount, and half earn less
- The blue box represents the middle 50% of employees, whose salaries range between $1,100 and $1,600. This is known as the Interquartile Range (IQR)
- The minimum salary recorded is roughly $900, while the maximum non-outlier salary reaches $2,000
- There are no visible individual dots outside the whiskers. This suggests that the dataset contains no extreme values or anomalies that significantly deviate from the general salary structure

```python
df_scores2 = pd.DataFrame({
    'Class_A': [65, 70, 72, 75, 90],
    'Class_B': [60, 62, 64, 66, 68],
    'Class_C': [55, 58, 60, 62, 95]
})
df_scores2
```

```python
plt.figure(figsize=(10, 8))
plt.boxplot([df_scores2['Class_A'], df_scores2['Class_B'], df_scores2['Class_C']],
            tick_labels=['classA', 'classB', 'classC'],
            patch_artist=True)
plt.show()
```

- *This Multiple Box Plot visualizes the distribution of salaries (or scores) across three different departments (classA, classB, and classC)*
- Highest Earning Department (classA): This department leads with the highest median value (approx. 72). The middle 50% of employees are concentrated in a high-performance bracket between 70 and 75
- Most Consistent Distribution (classB): This department shows the least variance. The narrow box indicates that salaries are tightly clustered around 62 to 66, suggesting a highly standardized pay scale without any anomalies
- Lowest Median Performance (classC): This department has the lowest overall median at 60. Most employees in this group earn within the range of 58 to 62


Anomalies (Outliers):
- classA features an outlier at 90, representing an individual whose compensation significantly exceeds the departmental norm
- classC has a prominent outlier at 95. This is particularly striking as it is nearly 1.5 times the departmental average, highlighting a major deviation that might require further investigation (e.g., a specialist or a high-level manager)

```python
df_performance = pd.DataFrame({
    'Running': [12.1, 11.8, 12.0, 11.9, 12.2],
    'Swimming': [55, 54, 53, 52, 70],
    'Cycling': [30, 31, 29, 30, 30]
})
df_performance
```

```python
plt.figure(figsize=(7, 6))
plt.boxplot([df_performance['Running'], df_performance['Swimming'], df_performance['Cycling']],
            patch_artist=True,
            tick_labels=['running', 'swimming', 'cycling'])
plt.show()
```

- *This Multiple Box Plot provides a comparative study of performance across three sports: Running, Swimming, and Cycling*
- Running: Shows the lowest and most consistent results, tightly clustered between 11 and 13 units. The lack of outliers indicates a highly uniform performance level among participants
- Swimming: This sport features the highest overall scores, with a median of approximately 54 units. It also contains a significant outlier at 70 units, representing an exceptional performance that far exceeds the group average
- Cycling: Performance is highly stabilized around the 30-unit mark. While the core results are nearly identical (thin box), there are two outliers (one slightly above and one below), indicating minor deviations from the standard group performance
- *Swimming is the high-reward category with the greatest peak potential, whereas Running represents the most predictable and stable performance tier*
