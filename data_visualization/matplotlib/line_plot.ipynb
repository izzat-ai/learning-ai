```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# do'konning bir yillik sotuvi ma'lumotlari
df = pd.DataFrame({
    'oy': ['Yan', 'Fev', 'Mar', 'Apr', 'May', 'Iyn',
           'Iyl', 'Avg', 'Sen', 'Okt', 'Noy', 'Dek'],
    'sotuv': [12400, 13100, 15800, 14200, 16900, 18300,
              17500, 19200, 16100, 15400, 17800, 22100]
})
df
```

```python
plt.plot(df['oy'], df['sotuv'], color='r') # chiziq rangi qizil qilish

# grafikka nom berish
plt.title("2024-yil oylik sotuvlar")
plt.xlabel("oylar") # x o'qiga nom berish
plt.ylabel("sotuv (so'm)") # y o'qigan nom berish
plt.show()
```

- ushbu grafik - vaqt o'tishi bilan do'konning foydasi qanday bo'lganini ko'rsatadi
- yil davomida savdo barqaror o'sib borgan . Sotuvlar Yanvardagi eng past nuqtadan Dekabrdagi eng baland cho'qqiga ko'tarilgan . Deyarli 80% o'sish kuzatilgan
- grafikdan ikkita aniq pasayishni ham ko'rishimiz mumkin . Birinchisi aprel oyidagi ikkinchisi esa avgustdan oktabrgacha bo'lgan kuzgi pasayishdir

```python
np.random.seed(42)

# 200 nafar talabaning imtihon ballari (40 dan 100 gacha)
df5 = pd.DataFrame({
    'ball': np.random.randint(40, 101, size=200)
})
df5
```

```python
# bir oilaning oylik xarajatlari taqsimoti
df6 = pd.DataFrame({
    'kategoriya': ['Oziq-ovqat', 'Transport', 'Kiyim-kechak',
                   'Ko\'ngilochar', 'Kommunal', 'Boshqa'],
    'xarajat': [450000, 180000, 220000, 150000, 200000, 100000]
})
df6
```

```python
plt.pie(df6['xarajat'], labels=df6['kategoriya'], autopct='%1.1f%%', startangle=140, explode=(0.05, 0, 0, 0, 0, 0))
plt.title("Oylik xarajatlar taqsimoti")
plt.show()
```

- bir oilaning oylik harajatini deyarli 35% izi oziq-ovqatga ketadi . Kiyim-kechak va Kommunal to'lovlar oziq-ovqatdan keyingi eng yirik moliyaviy yuklamalar hisoblanadi
- ko'ngilochar narsalarga qilinadigan harajat , transport vositasiga qilinadigan sarflardan kam

```python
# uchta telefon brendining oylik sotuvlari ma'lumotlari
df7 = pd.DataFrame({
    'oy': ['Yan', 'Fev', 'Mar', 'Apr', 'May', 'Iyn',
           'Iyl', 'Avg', 'Sen', 'Okt', 'Noy', 'Dek'],
    'iphone': [850, 920, 780, 860, 910, 980,
               1050, 1100, 970, 890, 1200, 1450],
    'samsung': [720, 680, 750, 800, 770, 830,
                900, 950, 820, 780, 1050, 1200],
    'xiaomi': [430, 460, 410, 490, 520, 570,
               600, 650, 580, 510, 700, 850]
})
df7
```

```python
plt.plot(df7['oy'], df7['iphone'], color='blue', label='iphone')
plt.plot(df7['oy'], df7['samsung'], color='red', label='samsung')
plt.plot(df7['oy'], df7['xiaomi'], color='green', label='xiaomi')

plt.title("Telefon brendlari oylik sotuvlari")
plt.xlabel("oylar")
plt.ylabel("sotilgan (dona)")
plt.legend()
plt.show()
```

- barcha telefonlarning savdosini o'sish va pasayish oylari deyarli bir xil . Lekin iphone qolgan ikki telefondan ko'ra natijalari yuqori
- eng kam ko'rsatkichlarni xiaomi tashkil qilmoqda . Eng ko'p sotuvi 800 tagacha borgan , eng pasi 400 ta bo'lgan


```python
# onlayn do'konning 8 haftalik mijozlar ma'lumotlari
df8 = pd.DataFrame({
    'hafta': [1, 2, 3, 4, 5, 6, 7, 8],
    'yangi_mijozlar': [45, 62, 58, 74, 69, 83, 91, 88],
    'qaytgan_mijozlar': [120, 115, 132, 128, 145, 139, 158, 162]
})
df8
```

```python
plt.plot(df8['hafta'], df8['yangi_mijozlar'], color='blue', marker='o', linestyle='-', label='yangi_mijozlar')
plt.plot(df8['hafta'], df8['qaytgan_mijozlar'], color='#F97306', marker='s', linestyle='--', label='qaytgan_mijozlar')
plt.title("Haftalik mijozlar statistikasi")
plt.xlabel("hafta")
plt.ylabel("mijozlar soni")
plt.legend()
plt.show()
```

- onlayn do'konimizda 8 hafta ichida yangi mijozlar asta sekin o'sib borgan . Ammo qaytgan mijozlar yangi mijozlardan anchagina ko'p . Bu xizmat sifati yaxshi ekanligini va odamlar qayta murojat qilayotganini bildirishi mumkin
- 4-haftadan 5-haftaga o'tishda yangi mijozlar biroz kamaygan bo'lsa , huddi shu vaqt ichida qaytgan mijozlar ko'paygan

```python
# kompaniyaning bo'limlari ma'lumotlari
df9 = pd.DataFrame({
    'bo\'lim': ['Dasturlash', 'Dizayn', 'Marketing',
                'Moliya', 'HR', 'Sotuv'],
    'xodimlar': [34, 18, 22, 15, 12, 28],
    'ish_samarasi': [88, 76, 72, 81, 69, 84]
})
df9
```

```python
plt.bar(df9["bo'lim"], df9['ish_samarasi'], color=['blue', 'green', 'red', 'yellow', 'aqua', 'sienna'])
plt.title("Bo'limlar ish samarasi")
plt.xlabel("bo'limlar")
plt.ylabel("samaradorlik (%)")
plt.show()
```

- kampaniyaga eng ko'p samara olib kelayotgan bo'lim - dasturlash . U qariyb 80% dan yuqori foyda olib kelmoqda
- dasturlashdan keyingi bo'limlar esa moliya va sotuv bo'limlaridir . Ular 80% atrofidagi samaradorlikka erishgan
- samarasi past natijani ko'rsatgan bo'lim HR
- umumiy olganda barcha bo'limlarni samarasi yaxshi , hammasiniki 60% dan yuqori natijaga erishgan
