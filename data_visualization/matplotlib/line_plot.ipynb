```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# do'konning bir yillik sotuvi ma'lumotlari
df = pd.DataFrame({
    'oy': ['Yan', 'Fev', 'Mar', 'Apr', 'May', 'Iyn',
           'Iyl', 'Avg', 'Sen', 'Okt', 'Noy', 'Dek'],
    'sotuv': [12400, 13100, 15800, 14200, 16900, 18300,
              17500, 19200, 16100, 15400, 17800, 22100]
})
df
```

```python
plt.plot(df['oy'], df['sotuv'], color='r') # chiziq rangi qizil qilish

# grafikka nom berish
plt.title("2024-yil oylik sotuvlar")
plt.xlabel("oylar") # x o'qiga nom berish
plt.ylabel("sotuv (so'm)") # y o'qigan nom berish
plt.show()
```

- ushbu grafik - vaqt o'tishi bilan do'konning foydasi qanday bo'lganini ko'rsatadi

```python
# kompaniyaning 4 ta mintaqa bo'yicha choraklik sotuvi
df11 = pd.DataFrame({
    'chorak': ['Q1', 'Q2', 'Q3', 'Q4'],
    'shimoliy': [145000, 162000, 158000, 189000],
    'janubiy':  [98000, 112000, 125000, 134000],
    'sharqiy':  [76000, 88000, 95000, 102000],
    'g\'arbiy': [110000, 119000, 108000, 127000]
})
df11
```

```python
# kompaniyaning 7 yillik daromad va xarajatlari
df13 = pd.DataFrame({
    'yil': [2018, 2019, 2020, 2021, 2022, 2023, 2024],
    'daromad': [45000, 52000, 48000, 61000, 73000, 68000, 85000],
    'xarajat': [38000, 43000, 51000, 49000, 55000, 52000, 61000]
})
df13
```

```python
np.random.seed(42)

# kampaniyaning 12 oylik to'liq ma'lumotlari
df14 = pd.DataFrame({
    'oy': ['Yan', 'Fev', 'Mar', 'Apr', 'May', 'Iyn',
           'Iyl', 'Avg', 'Sen', 'Okt', 'Noy', 'Dek'],
    'sotuv':    [320, 415, 388, 470, 512, 598,
                 541, 623, 489, 445, 667, 789],
    'xarajat':  [210, 285, 260, 310, 340, 390,
                 355, 410, 320, 295, 430, 510],
    'mijozlar': [145, 198, 176, 223, 251, 312,
                 287, 334, 245, 212, 378, 445],
    'qaytish':  [23, 18, 31, 27, 19, 15,
                 22, 17, 28, 33, 14, 11]
})
df14
```

```python
# 2x2 grafik chizish
fig2, ax2 = plt.subplots(2, 2, figsize=(14, 10))

# oylik sotuv grafigi
ax2[0, 0].plot(df14['oy'], df14['sotuv'], color='blue')
ax2[0, 0].set_title("sotuvlar")

# oylik xarajatlar grafigi
ax2[0, 1].bar(df14['oy'], df14['xarajat'], color='orange')
ax2[0, 1].set_title("xarajatlar")

# oylik mijozlar grafigi
ax2[1, 0].plot(df14['oy'], df14['mijozlar'], color='green')
ax2[1, 0].set_title("mijozlar")

# oy bo'yicha qaytishlar grafigi
ax2[1, 1].bar(df14['oy'], df14['qaytish'], color='red')
ax2[1, 1].set_title("qaytarilgan mahsulotlar")

# grafiklar bir biriga tegmasligi uchun
plt.tight_layout()
plt.show()
```
