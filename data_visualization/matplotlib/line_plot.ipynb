```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# do'konning bir yillik sotuvi ma'lumotlari
df = pd.DataFrame({
    'oy': ['Yan', 'Fev', 'Mar', 'Apr', 'May', 'Iyn',
           'Iyl', 'Avg', 'Sen', 'Okt', 'Noy', 'Dek'],
    'sotuv': [12400, 13100, 15800, 14200, 16900, 18300,
              17500, 19200, 16100, 15400, 17800, 22100]
})
df
```

```python
plt.plot(df['oy'], df['sotuv'], color='r') # chiziq rangi qizil qilish

# grafikka nom berish
plt.title("2024-yil oylik sotuvlar")
plt.xlabel("oylar") # x o'qiga nom berish
plt.ylabel("sotuv (so'm)") # y o'qigan nom berish
plt.show()
```

- ushbu grafik - vaqt o'tishi bilan do'konning foydasi qanday bo'lganini ko'rsatadi
- yil davomida savdo barqaror o'sib borgan . Sotuvlar Yanvardagi eng past nuqtadan Dekabrdagi eng baland cho'qqiga ko'tarilgan . Deyarli 80% o'sish kuzatilgan
- grafikdan ikkita aniq pasayishni ham ko'rishimiz mumkin . Birinchisi aprel oyidagi ikkinchisi esa avgustdan oktabrgacha bo'lgan kuzgi pasayishdir

```python
# elektron do'konning mahsulot sotuvlari ma'lumotlari
df2 = pd.DataFrame({
    'mahsulot': ['Noutbuk', 'Telefon', 'Planshet', 'Qulaqchin', 'Soat'],
    'sotilgan': [142, 389, 95, 274, 183]
})
df2
```

```python
# IT sohasidagi hodimlarning tajribasi va maoshi haqidagi ma'lumotlar
df4 = pd.DataFrame({
    'tajriba_yil': [1, 2, 2, 3, 4, 4, 5, 6, 6, 7, 8, 9, 10, 10, 11],
    'maosh': [3200, 3800, 4100, 4500, 5200, 5800, 6400,
              7100, 7500, 8200, 9000, 9800, 10500, 11200, 12000]
})
df4
```

```python
plt.scatter(df4['tajriba_yil'], df4['maosh'], color='orange')
plt.title("Tajriba va maosh bog'liqligi")
plt.xlabel("tajriba (yil)")
plt.ylabel("maosh (million so'm)")
plt.show()
```

```python
np.random.seed(42)

# 200 nafar talabaning imtihon ballari (40 dan 100 gacha)
df5 = pd.DataFrame({
    'ball': np.random.randint(40, 101, size=200)
})
df5
```

```python
plt.hist(df5['ball'], bins=10, color='purple', edgecolor='white')
plt.title("Talabalar imtihon natijalari")
plt.xlabel("ball")
plt.ylabel("talabalar soni")
plt.show()
```

- Grafikda keskin chuqurliklar yoki anomal pasayishlar yo'q. Talabalarning ballari 40 dan 100 gacha bo'lgan barcha oraliqlarda deyarli bir xil taqsimlangan
- Eng baland ustun grafikning eng o'ng chekkasida — 94-100 ball oralig'ida turibdi. 200 ta talabadan 24 nafari imtihondan maksimal yoki unga juda yaqin ball olishga erishgan

```python
# bir oilaning oylik xarajatlari taqsimoti
df6 = pd.DataFrame({
    'kategoriya': ['Oziq-ovqat', 'Transport', 'Kiyim-kechak',
                   'Ko\'ngilochar', 'Kommunal', 'Boshqa'],
    'xarajat': [450000, 180000, 220000, 150000, 200000, 100000]
})
df6
```

```python
plt.pie(df6['xarajat'], labels=df6['kategoriya'], autopct='%1.1f%%', startangle=140, explode=(0.05, 0, 0, 0, 0, 0))
plt.title("Oylik xarajatlar")
plt.show()
```

- bir oilaning oylik harajatini deyarli 35% izi oziq-ovqatga ketadi . Kiyim-kechak va Kommunal to'lovlar oziq-ovqatdan keyingi eng yirik moliyaviy yuklamalar hisoblanadi
- ko'ngilochar narsalarga qilinadigan harajat , transport vositasiga qilinadigan sarflardan kam
