```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# do'konning bir yillik sotuvi ma'lumotlari
df = pd.DataFrame({
    'oy': ['Yan', 'Fev', 'Mar', 'Apr', 'May', 'Iyn',
           'Iyl', 'Avg', 'Sen', 'Okt', 'Noy', 'Dek'],
    'sotuv': [12400, 13100, 15800, 14200, 16900, 18300,
              17500, 19200, 16100, 15400, 17800, 22100]
})
df
```

```python
np.random.seed(42)

# kampaniyaning 12 oylik to'liq ma'lumotlari
df14 = pd.DataFrame({
    'oy': ['Yan', 'Fev', 'Mar', 'Apr', 'May', 'Iyn',
           'Iyl', 'Avg', 'Sen', 'Okt', 'Noy', 'Dek'],
    'sotuv':    [320, 415, 388, 470, 512, 598,
                 541, 623, 489, 445, 667, 789],
    'xarajat':  [210, 285, 260, 310, 340, 390,
                 355, 410, 320, 295, 430, 510],
    'mijozlar': [145, 198, 176, 223, 251, 312,
                 287, 334, 245, 212, 378, 445],
    'qaytish':  [23, 18, 31, 27, 19, 15,
                 22, 17, 28, 33, 14, 11]
})
df14
```

```python
# 2x2 grafik chizish
fig2, ax2 = plt.subplots(2, 2, figsize=(14, 10))

# oylik sotuv grafigi
ax2[0, 0].plot(df14['oy'], df14['sotuv'], color='blue')
ax2[0, 0].set_title("sotuvlar")

# oylik xarajatlar grafigi
ax2[0, 1].bar(df14['oy'], df14['xarajat'], color='orange')
ax2[0, 1].set_title("xarajatlar")

# oylik mijozlar grafigi
ax2[1, 0].plot(df14['oy'], df14['mijozlar'], color='green')
ax2[1, 0].set_title("mijozlar")

# oy bo'yicha qaytishlar grafigi
ax2[1, 1].bar(df14['oy'], df14['qaytish'], color='red')
ax2[1, 1].set_title("qaytarilgan mahsulotlar")

fig2.suptitle("Kompaniya yillik statistikasi",
             fontsize=16, fontweight='bold')

# grafiklar bir biriga tegmasligi uchun
plt.tight_layout()
plt.show()
```

```python
# IT sohasidagi ma'lumotlar
df15 = pd.DataFrame({
    'soha': ['AI/ML', 'Kibexavfsizlik', 'Veb dasturlash',
             'Mobil dasturlash', 'DevOps', 'Data Science'],
    'ish_joylari': [1840, 1250, 2100, 1680, 980, 1520],
    'ortacha_maosh': [4200, 3800, 2900, 3100, 3600, 4000],
    'talab_oshishi': [38, 29, 15, 21, 25, 33]
})
df15
```

```python
# bir qator va 3 ustundan iborat grafik chizish
fig3, ax3 = plt.subplots(1, 3, figsize=(16, 5))

ax3[0].barh(df15['soha'], df15['ish_joylari'], color='steelblue')
ax3[0].set_title("ish joylari soni")
ax3[0].set_xlim(0, 2500)
ax3[0].set_xlabel("ish joylari soni")

ax3[1].barh(df15['soha'], df15['ortacha_maosh'], color='mediumseagreen')
ax3[1].set_title("o'rtacha maosh")
ax3[1].set_xlim(0, 5000)
ax3[1].set_xlabel("o'rtacha maosh")

ax3[2].barh(df15['soha'], df15['talab_oshishi'], color='tomato')
ax3[2].set_title("talab oshishi")
ax3[2].set_xlim(0, 50)
ax3[2].set_xlabel("talab oshishi (%)")

fig3.suptitle("IT sohasidagi ma'lumotlar", fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
np.random.seed(15)

# O'zbekiston shaharlari demografik va iqtisodiy ko'rsatkichlari
df16 = pd.DataFrame({
    'shahar': ['Toshkent', 'Samarqand', 'Buxoro', 'Namangan',
               'Andijon', "Farg'ona", 'Nukus', 'Qarshi'],
    'aholi_soni': [2954, 564, 391, 523, 478, 441, 320, 285],  # mingda
    'orta_yosh': [29.5, 31.2, 30.8, 27.4, 28.1, 28.9, 26.8, 27.9],
    'ishsizlik': [7.2, 9.8, 8.5, 11.3, 10.7, 9.2, 12.1, 10.5]
})
df16
```
