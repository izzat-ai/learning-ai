```python
import numpy as np
import pandas as pd
```

```python
# matplotlib kutubxonasini chaqirish
import matplotlib.pyplot as plt
```

```python
# grafiklar colabni o'zida chiqishi uchun
%matplotlib inline
```

```python
df_sales = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [120, 150, 170, 160, 180, 200]
})
df_sales
```

```python
# oylar bo'yicha savdoning o'sishi
plt.plot(df_sales['Month'], df_sales['Sales'])
plt.show()
```

```python
df_students = pd.DataFrame({
    'Subject': ['Math', 'Physics', 'Chemistry', 'Biology'],
    'AverageScore': [78, 85, 74, 88]
})
df_students
```

```python
plt.bar(df_students['Subject'], df_students['AverageScore'])
plt.xlabel('subjects')
plt.ylabel('average score')
plt.title("fanlar bo'yicha taqqoslash grafigi")
plt.show()
```

```python
df_temperature = pd.DataFrame({
    'Day': list(range(1, 11)),
    'City_A': [20, 21, 19, 22, 23, 24, 25, 26, 24, 23],
    'City_B': [18, 19, 17, 18, 20, 21, 22, 23, 22, 21]
})
df_temperature
```

```python
plt.plot(df_temperature['Day'], df_temperature['City_A'], linestyle=':', color='b', marker='o')
plt.plot(df_temperature['Day'], df_temperature['City_B'], linestyle='-', color='r')
plt.xlabel('days')
plt.ylabel('temperature')
plt.legend(['City_A', 'City_B'], loc=0)
plt.show()
```

```python
df_employees = pd.DataFrame({
    'Department': ['IT', 'HR', 'Sales', 'Marketing', 'Finance'],
    'Employees': [15, 5, 20, 10, 8]
})
df_employees
```

- har bir bo'limdagi xodimlar sonini taqqoslash

```python
plt.bar(df_employees['Department'], df_employees['Employees'])
plt.xlabel('department')
plt.ylabel('employees')
plt.title('comparison of the number of employees')
plt.show()
```

```python
df_visits = pd.DataFrame({
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'],
    'Visits': [320, 450, 500, 480, 600, 750, 700]
})
df_visits
```

- *haftaning kunlari bo'yicha saytga kirishlar dinamikasini grafigini chizish*

```python
plt.plot(df_visits['Day'], df_visits['Visits'], linewidth=2.5, alpha=0.7, linestyle=':', color='b', marker='o')
plt.xlabel('day')
plt.ylabel('visits')
plt.title('visits per day')
plt.show()
```

```python
df_scores = pd.DataFrame({
    'Score': [55, 60, 65, 70, 75, 80, 85, 90, 95, 100,
              72, 68, 77, 83, 91, 88, 66, 74]
})
df_scores
```

- *talabalar baholarining taqsimot grafigi*

```python
plt.hist(df_scores['Score'], bins=5, edgecolor='black')
plt.xlabel('score')
plt.ylabel('frequency')
plt.title('grade distribution')
plt.show()
```

```python
df_ads = pd.DataFrame({
    'Ad_Budget': [100, 150, 200, 250, 300, 350, 400],
    'Revenue': [1000, 1400, 1800, 2100, 2600, 3000, 3400]
})
df_ads
```

```python
plt.scatter(df_ads['Ad_Budget'], df_ads['Revenue'], color='green', marker='o', s=100, alpha=0.7)
plt.show()
```
