```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_sites = pd.DataFrame({
    'Day': [1, 2, 3, 4, 5],
    'Site_A': [120, 150, 170, 160, 180],
    'Site_B': [80, 90, 95, 100, 110],
    'Site_C': [200, 210, 190, 220, 230]
})
df_sites
```

```python
plt.plot(df_sites['Day'], df_sites['Site_A'], marker='o')
plt.plot(df_sites['Day'], df_sites['Site_B'], marker='>')
plt.plot(df_sites['Day'], df_sites['Site_C'], marker='.')
plt.xlabel('days')
plt.ylabel('traffic')
plt.legend(['Site_A', 'Site_B', 'Site_C'])
plt.show()
```

- *This multi-line chart illustrates the traffic dynamics of three different websites (Site_A, Site_B, Site_C) over a 5-day period*
- The green line representing Site_C holds the dominant position throughout the period, reaching a peak of 230 traffic units on Day 5. Despite a minor dip on Day 3, its overall trajectory remains upward
- Site_A, shown in blue, demonstrates a consistent growth pattern, starting at 120 units on Day 1 and climbing to 180 units by Day 5
- The orange line for Site_B shows the lowest traffic volume among the three. Its growth is much slower, moving from 80 to 110 units over the 5 days
- *All three websites show a positive growth trend. However, Site_C continues to lead the market significantly, while Site_B struggles to close the gap with the other two competitors*

```python
df_weather2 = pd.DataFrame({
    'Day': [1, 2, 3, 4, 5],
    'Tashkent': [30, 32, 33, 31, 34],
    'Samarkand': [28, 29, 30, 29, 31],
    'Bukhara': [35, 36, 37, 36, 38]
})
df_weather2
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].plot(df_weather2['Day'], df_weather2['Tashkent'], color='#F97306')
ax[1].plot(df_weather2['Day'], df_weather2['Samarkand'], color='#C875C4')
ax[2].plot(df_weather2['Day'], df_weather2['Bukhara'], color='#9ACD32')

# har bir grafikni nomlash
ax[0].set_title('Tashkent')
ax[1].set_title('Samarkand')
ax[2].set_title('Bukhara')

plt.show()
```

- *This visualization, created using subplots, compares the 5-day temperature trends across three major cities: Tashkent, Samarkand, and Bukhara*
- The temperature rose from 30.0C on Day 1 to a peak of 34.0C on Day 5. There was a notable fluctuations, specifically a sharp drop on Day 4 before reaching the final peak
- Follows a similar trend but remains generally cooler, with temperatures ranging between 28.0C and 31.0C. The mid-period dip is also visible here
- This is the warmest city in the dataset. Starting at 35.0C, it reached a high of 38.0C by the end of the period. Despite the shared Day 4 dip, its absolute values remain the highest
- *The synchronized temperature patterns across all three charts (peaking on Day 3 and dipping on Day 4) suggest a shared regional weather event or atmospheric change affecting the entire area simultaneously*

```python
df_exam = pd.DataFrame({
    'Student': [1, 2, 3, 4, 5],
    'Math': [78, 85, 90, 88, 92],
    'Physics': [80, 82, 88, 86, 90],
    'English': [75, 80, 83, 85, 87]
})
df_exam
```

```python
plt.plot(df_exam['Student'], df_exam['Math'], color='#030764', linestyle=':', linewidth=2.5)
plt.plot(df_exam['Student'], df_exam['Physics'], color='#008000', linestyle='--')
plt.plot(df_exam['Student'], df_exam['English'], color='#C20078')
plt.xticks(df_exam['Student'])
plt.xlabel('Students')
plt.ylabel('Grades')
plt.legend(['Math', 'Physics', 'English'])
plt.show()
```

- *This line chart illustrates the performance trends of five students across three different subjects*
- Represented by the dark blue dotted line. This subject shows the highest rate of improvement and the best overall scores. While the first student scored 78, the fifth student achieved the peak result of 92
- Indicated by the green dashed line. Grades in this subject fluctuate between 80 and 90. The third and fifth students recorded the strongest performances in this category
- Shown as a solid magenta line. While English scores show the most consistent and steady growth, they remain lower in absolute terms compared to the other subjects, ranging from 75 to 87
- *A slight dip is observed across almost all subjects after the third student. This suggests that the fourth student's performance was slightly below the group average before a recovery was seen with the fifth student*

```python
df_products = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr'],
    'Phone': [300, 320, 310, 330],
    'Laptop': [150, 160, 170, 165],
    'Tablet': [200, 210, 205, 215]
})
df_products
```

```python
fig2, ax2 = plt.subplots(2, 2, figsize=(10, 8))
ax2[0, 0].plot(df_products['Month'], df_products['Phone'], color='#D2691E')
ax2[0, 0].set_title('phone')

ax2[0, 1].plot(df_products['Month'], df_products['Laptop'], color='#3D1C02')
ax2[0, 1].set_title('laptop')

ax2[1, 0].plot(df_products['Month'], df_products['Tablet'], color='#800080')
ax2[1, 0].set_title('tablet')

ax2[1, 1].axis('off')

plt.show()
```

- *This visualization showcases the sales trends for three different technology categories during the first four months of the year*
- Sales started at 300 units in January and peaked at 330 units in April. There was an upward trend in February, followed by a temporary dip in March
- Sales reached their highest point in March (170 units) but experienced a decline in April. Overall, the performance remains better than the January starting point (150 units)
- This category follows a similar pattern to phones, with a dip in March followed by a significant surge in April, reaching 215 units
- *A common trend across all devices is a sales boost in February. While March was a relatively slow month for all categories, April saw a strong recovery and growth in the phone and tablet markets*

```python
df_fin = pd.DataFrame({
    'Month': [1, 2, 3, 4, 5],
    'Revenue': [1000, 1200, 1400, 1300, 1500],
    'Cost': [600, 700, 800, 750, 850]
})
df_fin
```

```python
plt.plot(df_fin['Month'], df_fin['Revenue'], color='#008000', marker='o')
plt.plot(df_fin['Month'], df_fin['Cost'], color='#E50000', marker='>')

plt.xticks(df_fin['Month'])

plt.legend(['Revenue', 'Cost'])

plt.xlabel('Month')
plt.ylabel('Amount')
plt.show()
```

- *This line chart tracks the relationship between Revenue and Cost over a 5-month period*
- There is a clear overall upward trajectory in revenue. Starting at 1,000 units in Month 1, it reached its peak of 1,500 units by Month 5, despite a slight dip in Month 4
- Costs have also increased over time but remained consistently lower than revenue, ranging between 600 and 850 units
- The widening gap between the two lines indicates increasing operational efficiency. The maximum net profit was achieved in Month 5, where the distance between revenue and cost is greatest
- *The business shows healthy growth and stability. The revenue consistently outperforms costs, and the increasing margin suggests improving profitability as the months progress*

```python
df_sports = pd.DataFrame({
    'Match': [1, 2, 3, 4, 5],
    'Football': [2, 3, 1, 4, 2],
    'Basketball': [80, 85, 78, 90, 88],
    'Volleyball': [25, 23, 26, 24, 27]
})
df_sports
```

```python
fig3, ax3 = plt.subplots(3, 1, figsize=(6, 10))
ax3[0].plot(df_sports['Match'], df_sports['Football'], color='#15B01A')
ax3[0].set_title('Football')
ax3[0].set_xticks(df_sports['Match'])

ax3[1].plot(df_sports['Match'], df_sports['Basketball'], color='#A52A2A', linestyle='--')
ax3[1].set_title('Basketball')
ax3[1].set_xticks(df_sports['Match'])

ax3[2].plot(df_sports['Match'], df_sports['Volleyball'], color='#6E750E', linestyle=':')
ax3[2].set_title('Volleyball')
ax3[2].set_xticks(df_sports['Match'])

plt.show()
```

- *This set of subplots compares the performance of three sportsFootball, Basketball, and Volleyballacross a series of 5 matches*
- Displays high volatility. The team reached a peak performance of 4 units in Match 4, following a significant slump to 1 unit in Match 3
- Scores are stabilized in a higher range (78-90 points). Similar to the football trend, Match 4 represents the peak performance, while Match 3 shows the lowest score in the series
- Unlike the other two, volleyball shows a more tempered progression. The results fluctuate moderately between 23 and 27 units, indicating a more consistent performance level
- *There is a synchronized trend across all sports where Match 3 appears to be a universal low point, while Match 4 marks a significant recovery and peak for the teams*
