```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_devices = pd.DataFrame({
    'Device': ['Mobile', 'Desktop', 'Tablet', 'Other'],
    'Users': [520, 310, 120, 50]
})
df_devices
```

```python
df_sum = df_devices.groupby('Device')['Users'].sum()
explode = (0, 0.1, 0, 0)
plt.pie(df_sum, labels=df_sum.index, autopct='%1.1f%%', shadow=True, explode=explode)
plt.title('Devices')
plt.show()
```

- *This pie chart visualizes the distribution of users across various device categories:*
- Mobile: This is the dominant category, accounting for more than half of the total user base at 52.0%. The "explode" feature is applied to this segment to emphasize that it is the primary source of traffic
- Desktop: Ranks second, representing 31.0% of the total distribution
- Tablet: Holds a smaller but notable share, used by 12.0% of the audience
- Other: Represents the smallest segment, with only 5.0% of users accessing the platform via alternative devices

```python
df_budget = pd.DataFrame({
    'Category': ['Rent', 'Food', 'Transport', 'Entertainment', 'Savings'],
    'Amount': [1500, 800, 300, 400, 600]
})
df_budget
```

```python
df_budget_sum = df_budget.groupby('Category')['Amount'].sum()
df_budget_sum
```

```python
plt.pie(df_budget_sum, labels=df_budget_sum.index, autopct='%1.1f%%', shadow=True, explode=(0, 0, 0.1, 0, 0))
plt.title('Budget Categories')
plt.show()
```

- *This pie chart visualizes the distribution of monthly budget expenditures across several categories:*
- Rent: Represents the largest portion of the budget at 41.7%. This segment is "exploded" to emphasize it as the primary monthly expense
- Food: The second-highest expenditure, accounting for 22.2% of the total budget
- Savings: A healthy 16.7% of the monthly income is being set aside for future use
- Entertainment: Leisure and entertainment activities take up 11.1% of the monthly spending
- Transport: The smallest expense category, making up only 8.3% of the budget
- *Fixed living costs (Rent and Food) dominate the budget, consuming 63.9% of the total income. The savings rate of 16.7% indicates a disciplined financial approach toward long-term goals*

```python
df_students = pd.DataFrame({
    'Grade': ['A', 'B', 'C', 'D', 'F'],
    'Count': [18, 25, 30, 15, 7]
})
df_students
```

```python
df_st_sum = df_students.groupby('Grade')['Count'].sum()
df_st_sum
```

```python
plt.pie(df_st_sum, labels=df_st_sum.index, autopct='%1.1f%%', shadow=True, explode=(0, 0, 0.1, 0, 0), colors=['#F5DEB3', '#C1F80A', '#008000', '#40E0D0', '#FF0000'])
plt.title('Student Grades')
plt.show()
```

- *This pie chart visualizes the distribution of student grades for a specific course or exam:*
- C (Average): The dominant category, representing 31.6% of the students. It is highlighted with an "explode" effect to show it is the most common grade.
- B (Good): Accounted for by 26.3% of the student population, showing a solid performance from a quarter of the class.
- A (Excellent): A significant portion of students, 18.9%, achieved the highest possible grade.
- D (Poor): Roughly 15.8% of the students scored on the lower end of the passing spectrum
- F (Fail): The smallest segment at 7.4%, representing students who did not meet the minimum requirements
