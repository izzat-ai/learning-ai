```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')

# Pairplot chizish
# hue='species' - gullarni turi bo'yicha ranglarga ajratadi
sns.pairplot(iris, hue='species', palette='husl', markers=["o", "s", "D"])

plt.suptitle("Iris gullari o'lchamlari o'rtasidagi bog'liqlik", y=1.02)
plt.show()
```
