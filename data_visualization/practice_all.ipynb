```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
df_metrics = pd.DataFrame({
    "age": [21, 25, 30, 35, 40, 45],
    "salary": [500, 700, 1200, 1600, 2000, 2300],
    "experience": [1, 3, 6, 9, 12, 15]
})
df_metrics
```

```python
metrics_corr = df_metrics.corr()
metrics_corr
```

```python
sns.heatmap(metrics_corr, annot=True)
plt.show()
```

- The image displays a Correlation Matrix Heatmap for three variables: age, salary, and experience.
- Perfect Correlation: The correlation coefficient between all pairs of variables is exactly 1. This indicates a perfect positive linear relationship.
- Interpretation: As age increases, salary and experience increase in perfect lockstep. There is no variance or noise in the data.
- Observation: In real-world data science, seeing a correlation of 1 across different features (not just the diagonal) is highly unusual. It suggests that the features might be redundant or derived directly from each other using a fixed multiplier.
- Color Scale: The color bar on the right shows an extremely narrow range (from 0.9975 to 1.0000), emphasizing that there is virtually no statistical difference in how these variables move relative to one another

```python
df_temperature = pd.DataFrame({
    "day": ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
    "temp": [22, 24, 23, 25, 26, 27, 28]
})
df_temperature
```

```python
plt.plot(df_temperature['day'], df_temperature['temp'])
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.show()
```

- The line chart illustrates the temperature variations over a week based on the day and temp columns:
- Overall Trend: There is a clear upward trend in temperature throughout the week. The lowest temperature is recorded on Monday (22C), and the peak is reached on Sunday (28C).

Daily Breakdown:
- Monday - Tuesday: The temperature rises from 22C to 24C.
- Wednesday: A slight dip is observed, dropping to 23C, which is the only decrease during the week.
- Thursday - Sunday: From Thursday (25C) onwards, the temperature increases steadily and consistently until the end of the week.
- Conclusion: Aside from a minor fluctuation on Wednesday, the weather shows a consistent warming trend, leading to a significantly warmer weekend

```python
df_students = pd.DataFrame({
    "faculty": ["IT", "Economy", "Medicine", "Law"],
    "students": [320, 210, 180, 150]
})
df_students
```

```python
df_study = pd.DataFrame({
    "study_hours": [1, 2, 3, 4, 5, 6, 7],
    "score": [50, 55, 65, 70, 78, 85, 90]
})
df_study
```

Key Data Points:
- Studying for 1 hour results in a score of approximately 50.

```python
df_sales = pd.DataFrame({
    "month": ["Jan", "Feb", "Mar", "Apr"],
    "laptop": [120, 150, 170, 160],
    "phone": [200, 220, 210, 230]
})
df_sales
```

```python
plt.plot(df_sales['month'], df_sales['laptop'], label='Laptop')
plt.plot(df_sales['month'], df_sales['phone'], label='Phone')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].plot(df_sales['month'], df_sales['laptop'])
ax[0].set_title('laptop_sales')

ax[1].plot(df_sales['month'], df_sales['phone'])
ax[1].set_title('phone_sales')

plt.show()
```

- Laptop Sales: Peaked in March (170) before a slight decline in April.
- Phone Sales: Experienced a dip in March but surged to a high of 230 in April

```python
df_health = pd.DataFrame({
    "height": [165, 170, 175, 180, 185],
    "weight": [60, 65, 70, 78, 85],
    "age": [20, 22, 25, 27, 30]
})
df_health
```

```python
fig1, ax1 = plt.subplots(1, 2, figsize=(10, 5))
ax1[0].scatter(df_health['age'], df_health['height'])
ax1[0].set_title('height')

ax1[1].scatter(df_health['age'], df_health['weight'])
ax1[1].set_title('weight')

fig1.supxlabel('age')
plt.show()
```

- The visualization displays two subplots showing the relationship between age and physical attributes (height and weight):
- Overall Trend: Both plots exhibit a strong positive linear correlation. As the age increases, both height and weight show a steady upward trend.
- Height vs. Age: At age 20, the height starts at 165 cm and rises consistently to reach 185 cm by age 30.
- Weight vs. Age: Similarly, weight increases proportionally from 60 kg at age 20 to 85 kg by age 30.
- Conclusion: The graphs indicate that age is a significant predictor of both height and weight in this dataset. The near-perfect alignment of the data points suggests a very stable and predictable growth pattern over the observed decade

```python
df_market = pd.DataFrame({
    "brand": ["Samsung", "Apple", "Xiaomi", "Other"],
    "share": [35, 30, 25, 10]
})
df_market
```

```python
brands_sum = df_market.groupby('brand')['share'].sum()
brands_sum
```

```python
labels = ['apple', 'other', 'samsung', 'xiaomi']
plt.pie(brands_sum, shadow=True, autopct='%1.1f%%', labels=labels)
plt.show()
```

The pie chart visualizes the market share distribution among different brands:
- Market Leader: Samsung holds the largest portion of the market at 35.0%.
- Runner-up: Apple follows closely with a 30.0% market share.
- Xiaomi: This brand controls exactly a quarter of the market at 25.0%.
- Others: All other minor brands combined account for the remaining 10.0%.
- Conclusion: The chart highlights that Samsung and Apple are the dominant players, collectively controlling 65% of the market. Xiaomi maintains a strong competitive position, while other brands represent a smaller segment of the industry
