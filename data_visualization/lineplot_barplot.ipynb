```python
df_energy = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Electricity': [320, 300, 350, 400, 420, 450],
    'Gas': [200, 180, 210, 230, 240, 260]
})
df_energy
```

```python
plt.plot(df_energy['Month'], df_energy['Electricity'], linestyle=":", marker='o', color='r')
plt.plot(df_energy['Month'], df_energy['Gas'], color='b', marker='v')
plt.xlabel('months')
plt.ylabel('energy')
plt.title('Monthly Gas and Electricity Consumption Trends')
plt.legend(['electricity', 'gas'], loc=0)
plt.show()
```

- Both energy types (electricity and gas) show a significant upward trend compared to the beginning of the year, with a steady increase starting from March
- Electricity consumption (red dotted line) consistently remains higher than gas. It reached its peak in June at 450 units
- A slight decrease in both resources was observed in February, likely due to fewer days in the month or temporary mild weather
- The sharp rise in electricity between April and June suggests an increased use of cooling systems (air conditioning) as the summer season approached
- Gas consumption (blue line) grew more slowly and steadily, with the June figure (260 units) being approximately 30% higher than in January

```python
df_salary = pd.DataFrame({
    'Profession': ['Teacher', 'Engineer', 'Doctor', 'Designer', 'Programmer'],
    'AverageSalary': [600, 1200, 1500, 1000, 1800]
})
df_salary
```

```python
plt.bar(df_salary['Profession'], df_salary['AverageSalary'], color=['#ced4da', '#ced4da', '#ced4da', '#ced4da', '#2a9d8f'])
plt.xlabel('profession')
plt.ylabel('salary')
plt.title('Average Salary by Profession')
plt.show()
```

- Programmers earn the highest average salary, reaching approximately 1,800 units, significantly outperforming other professions
- Doctors follow in second place with a substantial average salary of 1,500 units
- Engineers and Designers earn moderate salaries, at 1,200 and 1,000 units respectively
- Teachers have the lowest average salary among the group, at around 600 units, which is three times less than that of programmers
- The use of a distinct teal color for the "Programmer" bar effectively emphasizes the leading profession in this dataset
- *There is a clear salary gap between tech-oriented roles and traditional public service roles like teaching*

```python
df_orders = pd.DataFrame({
    'Day': [1, 2, 3, 4, 5, 6, 7],
    'Orders': [45, 50, 47, 60, 65, 70, 68]
})
df_orders
```

```python
plt.plot(df_orders['Day'], df_orders['Orders'], linewidth=0.7, color='#FC5A50', marker='o')
plt.xlabel('days')
plt.ylabel('orders')
plt.title('Daily Orders')
plt.show()
```

- There is a clear upward trend in the number of orders throughout the week, growing from 45 orders on Day 1 to 68 orders by Day 7
- The highest volume of orders occurred on Day 6, reaching a peak of 70 orders
- A slight dip is observed between Day 2 and Day 3 (dropping from 50 to 47 orders), followed by a significant and steady recovery
- The most rapid growth phase took place between Day 3 and Day 6, indicating high activity during the second half of the week

```python
df_population = pd.DataFrame({
    'City': ['Tashkent', 'Samarkand', 'Bukhara', 'Andijan'],
    'Population': [3000000, 550000, 280000, 450000]
})
df_population
```

```python
# eng katta qiymat yashil rangda , kichiklari qizil rangda
colors = ['#00FF00' if x==max(df_population['Population']) else '#C20078' for x in df_population['Population']]

# grafik chizish
plt.bar(df_population['City'], df_population['Population'], width=0.7, color=colors)
plt.xlabel('cities')
plt.ylabel('population')
plt.title('Population of Cities')
plt.show()
```

- Tashkent stands as the clear leader in terms of population, significantly surpassing the other cities with over 3 million inhabitants
- Samarkand and Andijan show relatively similar population figures, both hovering around the 500,000 mark
- Among the cities presented, Bukhara has the lowest population, estimated at nearly 300,000

```python
df_temperature_year = pd.DataFrame({
    'Year': [2019, 2020, 2021, 2022, 2023],
    'AvgTemperature': [14.5, 14.8, 15.2, 15.6, 16.0]
})
df_temperature_year
```

```python
plt.plot(df_temperature_year['Year'], df_temperature_year['AvgTemperature'], linestyle='--', linewidth=2.5, marker='o', color='#029386')
plt.xticks(df_temperature_year['Year'])
plt.xlabel('years')
plt.ylabel('temperature')
plt.title('Average Temperature Over the Years')
plt.show()
```

- The chart illustrates a steady and sharp increase in average temperature over the years
- By 2023, the temperature reached its maximum peak of 16.0C
- This represents a significant rise compared to the 2019 starting point of 14.5C, highlighting a clear upward trajectory
- The average temperature for this specific period (15.2C) was observed during the year 2021

```python
df_sales_category = pd.DataFrame({
    'Category': ['Electronics', 'Clothing', 'Food', 'Books'],
    'Sales': [12000, 8000, 15000, 4000]
})
df_sales_category
```

```python
colors = ['C0', 'C1', 'C2', 'C3']
plt.bar(df_sales_category['Category'], df_sales_category['Sales'], color=colors, width=0.7)
plt.xlabel('categories')
plt.ylabel('sales')
plt.title('Sales by Category')
plt.show()
```

- The "Food" category dominates the sales performance, reaching a peak of over 15,000 units
- "Electronics" follows as the second most successful category, with a solid performance of 12,000 units
- The "Clothing" sector shows a moderate demand, accounting for approximately 8,000 units in sales
- "Books" recorded the lowest sales volume among all categories, with only 4,000 units sold
