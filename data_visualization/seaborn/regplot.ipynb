This page teaches regplot. The necessary DFs are created using AI. The conclusions of the graphs are improved using AI.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(0)

df = pd.DataFrame({
    "study_hours": np.random.uniform(1, 10, 100),
    "exam_score": np.random.uniform(50, 90, 100)
})
df
```

```python
sns.regplot(data=df, x='study_hours', y='exam_score')
plt.title('Study Hours vs Exam Score')
plt.show()
```

- The data points are highly scattered with no clear pattern, and the regression line is nearly flat. This indicates a lack of strong linear correlation between study hours and exam scores in this specific dataset
- The line shows a very slight downward tilt, suggesting a weak negative trend. Effectively, this means that increasing study hours does not necessarily lead to higher exam scores here
- There is high variance in the scores regardless of the hours studied. For example, students studying for 2 hours can score as high as 85 or as low as 52, and the same applies to those studying nearly 10 hours

```python
df2 = pd.DataFrame({
    "ad_spend": np.random.uniform(100, 5000, 200),
    "sales": np.random.uniform(2000, 15000, 200)
})
df2
```

```python
sns.regplot(data=df2, x='ad_spend', y='sales')
plt.title('Ad Spend vs Sales')
plt.show()
```

- The regression line is nearly flat, indicating that there is no significant linear relationship between advertising spend and sales volume in this dataset

```python
np.random.seed(1)

df3 = pd.DataFrame({
    "hours": np.random.randint(1, 10, 80),
    "score": np.random.randint(40, 100, 80)
})
df3
```

```python
sns.regplot(data=df3, x='hours', y='score')
plt.title('Hours of Study vs Exam Score')
plt.show()
```

- There is no significant linear relationship between study hours and exam scores.
- There are students who studied for 1-2 hours and scored 85+ points, as well as those who studied for 8-10 hours and scored around 50-60 points.
