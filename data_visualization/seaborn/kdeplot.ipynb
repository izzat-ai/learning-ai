```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(0)

df = pd.DataFrame({
    "salary": np.random.normal(2500, 600, 500)
})
df
```

```python
sns.kdeplot(df['salary'], fill=True)
plt.title("Salary Distribution")
plt.show()
```

- The most frequent salary is approximately 2400â€“2500 units
- The bulk of the employees or data points fall within the 1500 to 3500 range

```python
df2 = pd.DataFrame({
    "order_amount": np.random.gamma(2, 150, 700),
    "customer_type": np.random.choice(["New", "Returning"], 700)
})
df2
```

```python
sns.kdeplot(data=df2, x='order_amount', hue='customer_type', fill=True)
plt.title("Order Amount Distribution")
plt.show()
```

- Both New and Returning customers follow a right-skewed distribution. This is typical for e-commerce, where most orders are of low value, with a "long tail" of high-value transactions
- Interestingly, there's a slight bump in the Returning customer density around 700-800 units, suggesting a segment of loyal customers who place higher-than-average value orders compared to first-time buyers

```python
df3 = pd.DataFrame({
    "salary": np.random.normal(3000, 700, 600),
    "department": np.random.choice(["IT", "Sales"], 600)
})
df3
```

```python
sns.kdeplot(data=df3, x='salary', hue='department', fill=True)
plt.title("Salary Distribution by Department")
plt.show()
```

- Both departments follow a Normal Distribution. The overall salary structures are healthy, with most employees earning around the departmental average
- IT (Blue): The peak density is centered around 3200â€“3300 units. IT professionals generally have a higher modal salary compared to the Sales team
- Sales (Orange): The peak is shifted slightly to the left, around 2800â€“2900 units. This indicates that the most common salary in Sales is lower than in IT

```python
df4 = pd.DataFrame({
    "prediction_score": np.concatenate([
        np.random.beta(2, 6, 400),
        np.random.beta(6, 2, 400)
    ]),
    "dataset": ["Train"]*400 + ["Production"]*400
})
df4
```

```python
sns.kdeplot(data=df4, x='prediction_score', hue='dataset', fill=True)
plt.title("Prediction Score Distribution")
plt.show()
```

- The plot reveals a massive shift between the Train (blue) and Production (orange) distributions. The model was trained on data that mostly yielded scores between 0.1 and 0.3, yet in production, it is producing scores between 0.7 and 0.9
- The lack of overlap indicates that the model is operating in an environment it was not trained for. This is a classic example of model performance degradation, where predictions in production are likely inaccurate

```python
df5 = pd.DataFrame({
    "watch_time": np.random.gamma(2, 30, 700),
    "user_type": np.random.choice(["Free", "Premium"], 700)
})
df5
```

```python
sns.kdeplot(data=df5, x='watch_time', hue='user_type', fill=True)
plt.title("Watch Time Distribution")
plt.show()
```

- The watch time for both Free and Premium users follows a right-skewed distribution. Most users watch for a short duration, while only a small percentage are "power users" with very high watch times
- The "Long Tail" of the distribution is more populated by Premium users. This suggests that the subscription model effectively encourages longer viewing sessions and deeper platform engagement
