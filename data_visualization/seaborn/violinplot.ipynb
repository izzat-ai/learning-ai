```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(0)

df = pd.DataFrame({
    "salary": np.random.normal(3000, 700, 400),
    "department": np.random.choice(["IT", "Sales"], 400)
})
df
```

```python
sns.violinplot(data=df, x='salary', y='department')
plt.title('Salary Distribution by Department')
plt.show()
```

- Sales: The widest part of the violin is between 2500 and 3000, indicating where most employee salaries are clustered
- IT: The distribution is bulkier around the 3000â€“3500 range, suggesting a more standardized pay structure for the majority of the team
- The thicker internal bar (representing the middle 50% of data) is slightly more spread out in Sales, indicating a higher variance in pay compared to IT

```python
np.random.seed(1)

df1 = pd.DataFrame({
    "monthly_income": np.random.normal(3500, 800, 500),
    "department": np.random.choice(["IT", "Marketing", "Sales"], 500)
})
df1
```

```python
sns.violinplot(data=df1, x='monthly_income', y='department')
plt.title('Monthly Income Distribution by Department')
plt.show()
```

- Marketing holds the highest median income at approximately 3500â€“3600 units
- Sales follows with a median near 3400 units
- IT shows the lowest median among the three, centered around 3100â€“3200 units

```python
df2 = pd.DataFrame({
    "watch_time": np.random.gamma(2, 25, 600),
    "subscription": np.random.choice(["Free", "Premium"], 600)
})
df2
```

```python
sns.violinplot(data=df2, x='watch_time', y='subscription')
plt.title('Watch Time Distribution by Subscription Type')
plt.show()
```

- The median watch time for both Free and Premium subscribers is nearly identical, hovering around 35-40 minutes
- Free subscribers show a very high concentration in the lower time range (20â€“60 minutes), evidenced by the bulging base of the violin.
- Premium subscribers have a slightly more distributed engagement pattern, with a wider "neck" extending into the higher watch time categories

```python
df3 = pd.DataFrame({
    "delivery_time": np.random.normal(40, 10, 550),
    "service_type": np.random.choice(["Standard", "Express"], 550)
})
df3
```

```python
sns.violinplot(data=df3, x='delivery_time', y='service_type')
plt.title('Delivery Time Distribution by Service Type')
plt.show()
```

- The median delivery time for the Express service is approximately 40 minutes
- The Standard service has a slightly higher median, centered around 42-43 minutes
- Express shows a tighter, more symmetric distribution around the median, indicating a more predictable and reliable delivery window
- Standard delivery has a wider spread, implying higher variability in delivery performance
