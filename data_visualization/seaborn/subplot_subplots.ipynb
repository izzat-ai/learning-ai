```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# onlayn do'kon ma'lumotlari
shop = pd.DataFrame({
    "month":["Jan","Feb","Mar","Apr","May","Jun"],
    "sales":[12000,13500,15000,17000,18200,19500],
    "orders":[210,240,265,290,310,340]
})
shop
```

```python
plt.figure(figsize=(12, 6))

plt.title("Onlayn do'kon ma'lumotlari")

plt.subplot(1, 2, 1)
sns.lineplot(data=shop, x='month', y='sales')
plt.xlabel("oylar", fontsize=11, labelpad=8)

plt.subplot(1, 2, 2)
sns.barplot(data=shop, x='month', y='orders')
plt.xlabel("oylar", fontsize=11, labelpad=8)

plt.show()
```

- Birinchi grafikda har oygi bo'lgan savodlarni ko'rishimiz mumkin . Do'konda bo'lgan savo oyma-oy o'sib borgan va Iyun oyida eng yuqori natijani qayd etgan
- Ikkinchi grafikda esa oylar bo'yicha buyurtmalar oshganini ko'rishimiz mumkin . Bu ham huddi savo oshgani sari yuqorilab bormoqda .

###Subplots

```python
# KPI dashboard
company = pd.DataFrame({
    "month":["Jan","Feb","Mar","Apr","May","Jun"],
    "sales":[12000,14000,15500,17000,18500,20000],
    "profit":[2200,2600,2800,3100,3500,3900],
    "customers":[180,200,225,250,270,300]
})
company
```

```python
fig, ax = plt.subplots(2, 2, figsize=(14, 8))
sns.lineplot(ax=ax[0, 0], data=company, x='month', y='sales')
ax[0, 0].set_title("Savdo")

sns.scatterplot(ax=ax[0, 1], data=company, x='sales', y='profit')
ax[0, 1].set_title("Savdo va foyda")
plt.tight_layout()

sns.barplot(ax=ax[1, 0], data=company, x='month', y='profit')
ax[1, 0].set_title("Oylar bo'yicha foyda")

sns.histplot(ax=ax[1, 1], data=company, x='customers')
ax[1, 1].set_title("Mijozlar taqsimoti")

plt.show()
```

- Birinchi grafikda oylar bo'yicha savdoning o'sib borishini ko'rish mumkin
- Ikkinchi grafikda , savdo oshgani sari foyda ham shunga mos ravishda oshadi
- Uchinchi grafikda Savdo kabi foyda ham har oy barqaror ravishda o'sib borgan
- To'rtinchi grafik oylar kesimida mijozlar soni asosan ikki xil oraliqda ko'proq takrorlangan
