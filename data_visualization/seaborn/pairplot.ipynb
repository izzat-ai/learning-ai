```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame({
    "age": np.random.randint(18, 60, 120),
    "salary": np.random.randint(300, 3000, 120),
    "experience": np.random.randint(0, 20, 120),
    "gender": np.random.choice(["Male", "Female"], 120)
})
df
```

```python
sns.pairplot(df, hue='gender')
plt.show()
```

Diagonal Plots (KDE - Distribution):

- Age: Both genders follow a fairly uniform distribution between 20 and 60 years. There is a slight peak for females around their late 40s.

- Salary: The pay range is broad (500 to 3000). Females show a more pronounced peak around the 1500 mark, whereas male salaries are more evenly spread.

- Experience: For males, there is a clear peak around 12â€“15 years of experience.

- 2. Scatter Plots (Correlation):

Salary vs. Experience: Surprisingly, there is no strong linear correlation between experience and salary. The data points are highly dispersed, suggesting that in this dataset, more experience does not automatically guarantee a higher salary.

Age vs. Salary: Similarly, age does not appear to be a primary predictor of income. High and low earners are found across all age groups.

- 3. Gender Insights:

The overlap between blue (Male) and orange (Female) points is significant. This indicates that gender does not form distinct clusters or "islands" in terms of how age, salary, and experience interact. The demographic groups are treated or behave similarly within these variables

```python
np.random.seed(42)

df1 = pd.DataFrame({
    "age": np.random.randint(20, 65, 150),
    "income": np.random.randint(1500, 9000, 150),
    "experience": np.random.randint(0, 40, 150),
    "education_level": np.random.choice(
        ["High School", "Bachelor", "Master"], 150
    )
})
df1
```

```python
df1['education_level'].unique()
```

```python
sns.pairplot(df1, hue='education_level')
plt.show()
```

- Diagonal Plots (KDE - Distribution):

Age: The High School group (orange) tends to be older, peaking around 60. Bachelor (blue) and Master (green) graduates are generally younger, peaking around 40.

Income: The High School group has a distinct income peak around 5000â€“6000. Bachelor and Master holders show a more distributed income range, suggesting more variability in their earnings.

Experience: High School graduates show the highest density in long-term experience (peaking near 30 years), which correlates with their higher age.

- 2. Scatter Plots (Correlation):

Age vs. Experience: There is a visible positive correlation between age and experience. As expected, older individuals in the dataset tend to have more years of work experience.

Income vs. Experience: There is no strong linear correlation between experience and income. The dispersion of points suggests that having more years of experience doesn't automatically result in a significantly higher salary in this specific population.

- 3. Education Insights:

The significant overlap between the three colors indicates that education level is not a strict separator for income or experience. While there are distributional differences, the groups are not distinct enough to form isolated clusters, suggesting other factors (like industry or role) might play a larger role in determining income

```python
np.random.seed(7)

df2 = pd.DataFrame({
    "area": np.random.randint(40, 220, 180),
    "rooms": np.random.randint(1, 6, 180),
    "distance_center": np.random.randint(1, 35, 180),
    "price_category": np.random.choice(
        ["Low", "Medium", "High"], 180
    )
})
df2
```

```python
df2['price_category'].unique()
```

```python
sns.pairplot(df2, hue='price_category')
plt.show()
```

Diagonal Plots (KDE - Distribution):

- Area: The distributions for all price categories overlap significantly. While High price homes (orange) show a slightly wider range, there's no sharp distinction based on area alone.

- Rooms: A clearer trend is visible here. High price categories (orange) peak around 4 and 5 rooms, while Low price homes (blue) are more frequently associated with 1 or 2 rooms.

- Distance_center: The density plots for distance from the center are nearly identical across all price tiers, suggesting that distance alone doesn't cleanly separate these price segments.

2. Scatter Plots (Correlation):

- Area vs. Rooms: There is a clear positive correlation. As the number of rooms increases, the total area typically increases as well.

- Distance_center vs. Others: There is no visible correlation between distance from the center and house size or number of rooms. Points are scattered randomly across the grid.

- Price Overlap: The significant mixing of colors (Blue, Orange, Green) across all scatter plots indicates that these three features together are not sufficient to perfectly classify the price of a house. The classes are not linearly separable in this 3D feature space

```python
np.random.seed(21)

df3 = pd.DataFrame({
    "tenure_months": np.random.randint(1, 72, 200),
    "monthly_charges": np.random.randint(20, 130, 200),
    "total_charges": np.random.randint(100, 9000, 200),
    "churn": np.random.choice(["Yes", "No"], 200)
})
df3
```

```python
df3['churn'].unique()
```

```python
sns.pairplot(df3, hue='churn')
plt.show()
```

Diagonal Plots (Distributions):

- Tenure Months: Non-churned customers (blue) show a higher density at longer tenures. Churned customers (orange) tend to drop off earlier in their lifecycle.

- Monthly Charges: Churned customers seem to have a slightly higher density in the higher monthly billing range, suggesting that high costs might be a factor for attrition.

- Total Charges: Both distributions are right-skewed, but non-churners reach higher total values due to their longer loyalty.2. Scatter Plots (Correlations):Tenure vs.

- Total Charges: There is a very strong positive correlation here. This is expected as $Total \approx Tenure \times Monthly$.Monthly vs. Total Charges: A positive "fan-shaped" correlation is visible, reflecting that total spend increases with monthly price but varies based on how long the customer stayed.

- Churn Insights: Orange points (churn) are scattered throughout but are more concentrated in the lower tenure and higher monthly charge zones
