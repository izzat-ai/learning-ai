**This page will explore displot. DataFrames are created using AI.**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)
ages = np.random.randint(18, 65, 200)

df = pd.DataFrame({"age": ages})
df
```

```python
sns.displot(df['age'], kde=True)
plt.title("Distribution of Age")
plt.show()
```

```python
salaries = np.random.normal(800, 150, 300)
df_salary = pd.DataFrame({"salary": salaries})
df_salary
```

```python
sns.displot(df_salary, x='salary', kde=True)
plt.title("Distribution of Salaries")
plt.show()
```

- The data ranges from 400 to 1200 units
- Most observations are concentrated between 650 and 950.
- Extreme values (below 500 or above 1100) are rare, indicating few outliers in this specific dataset

```python
credit = pd.DataFrame({
    "loan_amount": np.random.gamma(2, 200, 300),
    "default": np.random.choice(["Yes", "No"], 300)
})
credit
```

```python
sns.displot(credit, x='loan_amount', hue='default', kde=True)
plt.title("Distribution of Loan Amounts")
plt.show()
```

- These customers represent the majority across almost all price points. The peak density is observed around the 200–300 range.
- Interestingly, defaults are also concentrated in the smaller loan brackets. The frequency of defaults drops significantly as the loan amount increases beyond 800.
- There are very few orange bars/lines beyond the 1000 mark, suggesting that high-value loans are either less common or subject to stricter vetting, resulting in fewer defaults

```python
students = pd.DataFrame({
    "score": np.random.normal(70, 10, 200),
    "class": np.random.choice(["A", "B"], 200)
})
students
```

```python
sns.displot(students, x='score', hue='class', multiple='stack', kde=True)
plt.title("Distribution of Scores")
plt.show()
```

- The peak (mode) for both classes is centered around the 70-75 score range. This suggests that the average performance level is consistent across both groups.
- Class B (blue) has a much larger sample size than Class A (orange). This is evident as the blue portion of the bars is consistently taller across the entire score range

```python
df2 = pd.DataFrame({
    "salary": np.random.normal(900, 180, 400),
    "department": np.random.choice(["IT", "Sales", "HR"], 400)
})
df2
```

```python
sns.displot(df2, x='salary', hue='department', kde=True)
plt.title("Distribution of Salaries by Department")
plt.show()
```

- The salaries across all three departments follow a Normal Distribution. The bell-shaped curves (KDE lines) are smooth and symmetrical, indicating a well-structured pay scale
- This department has the largest headcount. The distribution starts at a lower salary point (around 400) compared to others, with a peak around 850.
- The IT distribution is shifted to the right. Its peak is around the 950–1000 range, indicating a higher average salary compared to Sales and HR.
- Sales sits in the middle ground, with its mode centered around 900 units

```python
df3 = pd.DataFrame({
    "purchase_amount": np.random.gamma(2, 250, 350),
    "customer_type": np.random.choice(["New", "Returning"], 350)
})
df3
```

```python
sns.displot(df3, x='purchase_amount', hue='customer_type', kde=True)
plt.title("Distribution of Purchase Amounts by Customer Type")
plt.show()
```

- where the majority of transactions are small, with a few high-value purchases forming a long tail on the right
- This group shows a higher frequency of purchases across almost all price brackets. Their distribution peak is slightly broader, indicating a more established spending pattern.
- These customers are heavily concentrated in the lower transaction tiers (200–500 units) and drop off faster as the price increases

```python

df_1 = pd.DataFrame({
    "salary": np.random.normal(2000, 500, 300)
})
df_1
```

```python
# ecdf grafigini chizish
sns.displot(df_1, x="salary", kind="ecdf")
plt.show()
```

```python
df_2 = pd.DataFrame({
    "purchase_amount": np.random.gamma(2, 150, 500),
    "customer_type": np.random.choice(["New", "Returning"], 500)
})
df_2
```

```python
sns.displot(df_2, x="purchase_amount", hue="customer_type", kind="ecdf")
plt.show()
```

- The purchasing behavior of both "Returning" and "New" customers is remarkably similar. The ECDF curves overlap for the majority of the distribution, indicating that their spending patterns are nearly identical
- In both groups, 50% of the customers spend approximately 250 units or less
- About 80% of transactions for both segments fall below the 500-unit threshold
- The steep slope between 0 and 500 units suggests that this is the primary "active zone" for sales. Marketing efforts should focus on moving customers from this 250–500 bracket into higher spending tiers

```python
np.random.seed(42)

df4 = pd.DataFrame({
    "exam_score": np.random.normal(65, 10, 300),
    "class": np.random.choice(["A", "B"], 300)
})
df4
```

```python
sns.displot(data=df4, x='exam_score', hue='class', kind='ecdf')
plt.title("Distribution of Exam Scores by Class")
plt.show()
```

- Both classes have a median score of approximately 65–67. Half of the students in each class scored below this mark
- About 80% of students in both groups scored below 75 units
- In the high-score region (above 80), Class B (orange line) extends further to the right, reaching up to 100. This suggests that while both classes are similar on average, Class B contains a few individual outliers with exceptionally high scores

```python
df5 = pd.DataFrame({
    "delivery_time": np.random.gamma(2, 20, 400),
    "service": np.random.choice(["Standard", "Express"], 400)
})
df5
```

```python
sns.displot(data=df5, x='delivery_time', hue='service', kde='ecdf')
plt.title("Distribution of Delivery Times by Service Type")
plt.show()
```

- Both service types exhibit a right-skewed (positive skew) distribution. Most deliveries are completed quickly, but there is a long tail of delayed deliveries extending beyond 100 units
- The peak density is shifted towards the left, centered around 20–25 units, confirming it is generally faster
- The mode is slightly higher, around 30 units. The orange bars are also more prominent in the 40–60 unit range compared to Express

```python
df6 = pd.DataFrame({
    "session_duration": np.concatenate([
        np.random.normal(5, 1.2, 250),
        np.random.normal(7, 1.5, 250)
    ]),
    "version": ["Old"]*250 + ["New"]*250
})
df6
```

```python
sns.displot(data=df6, x='session_duration', hue='version', kde='ecdf')
plt.title("Distribution of Session Durations by Version")
plt.show()
```

- The peak density is centered around 5 units of time. Most users spend between 3 and 7 minutes per session
- The entire distribution is shifted to the right, with a mode around 7-7.5 units. This indicates a significant improvement in session duration
