**This page will explore displot. DataFrames are created using AI.**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)
ages = np.random.randint(18, 65, 200)

df = pd.DataFrame({"age": ages})
df
```

```python
sns.displot(df['age'], kde=True)
plt.title("Distribution of Age")
plt.show()
```

```python
salaries = np.random.normal(800, 150, 300)
df_salary = pd.DataFrame({"salary": salaries})
df_salary
```

```python
sns.displot(df_salary, x='salary', kde=True)
plt.title("Distribution of Salaries")
plt.show()
```

- The data ranges from 400 to 1200 units
- Most observations are concentrated between 650 and 950.
- Extreme values (below 500 or above 1100) are rare, indicating few outliers in this specific dataset

```python
credit = pd.DataFrame({
    "loan_amount": np.random.gamma(2, 200, 300),
    "default": np.random.choice(["Yes", "No"], 300)
})
credit
```

```python
sns.displot(credit, x='loan_amount', hue='default', kde=True)
plt.title("Distribution of Loan Amounts")
plt.show()
```

- These customers represent the majority across almost all price points. The peak density is observed around the 200–300 range.
- Interestingly, defaults are also concentrated in the smaller loan brackets. The frequency of defaults drops significantly as the loan amount increases beyond 800.
- There are very few orange bars/lines beyond the 1000 mark, suggesting that high-value loans are either less common or subject to stricter vetting, resulting in fewer defaults

```python
students = pd.DataFrame({
    "score": np.random.normal(70, 10, 200),
    "class": np.random.choice(["A", "B"], 200)
})
students
```

```python
sns.displot(students, x='score', hue='class', multiple='stack', kde=True)
plt.title("Distribution of Scores")
plt.show()
```

- The peak (mode) for both classes is centered around the 70-75 score range. This suggests that the average performance level is consistent across both groups.
- Class B (blue) has a much larger sample size than Class A (orange). This is evident as the blue portion of the bars is consistently taller across the entire score range

```python
np.random.seed(1)

df1 = pd.DataFrame({
    "age": np.random.randint(18, 70, 300)
})
df1
```

```python
sns.displot(df1['age'], kde=True)
#plt.title
plt.show()
```

- Young Adults (25–35): The first significant cluster is seen here, representing a large portion of the sample.
- 40–45: Another peak occurs in this range.
- 65–70: interestingly, there is a late surge in count towards the end of the age range, indicating a strong presence of older individuals in the dataset

```python
df2 = pd.DataFrame({
    "salary": np.random.normal(900, 180, 400),
    "department": np.random.choice(["IT", "Sales", "HR"], 400)
})
df2
```

```python
sns.displot(df2, x='salary', hue='department', kde=True)
plt.title("Distribution of Salaries by Department")
plt.show()
```

- The salaries across all three departments follow a Normal Distribution. The bell-shaped curves (KDE lines) are smooth and symmetrical, indicating a well-structured pay scale
- This department has the largest headcount. The distribution starts at a lower salary point (around 400) compared to others, with a peak around 850.
- The IT distribution is shifted to the right. Its peak is around the 950–1000 range, indicating a higher average salary compared to Sales and HR.
- Sales sits in the middle ground, with its mode centered around 900 units

```python
df3 = pd.DataFrame({
    "purchase_amount": np.random.gamma(2, 250, 350),
    "customer_type": np.random.choice(["New", "Returning"], 350)
})
df3
```

```python
sns.displot(df3, x='purchase_amount', hue='customer_type', kde=True)
plt.title("Distribution of Purchase Amounts by Customer Type")
plt.show()
```

- where the majority of transactions are small, with a few high-value purchases forming a long tail on the right
- This group shows a higher frequency of purchases across almost all price brackets. Their distribution peak is slightly broader, indicating a more established spending pattern.
- These customers are heavily concentrated in the lower transaction tiers (200–500 units) and drop off faster as the price increases
