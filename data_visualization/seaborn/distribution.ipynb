```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

####Histplot

```python
# xodimlarning oylik maoshi
employees = pd.DataFrame({
    "salary": [
        700, 850, 900, 950, 1000,
        1100, 1200, 1250, 1300, 1400,
        1500, 1600, 1700, 1800, 2000,
        2200, 2400, 2600, 2800, 3000
    ]
})

employees.head()
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(data=employees, x='salary', bins=7, kde=True, color='skyblue')

plt.title('Salary Distribution')

plt.show()
```

- Xodimlarning aksari 700 dan 1300 gacha ko'p maosh oladi . Shunga qaramay tahminan 70-75% xodimlar 2000 dan kam maosh oladi

```python
# talabalarning imtihon ballari
students = pd.DataFrame({
    "score":[
        45,52,55,58,60,62,65,67,70,72,
        74,76,78,80,82,84,86,88,90,95
    ],
    "gender":[
        "Male","Female","Male","Female","Male",
        "Female","Male","Female","Male","Female",
        "Male","Female","Male","Female","Male",
        "Female","Male","Female","Male","Female"
    ]
})

students.head()
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(data=students, x='score', hue='gender', bins=8, kde=True)

plt.show()
```

- Qiz talabalar va o'gil bola talabalarning ma'lumotlari deyarli bir xil . Ikkala guruhning eng yuqori bali tahminan 75 ball bo'lgan
- 55 dan past ball olganlar faqat o'g'il bola talabalar hisoblanadi . Lekin bunday ball olganlar kam

####Displot

```python
# uy narxlari
houses = pd.DataFrame({
    "price":[
        90,95,100,110,120,130,
        140,150,160,170,180,190,
        200,210,220,240,260,280
    ],
    "city":[
        "Tashkent","Samarkand","Tashkent","Samarkand",
        "Tashkent","Samarkand","Tashkent","Samarkand",
        "Tashkent","Samarkand","Tashkent","Samarkand",
        "Tashkent","Samarkand","Tashkent","Samarkand",
        "Tashkent","Samarkand"
    ]
})

houses.head()
```

```python
sns.displot(data=houses, x='price', hue='city', bins=9, kde=True, height=5, aspect=2, multiple="dodge")
plt.show()
```

- Toshkentdagi uylar narxi pastki diapazonda to'plangan . Narx oshgan sari uylar kamaygan
- Eng yuqori narxlar esa Samarqandda kuzatilgan
