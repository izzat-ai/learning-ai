```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# xodimlar maoshi
df_salary = pd.DataFrame({
    "salary": np.random.normal(loc=8_000_000, scale=2_000_000, size=500).round(-3)
})
df_salary
```

```python
sns.histplot(data=df_salary, x=df_salary['salary'], bins=20, kde=True)
#plt.title("Maosh tarqoqligi grafigi")
plt.show()
```

```python
# talabalar bali
df_scores = pd.DataFrame({
    "score": np.clip(np.random.normal(70, 12, 400), 0, 100)
})
df_scores
```

```python
sns.kdeplot(data=df_scores, x=df_scores['score'], fill=True)
plt.show()
```
