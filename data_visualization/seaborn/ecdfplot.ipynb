**This page will explore ecdfplot. DataFrames are created using AI and graphical analysis is also being improved using AI.**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(1)

df_students = pd.DataFrame({
    "score": np.random.normal(70, 10, 120).clip(30, 100),
    "group": np.random.choice(["Online", "Offline"], 120)
})
df_students
```

```python
sns.ecdfplot(x="score", hue="group", data=df_students)
plt.show()
```

- According to the graph, the Online group is more likely to score high
- At almost all score points (60 to 85), the Online group line (blue) is located to the right and lower than the Offline group line (orange), which means that members of this group scored higher
- The average (median) score of the Online group is approximately 4 points higher than the Offline group

```python
np.random.seed(2)

df_salary = pd.DataFrame({
    "salary": np.concatenate([
        np.random.normal(3000, 500, 100),
        np.random.normal(4500, 700, 100)
    ]),
    "department": ["Support"]*100 + ["Engineering"]*100
})
df_salary
```

```python
sns.ecdfplot(data=df_salary, x='salary', hue='department')
plt.show()
```

- Salaries are significantly higher in the Engineering department
- There are many people in the Support department who earn less than 4,000
- At the 4000 point, the line is just beginning (at about 0.2 or 20% of the share). That is, only 20% of this section earns less than 4000

```python
np.random.seed(3)

df_delivery = pd.DataFrame({
    "time_minutes": np.concatenate([
        np.random.exponential(30, 150),
        np.random.exponential(20, 150)
    ]),
    "city": ["Region"]*150 + ["Tashkent"]*150
})
df_delivery
```

```python
sns.ecdfplot(data=df_delivery, x='time_minutes', hue='city')
plt.title("Delivery time")
plt.show()
```

- Delivery service in Tashkent is faster
- The Tashkent line (orange) is to the left of the Region line (blue) almost throughout the entire graph, which means that orders arrive faster in Tashkent
- The share in 40 minutes has reached approximately 0.85-0.90 (85-90%). That is, almost 90% of orders in Tashkent are delivered within 40 minutes
- The share in 40 minutes is approximately 0.80 (80%). That is, 80% of orders in the regions are delivered in less than 40 minutes
