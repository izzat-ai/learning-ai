```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
np.random.seed(42) # Natija har safar bir xil chiqishi uchun

n = 200 # Ma'lumotlar soni
df_work = pd.DataFrame({
    'Department': np.random.choice(['IT', 'HR', 'Sales', 'Marketing'], n),
    'Salary': np.random.normal(5000, 1500, n), # O'rtacha 5000, og'ish 1500
    'Experience': np.random.randint(1, 21, n),
    'Gender': np.random.choice(['Male', 'Female'], n)
})

# Outlier qo'shamiz (Juda katta maosh oladigan bir nechta xodim)
outliers = pd.DataFrame({
    'Department': ['IT', 'Sales'],
    'Salary': [12000, 13500],
    'Experience': [25, 30],
    'Gender': ['Male', 'Female']
})

df_work = pd.concat([df_work, outliers], ignore_index=True)
df_work
```

```python
sns.boxplot(data=df_work, x='Salary')
plt.show()
```

- The Boxplot visualization of the Salary variable provides several key statistical insights:
- Median: The central vertical line within the box indicates that the median salary is approximately 5000.
- Interquartile Range (IQR): The box represents the middle 50% of the data, with the first quartile (Q1) at roughly 4000 and the third quartile (Q3) at 6000.
- Outliers: Individual circles located beyond the whiskers (above 9000 and one near 0) represent outliers. These are data points that deviate significantly from the rest of the distribution, likely reflecting the high-salary entries we manually added to the dataset.
- Data Spread: The whiskers extend to roughly 2000 and 8700, defining the boundaries within which most of the salary data points fall, excluding extreme values

```python
# kategoriya bo'yicha boxplot
sns.boxplot(data=df_work, x='Salary', y='Department')
plt.show()
```

- The visualization presents a horizontal boxplot comparing salary distributions across different departments.
- Median Salary: The Marketing department shows the highest median salary among all groups, while the HR department has the lowest median pay.
- Salary Variance (IQR): The Sales department exhibits the widest interquartile range (the width of the blue box), indicating a higher variance in pay among its staff compared to other departments.

Outliers:

- The Sales department has a significant outlier earning approximately 13,500.
- IT features two high-earning outliers around the 11,000 and 12,000 marks.
- Interestingly, HR has a low-end outlier with a salary near zero.
- Whiskers: The whiskers for IT and Sales are extended further toward the higher end, suggesting a wider overall range of "normal" salaries in these fields compared to HR or Marketing

```python
sns.boxplot(data=df_work, x='Salary', y='Department', hue='Gender')
plt.show()
```

- This visualization is a grouped horizontal boxplot, showcasing salary distributions categorized by both Department and Gender using the hue parameter.
- Inter-Departmental Comparison: On average, the Marketing department maintains a relatively higher and more consistent median salary compared to other sectors.

Gender-Based Insights:

- In Sales, while the median pay is nearly equal for both genders, there is a prominent female outlier earning a significantly higher salary (approx. 13,500).
- In IT, the median salary for males is slightly higher than for females. Furthermore, all high-earning outliers in this department are male.
- In HR, females have a slightly higher median salary than males, whereas the only low-end outlier (near zero) belongs to a male employee.
- Spread and Variance: The male salary distribution in Sales shows higher variance compared to females, as indicated by the wider interquartile range (blue box)
