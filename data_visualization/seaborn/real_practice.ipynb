```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# onlayn do'konning kunlik savdosi ma'lumotlari
sales = pd.DataFrame({
    "date": [
        "2025-05-01","2025-05-02","2025-05-03","2025-05-04",
        "2025-05-05","2025-05-06","2025-05-07","2025-05-08",
        "2025-05-09","2025-05-10","2025-05-11","2025-05-12",
        "2025-05-13","2025-05-14"
    ],
    "daily_sales": [
        4200, 4500, 4700, 5200,
        5000, 6100, 6400, 6000,
        6700, 7200, 7100, 7600,
        7900, 8300
    ]
})

sales["date"] = pd.to_datetime(sales["date"])

sales.head()
```

```python
plt.figure(figsize=(12, 6))
plt.title("Kunlik savdo hajmi")
plt.xlabel("Sana")
plt.ylabel("Savdo hajmi")
plt.grid()

sns.lineplot(x=sales['date'], y=sales['daily_sales'])

plt.show()
```

```python
# uchta mahsulotning oylik savdosi ma'lumotlari
products = pd.DataFrame({
    "month": [
        "Jan","Feb","Mar","Apr","May","Jun",
        "Jan","Feb","Mar","Apr","May","Jun",
        "Jan","Feb","Mar","Apr","May","Jun"
    ],
    "product": [
        "Laptop","Laptop","Laptop","Laptop","Laptop","Laptop",
        "Phone","Phone","Phone","Phone","Phone","Phone",
        "Tablet","Tablet","Tablet","Tablet","Tablet","Tablet"
    ],
    "sales": [
        120,135,150,170,180,210,
        220,240,260,250,290,320,
        80,90,95,100,110,125
    ]
})
products.shape
```

```python
products['product'].unique()
```

```python
plt.figure(figsize=(12, 6))
plt.title("Oylik savdo hajmi")
plt.xlabel("Oy")
plt.ylabel("Savdo hajmi")
plt.grid()

sns.lineplot(x=products['month'], y=products['sales'], hue=products['product'], marker='o', linewidth=3)

plt.show()
```
