#####**Lineplot**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# onlayn do'konning kunlik savdosi ma'lumotlari
sales = pd.DataFrame({
    "date": [
        "2025-05-01","2025-05-02","2025-05-03","2025-05-04",
        "2025-05-05","2025-05-06","2025-05-07","2025-05-08",
        "2025-05-09","2025-05-10","2025-05-11","2025-05-12",
        "2025-05-13","2025-05-14"
    ],
    "daily_sales": [
        4200, 4500, 4700, 5200,
        5000, 6100, 6400, 6000,
        6700, 7200, 7100, 7600,
        7900, 8300
    ]
})

sales["date"] = pd.to_datetime(sales["date"])

sales.head()
```

```python
plt.figure(figsize=(12, 6))
plt.title("Kunlik savdo hajmi")
plt.xlabel("Sana")
plt.ylabel("Savdo hajmi")
plt.grid()

sns.lineplot(x=sales['date'], y=sales['daily_sales'])

plt.show()
```

- Har kuni online do'konda savdo ko'paymoqda
- Har 3-4 kunda bir marta vaqtinchalik pasayishlar kuzatilgan
- Eng ko'p savdo davr ohirida , 14-mayda qayd etilgan .

```python
# uchta mahsulotning oylik savdosi ma'lumotlari
products = pd.DataFrame({
    "month": [
        "Jan","Feb","Mar","Apr","May","Jun",
        "Jan","Feb","Mar","Apr","May","Jun",
        "Jan","Feb","Mar","Apr","May","Jun"
    ],
    "product": [
        "Laptop","Laptop","Laptop","Laptop","Laptop","Laptop",
        "Phone","Phone","Phone","Phone","Phone","Phone",
        "Tablet","Tablet","Tablet","Tablet","Tablet","Tablet"
    ],
    "sales": [
        120,135,150,170,180,210,
        220,240,260,250,290,320,
        80,90,95,100,110,125
    ]
})
products.shape
```

```python
products['product'].unique()
```

```python
plt.figure(figsize=(12, 6))
plt.title("Oylik savdo hajmi")
plt.xlabel("Oy")
plt.ylabel("Savdo hajmi")
plt.grid()

sns.lineplot(x=products['month'], y=products['sales'], hue=products['product'], marker='o', linewidth=3)

plt.show()
```

####**Barplot**

```python
# bo'limlar bo'yicha o'rtacha oylik maosh ma'lumotlari
employees = pd.DataFrame({
    "department": [
        "IT", "IT", "IT",
        "HR", "HR", "HR",
        "Finance", "Finance", "Finance",
        "Marketing", "Marketing", "Marketing"
    ],
    "salary": [
        1800, 2100, 2000,
        1200, 1300, 1250,
        1700, 1750, 1850,
        1400, 1500, 1450
    ]
})
employees.shape
```

```python
employees.head()
```

```python
employees['department'].unique()
```

```python
plt.figure(figsize=(12, 6))
plt.title("O'rtacha oylik maosh")
plt.xlabel("Bo'limlar")
plt.ylabel("Maosh")
plt.grid()

sns.barplot(x=employees['department'], y=employees['salary'], hue=employees['department'], palette='viridis', legend=False)

plt.show()
```

####**Heatmap**

```python
# talabalarning turli fanlardan baholari va davomati ma'lumotlari
students = pd.DataFrame({
    "math": [78, 85, 90, 72, 88, 95, 67, 80, 91, 76],
    "physics": [75, 82, 91, 70, 86, 93, 65, 79, 89, 74],
    "chemistry": [80, 84, 89, 73, 87, 94, 68, 81, 90, 77],
    "english": [65, 70, 74, 68, 72, 78, 60, 69, 75, 67],
    "attendance": [90, 95, 98, 88, 96, 99, 85, 92, 97, 89]
})

students.head()
```

```python
students.corr()
```

```python
plt.figure(figsize=(12, 6))
plt.title("Talabalarning baholari va davomati")

sns.heatmap(students.corr(), annot=True, cmap='coolwarm', linewidths=3)

plt.show()
```

- aniq fanlardan birini yaxshi o'zlashtirgan talaba qolganlarini ham deyarli bir xil a'lo darajada o'zlashtirishini ko'rsatadi
- darsga kelish baho oshishining eng muhim omillaridan biridir

```python
# kompaniyaning turli hududlarda mahsulot sotuvlari ma'lumotlari
sales = pd.DataFrame({
    "Region": [
        "North","North","North",
        "South","South","South",
        "East","East","East",
        "West","West","West"
    ],
    "Product": [
        "Laptop","Phone","Tablet",
        "Laptop","Phone","Tablet",
        "Laptop","Phone","Tablet",
        "Laptop","Phone","Tablet"
    ],
    "Sales": [
        120,180,90,
        150,210,100,
        170,230,120,
        140,190,110
    ]
})

sales.shape
```

```python
sales.head()
```

```python
sales2 = pd.pivot_table(data=sales, index='Region', columns='Product', values='Sales')
sales2
```

```python
plt.figure(figsize=(12, 6))
plt.title("Hududlardagi sotuvlar")

sns.heatmap(sales2, annot=True, cmap='coolwarm', linewidths=3, fmt='.0f')

plt.show()
```

- Har bir hududda Telefon ko'p sotiladi . Planshetlar esa juda kam sotiladi
- Telefon eng ko'p sotiladigan hudud esa Sharq hisoblanadi , undan keyin Janub turadi


####**Scatterplot**

```python
# talabalarning o'qish natijalari ma'lumotlari
students = pd.DataFrame({
    "hours_studied":[8,12,15,18,20,22,10,14,16,19,24,11],
    "exam_score":[58,66,71,77,84,90,61,70,73,82,95,64],
    "gender":["Male","Female","Male","Female","Male","Female",
              "Male","Female","Male","Female","Male","Female"],
    "passed":["No","Yes","Yes","Yes","Yes","Yes",
              "No","Yes","Yes","Yes","Yes","Yes"]
})

students.shape
```

```python
students.head()
```

```python
students['passed'].unique()
```

```python
plt.figure(figsize=(10, 6))
plt.title("Talabalarning o'qish natijalari")
plt.xlabel("O'qish soati")
plt.ylabel("Imtihon natijasi")

sns.scatterplot(data=students, x='hours_studied', y='exam_score', hue='gender', style='passed')

plt.show()
```

- Talabalar qancha ko'p dars tayyorlashsa , imtihondan shunchalik yuqori ball olishadi . Buni grafikda kuchli ijobiy bog'liqlik borligidan bilishimiz mumkin
- Kam dars tayyorlagan , shu sababli past natija olib , imtihondan yiqilganlar ham bor . Va ular o'g'il bolalar
- Grafikdan 11-12 soat o'qib , minimum 63-65 ball bilan imtihondan o'tish mumkin . Imtihondan eng yuqori ball olgan talaba o'g'il bola , u 24 soat o'qib deyarli 100 ball olgan .

```python
# avtomobil ma'lumotlari
cars = pd.DataFrame({
    "engine_size":[1.2,1.5,1.6,1.8,2.0,2.2,2.5,3.0,1.4,1.7,2.4,3.2],
    "fuel_consumption":[5.5,6.2,6.8,7.3,8.1,8.6,9.4,10.5,5.9,6.9,8.9,11.2],
    "fuel_type":["Petrol","Petrol","Diesel","Diesel","Petrol","Diesel",
                 "Petrol","Petrol","Diesel","Petrol","Diesel","Petrol"],
    "transmission":["Manual","Automatic","Manual","Automatic",
                    "Automatic","Manual","Automatic","Automatic",
                    "Manual","Manual","Automatic","Automatic"]
})

cars.shape
```

```python
cars.head()
```

```python
plt.figure(figsize=(10, 6))
plt.title("Avtomobil ma'lumotlari")
plt.xlabel("Divigatel hajmi")
plt.ylabel("100 kmga yoqilg'i sarfi")

sns.scatterplot(data=cars, x='engine_size', y='fuel_consumption', hue='fuel_type', style='transmission')

plt.show()
```

- Divigatel hajmi qancha katta bo'lsa , yoqilg'i sarfi ham shunga yarasha ko'payib boradi
- Divigatel hajmi katta mashinalarni aksari benzinda yuradi . Hamda ular avtomat mashinalari hisoblanadi
- Divigatel hajmi kichkina bo'lgan mashinalarni ko'p qismi mexanik bo'ladi .
