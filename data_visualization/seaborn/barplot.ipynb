```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame({
    "gender": ["Male","Female","Male","Female","Male","Female"],
    "salary": [500, 650, 550, 700, 600, 680]
})
df
```

```python
sns.barplot(x='gender', y='salary', data=df)
plt.show()
```

```python
df2 = pd.DataFrame({
    "city": ["Tashkent","Tashkent","Samarkand","Samarkand","Tashkent","Samarkand"],
    "gender": ["Male","Female","Male","Female","Male","Female"],
    "salary": [700, 800, 600, 650, 750, 680]
})
df2
```

```python
sns.barplot(x='city', y='salary', hue='gender', data=df2)
plt.title("Salary by city and gender")
plt.show()
```

```python
df3 = pd.DataFrame({
    "campaign": ["A","A","B","B","C","C","A","B","C"],
    "revenue": [120,150,200,220,180,160,140,210,190]
})
df3
```

```python
sns.barplot(x='campaign', y='revenue', data=df3)
plt.title("Revenue by campaign")
plt.show()
```

```python
df4 = pd.DataFrame({
    "region": ["Center","Center","Suburb","Suburb","Rural","Rural"],
    "price": [200,220,150,160,90,100]
})
df4
```

```python
sns.barplot(x='region', y='price', data=df4, errorbar='sd')
plt.title("Price by region")
plt.show()
```

```python
df5 = pd.DataFrame({
    "department": ["IT","IT","HR","HR","Sales","Sales","IT","HR","Sales"],
    "salary": [1200,1300,800,850,900,950,1250,820,980]
})
df5
```

- *Which department has the highest average salary?*

```python
sns.barplot(x='department', y='salary', data=df5)
plt.title("Salary by department")
plt.show()
```

- The average salary in the IT department is much higher than in the HR and Sales departments. The average salary in the HR and Sales departments is not much different.

```python
df6 = pd.DataFrame({
    "city": ["Tashkent","Tashkent","Samarkand","Samarkand","Bukhara","Bukhara",
             "Tashkent","Samarkand","Bukhara"],
    "gender": ["Male","Female","Male","Female","Male","Female","Female","Male","Female"],
    "income": [600,750,500,650,450,600,720,520,580]
})
df6
```

- *Which city and gender has the highest average income?*

```python
sns.barplot(x='city', y='income', hue='gender', data=df6)
plt.title("Income by city and gender")
plt.show()
```

- The average income of men and women in Tashkent is slightly higher than the average incomes in Samarkand and Bukhara. In each city, women earn more than men. In Bukhara, the average incomes of men and women differ significantly.

```python
df7 = pd.DataFrame({
    "product": ["Phone","Phone","Laptop","Laptop","Tablet","Tablet",
                "Phone","Laptop","Tablet"],
    "region": ["North","South","North","South","North","South","North","South","North"],
    "sales": [200,180,300,280,150,140,220,310,160]
})
df7
```

- *Which product sells the most and in which region is the difference the biggest ?*

```python
sns.barplot(x='product', y='sales', hue='region', data=df7)
plt.title("Sales by product and region")
plt.show()
```

- We can see from this graph that laptops are the most sold devices compared to phones and tablets. In addition, laptop sales in the North and South are not very different.
