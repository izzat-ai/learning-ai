```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.DataFrame({
    "age": [18, 19, 20, 21, 22, 23, 24, 30, 35, 40]
})
df
```

```python
sns.histplot(data=df, x='age', kde=True)
plt.show()
```

```python
df2 = pd.DataFrame({
    "age": [20,22,24,26,28,30, 21,23,25,27,29,31],
    "gender": ["Male"]*6 + ["Female"]*6
})
df2
```

```python
sns.histplot(data=df2, x='age', hue='gender')
plt.show()
```

The histogram illustrates the age distribution by gender within a specific dataset.

Axes: The horizontal axis (X) represents age, while the vertical axis (Y) represents the frequency or count of individuals.

Age Range: The data covers individuals aged between 20 and 31 years.

Gender Comparison:

Age 2022 bin: Males (blue) are more frequent with a count of 2, compared to 1 female (orange).

Age 2931 bin: In this older category, females are more frequent with a count of 2, while males have a count of 1.

Mid-range (2228): The distribution is uniform across these bins, with typically 1 representative from each gender per category.

Conclusion: The population is slightly polarized, with higher counts at the youngest and oldest ends of the provided age spectrum, while the middle remains steady

```python
sns.histplot(data=df2, x='age', hue='gender', multiple='stack')
plt.show()
```

```python
sns.histplot(data=df2, x='age', hue='gender', stat='percent')
plt.show()
```

```python
df3 = pd.DataFrame({
    "salary": [500, 600, 650, 700, 800, 900, 1200, 1500, 2000]
})
df3
```

```python
sns.histplot(data=df3, x='salary', kde=True)
plt.show()
```

- The provided visualization is a combination of a histogram and a KDE plot showing the distribution of salary.
- Right-Skewed Distribution: The data exhibits a clear right skew (positive skew), where the majority of the observations are concentrated on the left side (lower salaries) with a long tail extending towards the right.
- Peak Frequency: The highest concentration of employees is in the 500 to 800 salary range, with a count of 4.
- Declining Trend: As the salary increases, the number of people earning that amount decreases significantly. Only single individuals are found in the higher brackets (above 1400).
- KDE Line: The smooth blue curve provides a continuous estimate of the probability density, highlighting that the distribution deviates from a standard normal bell curve

```python
np.random.seed(42)

df5 = pd.DataFrame({
    "salary": np.random.normal(loc=6_000_000, scale=1_500_000, size=500).astype(int),
    "experience_years": np.random.randint(0, 16, size=500)
})
df5
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df5, x='salary', hue='experience_years', multiple='stack', kde=True)
plt.show()
```

The provided visualization is a sophisticated histogram with overlaid KDE curves, showcasing the distribution of salary based on years of experience.

Distribution Shape (Normal Distribution): The overall pattern follows a Normal Distribution, resembling a classic bell curve. The majority of observations are centered around the average salary of approximately $6 \times 10^6$.

Experience Segmentation (hue='experience_years'):The legend indicates experience ranging from 0 to 15 years, represented by a gradient from light pink to dark purple.

Interestingly, the peak of the density curves for different experience levels remains largely centered around the same mean salary, suggesting a high concentration of staff within this pay bracket regardless of tenure.

Frequency and Density: The central bin reaches a peak count of nearly 70, highlighting that this is the most common salary range in the dataset.

Key Insight: Since the data is symmetrically distributed, there is no significant skewness. This suggests that a standard linear model might perform well on this feature without requiring log transformations

```python
np.random.seed(42)

subjects = ["Math", "Physics", "English"]

df6 = pd.DataFrame({
    "exam_score": np.random.randint(0, 101, size=400),
    "subject": np.random.choice(subjects, size=400, p=[0.35, 0.35, 0.30])
})
df6
```

```python
sns.histplot(data=df6, x='exam_score', hue='subject', multiple='fill', kde=True)
plt.show()
```

- The visualization is a filled stacked histogram with overlaid KDE curves, representing the proportional distribution of exam scores across different subjects.
- Plot Style (multiple="fill"): The Y-axis is normalized to a scale of 0 to 1.0 (0% to 100%), which allows for comparing the relative proportions of subjects within each score bin rather than absolute counts.

Subject Distribution (hue="subject"):

- English (Green): Occupies the bottom layer with a fairly consistent share of approximately 30% across the entire score range.
- Physics (Orange): Positioned in the middle; its relative share slightly increases in the middle-to-high score brackets (4080).
- Math (Blue): The top layer shows a significant presence in both the mid-range (3050) and high-range (80100) scores.
- KDE Curves: The trend lines track the probability density of each subject. The orange curve (Physics) shows a noticeable peak around the 20-score mark.
- Key Insight: The distribution for all subjects appears relatively uniform across the 0100 score range, with no single subject showing a strong skew toward very low or very high grades
