```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(0)

df_students = pd.DataFrame({
    "study_hours": np.random.randint(1, 10, 120),
    "exam_score": np.random.normal(65, 10, 120).astype(int)
})
df_students
```

```python
sns.jointplot(data=df_students, x='study_hours', y='exam_score')
plt.show()
```

- There is no linear increase in test scores with increasing study hours. For example, a student who studies for 1 hour may score 80, while a student who studies for 9 hours may score around 60. This means that the correlation between the variables is very weak or non-existent
- The highest frequency (peak) is observed around 4 hours and 8-9 hours
- The distribution of scores is close to a normal distribution (Bell curve). Most students scored between 60 and 75. Scores below 40 and above 85 are rare

```python
# with a regression line
sns.jointplot(data=df_students, x='study_hours', y='exam_score', kind='reg')
plt.show()
```

```python
np.random.seed(1)

df_marketing = pd.DataFrame({
    "ad_budget": np.random.randint(100, 5000, 200),
    "sales": np.random.randint(10, 300, 200)
})
df_marketing
```

```python
sns.jointplot(data=df_marketing, x='ad_budget', y='sales', kind='reg')
plt.show()
```

- The graph shows a very weak linear relationship between advertising budget and sales.
- a sharp increase in the budget does not always lead to a corresponding increase in sales
- The data points are spread out far enough from the regression line, indicating that many factors other than budget are affecting sales.

```python
np.random.seed(2)

df_shop = pd.DataFrame({
    "site_visits": np.random.randint(50, 5000, 300),
    "profit": np.random.randint(-100, 2000, 300)
})
df_shop
```

```python
sns.jointplot(data=df_shop, x='site_visits', y='profit', kind='reg')
plt.show()
```

- There is no significant linear relationship between traffic and profit
- An increase in the number of visitors to a site does not automatically guarantee an increase in profits
- There are several breakeven points between 100 and 1000 visits. There are also breakeven points above 2500 and 4500 visits. This means that some campaigns are not paying off, even though they are driving a lot of traffic.
