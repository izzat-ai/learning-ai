This page will teach you how to use countplot. DataFrames are created using AI.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_users = pd.DataFrame({
    "user_id": range(1, 16),
    "gender": [
        "Male", "Female", "Male", "Male", "Female",
        "Other", "Male", "Female", "Male", "Female",
        "Male", "Other", "Female", "Male", "Male"
    ]
})
df_users
```

```python
sns.countplot(x='gender', data=df_users, palette='Set2', hue='gender')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

- Total Sample Size: The dataset consists of 15 entries in total (8 Male, 5 Female, and 2 Other).
- Majority Class: The Male category is the dominant group with 8 counts, accounting for approximately 53% of the total population.
- Secondary Class: The Female category follows with 5 counts, representing roughly 33% of the dataset.
- Minority Class: The Other category has the lowest representation with only 2 counts (~13%).
- Data Insight: There is a noticeable class imbalance in the dataset. The number of males is significantly higher than other groups. From a machine learning perspective, the small sample size for the "Other" category might lead to poor generalization for this specific class if used in a predictive model

```python
df_orders = pd.DataFrame({
    "order_id": range(1, 21),
    "city": [
        "Tashkent", "Samarkand", "Tashkent", "Fergana", "Tashkent",
        "Bukhara", "Tashkent", "Fergana", "Samarkand", "Tashkent",
        "Andijan", "Tashkent", "Bukhara", "Samarkand", "Tashkent",
        "Fergana", "Tashkent", "Andijan", "Tashkent", "Tashkent"
    ]
})
df_orders
```

```python
sns.countplot(x='city', data=df_orders, palette='Set2', hue='city')
plt.title('City Distribution')
plt.xlabel('City')
plt.ylabel('Count')
plt.show()
```

- High Centralization: The chart clearly shows a heavy concentration of data in Tashkent. Out of 20 total observations, 10 (50%) belong to the capital city.
- Regional Disparity: There is a significant gap between Tashkent and the rest of the cities. Samarkand and Fergana have 3 entries each, while Bukhara and Andijan are the least represented with 2 entries each.
- Sampling Bias: This distribution suggests a geographical bias. If this dataset is intended to represent the entire country, it is currently skewed towards urban/poytaxt trends.
- Insight: For a machine learning model to be robust across different regions, we would need to collect more samples from the underrepresented cities to ensure it doesn't just "learn" Tashkent-specific patterns

```python
df_sales = pd.DataFrame({
    "customer_id": range(1, 25),
    "device": [
        "Android", "iOS", "Android", "Android", "iOS",
        "Android", "Android", "iOS", "Android", "Android",
        "iOS", "Android", "Android", "iOS", "Android",
        "Android", "Android", "iOS", "Android", "Android",
        "iOS", "Android", "Android", "Android"
    ],
    "purchased": [
        "Yes", "No", "Yes", "No", "Yes",
        "Yes", "No", "No", "Yes", "Yes",
        "No", "Yes", "Yes", "Yes", "No",
        "Yes", "No", "Yes", "Yes", "Yes",
        "No", "Yes", "No", "Yes"
    ]
})
df_sales
```

```python
sns.countplot(x='device', hue='purchased', data=df_sales, palette='Set2')
plt.title('Device and Purchase Distribution')
plt.xlabel('Device')
plt.ylabel('Count')
plt.show()
```

- Volume: Android users (17) significantly outnumber iOS users (7).
- Conversion: Android shows a high conversion rate of 70.6% (12 "Yes" vs 5 "No"). In contrast, iOS has a lower conversion rate of approximately 43% (3 "Yes" vs 4 "No").
- Insight: The Android platform is currently the primary driver of sales for this specific segment
