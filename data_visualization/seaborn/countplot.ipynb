```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_users = pd.DataFrame({
    "user_id": range(1, 16),
    "gender": [
        "Male", "Female", "Male", "Male", "Female",
        "Other", "Male", "Female", "Male", "Female",
        "Male", "Other", "Female", "Male", "Male"
    ]
})
df_users
```

```python
sns.countplot(x='gender', data=df_users, palette='Set2')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
