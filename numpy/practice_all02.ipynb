```python
# Determine how many elements are in each array
arr18 = np.array([1, 2, 3, 4, 5])
arr19 = np.array([[1, 2, 3], [4, 5, 6]])
arr20 = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

print("arr1 =", arr18.size)
print("arr2 =", arr19.size)
print("arr3 =", arr20.size)
```

```python
# Determine the data type of each array
arr25 = np.array([1, 2, 3])
arr26 = np.array([1.5, 2.5, 3.5])
arr27 = np.array(['a', 'b', 'c'])

print("arr1 =", arr25.dtype)
print("arr2 =", arr26.dtype)
print("arr3 =", arr27.dtype)
```

```python
arr28 = np.array([1, 2, 3, 4, 5])
print("original :", arr28.dtype)
```

```python
# convert array to float64 type
arr28_float64 = arr28.astype(np.float64)
print(arr28_float64)
print(arr28_float64.dtype)
```

```python
# Convert again to int32 type
arr28_int32 = arr28.astype(np.int32)
print(arr28_int32)
print("dtype =", arr28_int32.dtype)
```

```python
# Output the size of each array in bytes
arr30 = np.array([1.0, 2.0, 3.0], dtype=np.float16)
arr31 = np.array([1.0, 2.0, 3.0], dtype=np.float32)
arr32 = np.array([1.0, 2.0, 3.0], dtype=np.float64)

print("arr1 =", arr30.nbytes)
print("arr2 =", arr31.nbytes)
print("arr3 =", arr32.nbytes)
```
