```python
import numpy as np
```

```python
# Create a numpy array from this list and display it on the screen
data = [10, 20, 30, 40, 50]
arr1 = np.array(data)
arr1
```

```python
# create a numpy array
matrix = [[1, 2, 3],
          [4, 5, 6],
          [7, 8, 9]]
arr2 = np.array(matrix)
arr2
```

```python
# Extract the data type and shape of an array
print("dtype =", arr2.dtype)
print("shape =", arr2.shape)
```

```python
# Create an array of size 4x5 with all elements 0
arr3 = np.zeros((4, 5))
arr3
```

```python
# Create a 3x3x3 array (3D), all elements of which are 0
arr4 = np.zeros((3, 3, 3))
arr4
```

```python
# Output the size, shape, and total number of elements of an array
print("format =", arr4.ndim)
print("shape =", arr4.shape)
print("size =", arr4.size)
```

```python
# Create an array of size 5x4, all elements of which are 1
arr5 = np.ones((5, 4))
arr5
```

- *Create an array of size 2x6, all elements of which are 1, and of data type float*

```python
arr6 = np.ones((2, 6), dtype=np.float16)
print(arr6)
print("dtype =", arr6.dtype)
```

```python
# Create a 3x4 array with all elements 7
arr7 = np.full((3, 4), 7)
arr7
```

```python
# Create a 5x5 array with all elements 3.14
arr8 = np.full((5, 5), 3.14)
print(arr8)
print("dtype =", arr8.dtype)
```

```python
# Create a 5x5 identity matrix
arr9 = np.eye(5)
arr9
```

```python
# Create a 7x7 unit matrix
arr10 = np.eye(7)
arr10
```

```python
# Calculate the sum of diagonal elements
arr10_diags = np.diag(arr10)
arr10_diags.sum()
```

```python
# Create an array of numbers from 0 to 50
arr11 = np.arange(0, 50)
arr11
```

```python
# Create an array of numbers from 10 to 100, every 5 digits
arr12 = np.arange(10, 100, 5)
arr12
```
