```python
import numpy as np
```

```python
temperatures = np.array([
    15, 17, 22, 28, 33, 38, 40, 39, 34, 27, 20, 16, # Shahar A
    10, 12, 18, 24, 30, 35, 37, 36, 31, 23, 15, 11, # Shahar B
    25, 26, 28, 30, 32, 33, 34, 34, 33, 31, 28, 26  # Shahar C (Tropik)
])
temperatures.size
```

- View each city's data in a separate row

```python
data = temperatures.reshape(3, 12)
data
```

- Temperatures above 35 degrees Celsius are considered "Abnormally hot" due to global warming
- Filter only temperatures above 35 degrees Celsius in City B (City 2)

```python
data[data>35]
```

```python
city2_anomal = data[1][data[1]>35]
city2_anomal
```

- When analyzing the data for City B, it was found that there were a total of 2 cases where the temperature exceeded 35 degrees Celsius. This is the hottest time of the summer for that area

```python
# convert temperature from Celsius to Fahrenheit
temps_f = data*1.8 + 32
temps_f
```

```python
# determine the average summer temperature for each city
np.mean(temps_f[:, [5, 6, 7]], axis=1)
```

```python
# transpose each city to compare their temperatures
fahren = temps_f.T
fahren
```

```python
# determining the hottest city in January
np.argmax(fahren[0, :])
```

```python
# The temperature in all cities will increase by 2% next year. Predicting data for next year
new_year_temp = temps_f*1.02
new_year_temp
```

```python
# Combining old information and new predictions into one
data_3d = np.dstack((data, new_year_temp))
data_3d
```

- City B has a very high temperature volatility of 10 to 37. This can be a challenge for agriculture .
- The temperature variation in City C is very small (between 25 and 34 degrees Celsius). It is the most suitable city for tourism .
- Next year's forecast predicts that the hottest temperature in City A will exceed 106 Fahrenheit. This should be a warning to the health system .
