- Classify each temperature into a category:
  - Above 30: 'Hot'
  - Between 20-30: 'Warm'
  - Between 10-20: 'Cool'
  - Below 10: 'Cold

```python
temperatures = np.array([15, 28, 35, 10, 22, 38, 5, 31, 18, 42])

print("Temperatures:", temperatures)
```

```python
conditions = [
    (temperatures>30),
    (temperatures >= 20) & (temperatures <= 30),
    (temperatures >=10) & (temperatures <=20),
    (temperatures < 10)
]
second_conds = ['hot', 'warm', 'cool', 'cold']
np.select(conditions, second_conds, temperatures)
```

- Create a category based on age:
  - Under 18: 'Minor'
  - 18-30: 'Young Adult'
  - 31-50: 'Adult'
  - Over 50: 'Senior'

```python
ages = np.array([15, 22, 35, 8, 45, 17, 60, 12, 28, 50])

print("Ages:", ages)
```

```python
conditions_age = [
    (ages < 18),
    (ages>=18) & (ages<=30),
    (ages>=31) & (ages<=50),
    (ages>50)
]
select_conds = ['minor', 'young adult', 'adult', 'senior']
np.select(conditions_age, select_conds, ages)
```

```python
prices = np.array([1200, 450, 800, 350, 1500, 600, 2000, 300])

print("Prices:", prices)
```

```python
conditions_prices = [
    prices>1000,
    (prices>=500) & (prices<=1000),
    prices<500
]
select_conds_prices = ['premium', 'standard', 'budget']
np.select(conditions_prices, select_conds_prices, default=' ')
```
