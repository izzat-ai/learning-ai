```python
matrix = np.array([[2, 1],
                   [7, 4]])
matrix
```

```python
# Find the determinant and inverse of this matrix
det = np.linalg.det(matrix)
inv = np.linalg.inv(matrix)
print('det:', det)
print('inv:', inv)
```

```python
matrix*inv
```
