```python
scores = np.array([45, 67, 89, 34, 56, 78, 90, 23, 67, 88])
scores
```

```python
# Find all scores above 60
scores[scores>60]
```

```python
# Find the scores between 50 and 80 (including 50 and 80)
scores[(scores>=50) & (scores<=80)]
```

```python
prices = np.array([1200, 450, 800, 350, 1500, 600, 2000, 300])
prices
```

```python
# Change prices less than 500 to 0
prices[prices<500] = 0
```

```python
prices
```

```python
# Mark prices above 1000 as "Expensive" and the rest
# as "Affordable" (create a new array)
new_prices = np.where(prices>1000, "expensive", "affordable")
new_prices
```
