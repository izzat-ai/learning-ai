```python
arr2 = np.array([1, 2, 3, 4, 5])
arr3 = np.array([[10, 20, 30], [40, 50, 60]])
arr2
```

```python
arr3
```

```python
# dimensions
print("arr2:", arr2.ndim)
print("arr3:", arr3.ndim)
```

```python
# shapes
print("arr2:", arr2.shape)
print("arr3:", arr3.shape)
```

```python
# Create an array of even numbers from 0 to 20
arr4 = np.arange(0, 20, 2)
arr4
```

```python
# Create a 3x4 array with all elements 0
arr5 = np.zeros(shape=(3, 4))
arr5
```

```python
# Create an array of size 2x5, all elements of which are 1
arr6 = np.ones((2, 5))
arr6
```

```python
# Create a 4x4 identity matrix
arr7 = np.eye(4, 4)
arr7
```

```python
arr8 = np.arange(10, 110, 10)
arr8
```

```python
print(arr8[2]) # 30
print(arr8[-1]) # 100

# 30->70
print(arr8[2:6])
```

```python
# get every 2nd element
arr8[::2]
```

```python
arr9 = np.arange(10, 130, 10).reshape(3, 4)
arr9
```

```python
# Get the element in row 2, column 3
arr9[1, 2]
```

```python
# Extract all elements of the first row
arr9[0, :]
```

```python
# get all elements of the last column
arr9[:, -1]
```

```python
# Get the 2x2 sub-section: [[60, 70], [100, 110]]
arr9[1:, 1:3]
```
