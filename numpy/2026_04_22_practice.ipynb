```python
temperatures = np.array([5, 15, 25, 35, 10, 30, 40, 8, 22, 38])

print("Temperatures:", temperatures)
```

```python
temperatures.size
```

```python
# Return 'Cold' if below 15, 'Moderate' if between 15-30, 'Hot' if above 30
cond2 = [
    temperatures < 15, # cold
    (temperatures >= 15) | (temperatures <= 30), # moderate
    temperatures > 30 # hot
]
choices = ['cold', 'moderate', 'hot']
result = np.select(cond2, choices, default='Unknown')
result
```

- Create a category based on age:
  - Under 18: 'Minor'
  - 18-30: 'Young Adult'
  - 31-50: 'Adult'
  - Over 50: 'Senior'

```python
cond3 = [
    ages < 18, # minor
    (ages >= 18) & (ages <= 65), # young adult
    (ages >= 31) & (ages <= 50), # adult
    ages > 50 # senior
]
choices2 = ['minor', 'young adult', 'adult', 'senior']
result2 = np.select(cond3, choices2, default='Unknown')
result2
```

- Determine the level based on salary:
  - Below 5000: 'Junior'
  - 5000-10000: 'Middle'  
  - 10000-15000: 'Senior'
  - Above 15000: 'Lead'
  

```python
salaries = np.array([3000, 5000, 7000, 10000, 15000, 2500, 8000, 12000])

print("Salaries:", salaries)
```

```python
cond5 = [
    salaries < 5000, # junior
    (salaries >= 5000) & (salaries <=10_000), # middle
    (salaries >= 10_000) & (salaries <= 15_000), # senior
    salaries > 15_000 # lead
]
choices4 = ['junior', 'middle', 'senior', 'lead']
result4 = np.select(cond5, choices4, default='Unknown')
result4
```

```python
uniq_els, counts_el = np.unique(result4, return_counts=True)
print(uniq_els)
print(counts_el)
```
