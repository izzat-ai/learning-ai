```python
import numpy as np
```

```python
# Create an array of size 4x6, all elements of which are 7.5
arr1 = np.full((4, 6), 7.5)
arr1
```

```python
# Extract the shape and data type of the array
print("shape :", arr1.shape)
print("dtype :", arr1.dtype)
```

```python
# Create an array of size 5x5, all elements of which are 3.14, and of data type int32
arr2 = np.full((5, 5), 3.14, dtype=np.int32)
arr2
```

```python
# Output the total number of elements and size (in bytes) of an array
print("elements :", arr2.size)
print("bytes :", arr2.nbytes)
```

```python
# Create a 6x6 unit matrix
arr3 = np.eye(6, 6)
arr3
```

```python
# Calculate the sum of diagonal elements
arr3.diagonal().sum()
```

```python
# Calculate the sum of all non-diagonal elements
np.sum(arr3) - np.trace(arr3)
```

```python
# Create an 8x8 unit matrix
arr4 = np.eye(8, 8)
arr4
```

```python
# Multiply the matrix by 5
arr4 = arr4*5
arr4
```

```python
# Change diagonal elements to 0

# copy from original array
arr4_diag = arr4.copy()

# get indices of diagonal values
diag_idx = np.diag_indices(arr4_diag.shape[0])
diag_idx
```

```python
arr4_diag[diag_idx] = 0
```
