```python
data = np.array([
    [101, 5, 250, 15],
    [102, 1, 30, 2],
    [103, 12, 1200, 45],
    [104, 0, 0, 1],
    [105, 3, 150, 10],
    [106, 8, 600, 30],
    [107, 2, 40, 5],
    [108, 15, 2000, 60],
    [109, 1, 15, 1],
    [110, 4, 300, 20]
])
```

```python
print(data)
```

- Identify customers who visited the site but did not purchase anything and separate their data into a separate array

```python
# number of products purchased
num_products = data[:, 1]
num_products
```

```python
mask = num_products<1
```

```python
customer_not_buy = data[mask]
customer_not_buy
```

```python
# purchases
purchases = data[:, 2]
purchases
```

```python
purchase_mean = np.mean(purchases)

np.where(purchases>purchase_mean, 'vip', 'classic')
```

```python
purchase_log = np.log1p(purchases)
purchase_log
```

```python
# add new column to array
new_data5col = np.column_stack((data, purchase_log))
new_data5col
```

```python
# customers who spent less than 2 minutes on the site
time_mask = new_data5col[:, 3] < 2
time_mask
```

```python
clean_data = new_data5col[~time_mask]
clean_data
```
