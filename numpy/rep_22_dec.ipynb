####**slicing**

```python
arr4 = np.array([5, 12, 7, 20, 3, 15, 8, 25])
arr4
```

```python
# 10 dan katta qiymatlarni olish
arr4[arr4>10]
```

```python
# birinchi 5 ta elementni olish
arr4[:5]
```

```python
# ohirgi 4 ta elementni olish
arr4[-3:]
```

```python
# 5 va 20 oralig'idagi qiymatlarni olish
arr4[(arr4>5) & (arr4<20)]
```

```python
juftlar = []
for n in arr4:
  if n%2 == 0:
    juftlar.append(n)
juftlar
```

```python
# 10 dan kichik qiymatlarni 0 ga almashtirish
arr4[arr4 < 10] = 0
arr4
```

```python
arr7 = np.array([10, np.nan, 20, np.nan, 30, 40])
arr7
```

```python
np.sum(np.isnan(arr7))
```

```python
# nan bo'lmagan qiymatlarni o'rtachasini aniqlash
np.nanmean(arr7)
```

```python
nans = np.isnan(arr7)
arr7[nans] = 0
arr7
```

```python
arr7 = np.array([10, np.nan, 20, np.nan, 30, 40])
arr7
```

```python
np.sum(np.isnan(arr7))
```

```python
# na bo'lmagan qiymatlarni olish
arr7[~np.isnan(arr7)]
```

```python
arr7_mean = np.nanmean(arr7)
np.isnan(arr7)
```

```python
arr7[np.isnan(arr7)] = arr7_mean
arr7
```
