```python
arr2 = np.arange(10, 30).reshape(5, 4)
arr2
```

```python
# har bir qatordagi eng katta qiymatlarni aniqlash
max_ax = np.max(arr2, axis=1)
max_ax
```

```python
# har bir ustundagi o'rtacha qiymatni hisoblash
mean_col = np.mean(arr2, axis=0)
mean_col
```

```python
arr11 = np.random.uniform(-100, 100, 11)
arr11
```

```python
# har bir elementni musbat yoki manfiyligini aniqlash
np.sign(arr11)
```

```python
# massivning manfiy qiymatlarini musbat qilish
arr11_new = np.abs(arr11)
arr11_new
```

```python
print("minimum qiymat:", np.min(arr11_new))
print("maksimum qiymat:", np.max(arr11_new))
```

```python
# max va min qiymatlari farqi
np.max(arr11_new) - np.min(arr11_new)
```
