#####**non-repeatable and other actions**

```python
arr19 = np.array([4, 7, 2, 7, 4, 9, 2, 5, 7, 9, 1, 4])
arr19
```

```python
# tarorlanmas qiymatlarni olish
np.unique(arr19)
```

```python
val, couns = np.unique(arr19, return_counts=True)
val
```

```python
# qiymatlar necha marta takrorlanganini aniqledi
couns
```

```python
arr_main = np.array([10, 15, 20, 25, 30, 35, 40])
arr_check = np.array([15, 22, 30, 41])
```

```python
# arr_maindagi elementni qaysi birir arr_checkda mavjudligini aniqlash
np.isin(arr_main, arr_check)
```

```python
arr_A = np.array([1, 2, 3, 4, 5, 6, 7])
arr_B = np.array([2, 4, 6, 8])
```

```python
# ikkinchi massivda bo'lmagan arr_A ning qiymatlarini olish
np.setdiff1d(arr_A, arr_B)
```

```python
np.isin(arr_A, arr_B)
```

```python
arr_count = np.array([0, 1, 1, 2, 3, 2, 1, 0, 4, 2, 3, 3, 1])
arr_count
```

```python
# har bir element necha marta takrorlangan
val2, couns2 = np.unique(arr_count, return_counts=True)
couns2
```

```python
for val, count in zip(val2, couns2):
    print(f"{val} -> {count} marta")
```
