- We have 5 users of a streaming platform and viewing figures for 6 movie genres.
- The data represents the users' interest in each genre (score from 0 to 100) and their subscription status.

- The users interest scores for the 6 genres are represented as 6 rows in the array below.
- Genres: `[Action, Comedy, Drama, Sci-Fi, Horror, Romance]`

```python
import numpy as np
```

```python
raw_ratings = np.array([
    [85, 40, 20, 90, 10, 5],   # User 0
    [20, 95, 80, 10, 5, 75],   # User 1
    [0, 0, 0, 0, 0, 0],        # User 2 (new profile, no data yet)
    [60, 60, 60, 60, 60, 60],  # User 3
    [45, 55, 100, 30, 20, 85]  # User 4
])
raw_ratings
```

```python
# User monthly subscription fees (USD)
subscriptions = np.array([15.99, 9.99, 0.0, 12.99, 15.99])
subscriptions
```

```python
raw_ratings.shape
```

```python
# Determine the average of all rows except row 3
average_values = np.mean(raw_ratings[[0, 1, 3, 4], :], axis=0)
average_values
```

```python
# convert the main array to float type
raw_ratings = raw_ratings.astype(float)
raw_ratings
```

```python
# Complementing the user's scores in row 3 with the average of all users
raw_ratings[2] = average_values
raw_ratings
```

```python
# Determine which users have given a score above 70 for the "Drama" genre (column 3) and currently have an active subscription
drama_active = raw_ratings[(raw_ratings[:, 2]>70) & (subscriptions>0)]
drama_active
```

```python
# The platform wants to promote the "Action" and "Sci-Fi" genres. Increase action (column 1) and sci-fi (column 4)
# by 1.5 times
raw_ratings[:, [0, 3]] = raw_ratings[:, [0, 3]]*1.5
raw_ratings
```

```python
# to bring the data into the range of 0-1 to feed it into the model
raw_ratings = raw_ratings/100
raw_ratings
```

```python
# Find the highest rated genre (index) for each user
raw_ratings.argmax(axis=1)
```

```python
# put payments in column format
subscriptions = subscriptions.reshape(-1, 1)
subscriptions
```

```python
np.set_printoptions(suppress=True, precision=6)

# Combining user ratings and their subscription amounts into a single array
data = np.hstack((raw_ratings, subscriptions))
data
```
