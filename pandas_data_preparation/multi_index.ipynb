```python
import numpy as np
import pandas as pd
```

```python
ser1 = pd.Series([11, 12, 13, 14, 15, 16, 17, 18, 19],
                 index=[['a', 'a', 'a', 'b', 'b', 'c', 'c', 'd', 'd'], [1, 2, 3, 1, 3, 1, 2, 3, 2]])
ser1
```

```python
# a indeksining barcha qiymatlarini olish
ser1['a']
```

```python
# c indeksining barcha qiymatlarini olish
ser1['c']
```

```python
# a ning 3-qiymatini olish
ser1.loc['a', 3]
```

```python
# d ning ikkinchi qiymatini olish
ser1.loc['d', 2]
```

```python
index = pd.MultiIndex.from_tuples([
    ("Uzbekistan", "Tashkent"),
    ("Uzbekistan", "Fergana"),
    ("Kazakhstan", "Almaty"),
    ("Kazakhstan", "Astana")
], names=["Country", "City"])

df = pd.DataFrame({
    "Population": [3000000, 1200000, 2000000, 1000000]
}, index=index)
df
```

```python
# faqatgina Uzbning ma'lumotlarini olish
df.loc['Uzbekistan']
```

```python
# Uzbning Toshkent ma'lumotini olish
df.loc['Uzbekistan', 'Tashkent']
```

```python
index = pd.MultiIndex.from_product(
    [[2024], ["January", "February"], [1, 2]],
    names=["Year", "Month", "Day"]
)

df2 = pd.DataFrame({
    "Temp": [5, 6, 3, 4]
}, index=index)
df2
```

```python
# 2024-yilning Fevral ma'lumotlarini olish
df2.loc[2024, 'February', :]
```

```python
df3 = pd.DataFrame(
    {
        'baholar': [5, 4, 3, 5, 4, 5]
    },
    index=pd.MultiIndex.from_tuples(
        [
            ('9-A', 'Matematika'),
            ('9-A', 'Fizika'),
            ('9-B', 'Matematika'),
            ('9-B', 'Fizika'),
            ('10-A', 'Matematika'),
            ('10-A', 'Fizika'),
        ],
        names=['sinf', 'fan']
    )
)
df3
```

```python
# Har bir sinf bo‘yicha o‘rtacha bahoni hisoblash
df3.groupby(level='sinf').mean()
```

```python
df4 = pd.DataFrame(
    {
        'aholi': [2.5, 2.7, 8.3, 8.5, 5.1, 5.3]
    },
    index=pd.MultiIndex.from_tuples(
        [
            ('Uzbekistan', 'Tashkent', 2020),
            ('Uzbekistan', 'Tashkent', 2021),
            ('Turkey', 'Istanbul', 2020),
            ('Turkey', 'Istanbul', 2021),
            ('USA', 'Chicago', 2020),
            ('USA', 'Chicago', 2021),
        ],
        names=['mamlakat', 'shahar', 'yil']
    )
)
df4
```

```python
# har bir mamlakat bo‘yicha umumiy aholi o‘sishini aniqlash
df4.groupby(level='mamlakat')['aholi'].sum()
```

```python
df5 = pd.DataFrame(
    {
        'xodimlar': [25, 30, 12, 15, 40, 38]
    },
    index=pd.MultiIndex.from_tuples(
        [
            ('Google', 'AI'),
            ('Google', 'Marketing'),
            ('Amazon', 'AI'),
            ('Amazon', 'Marketing'),
            ('Meta', 'AI'),
            ('Meta', 'Marketing'),
        ],
        names=['kompaniya', 'bo‘lim']
    )
)
df5
```

```python
# Har bir kompaniyada eng ko‘p xodimga ega bo‘limni topish
df5.groupby(level='kompaniya')['xodimlar'].idxmax()
```

```python
df6 = pd.DataFrame({
    "Region": ["North","North","North","South","South","South"],
    "City": ["A","A","B","C","C","D"],
    "Sales": [120, 150, 90, 200, 210, 160]
}).set_index(["Region","City"])
df6
```

```python
df6.groupby(level=0)['Sales'].idxmax()
```

```python
df7 = pd.DataFrame({
    "Company": ["X","X","X","Y","Y","Y"],
    "Department": ["IT","HR","IT","Sales","Sales","HR"],
    "Employees": [20, 5, 25, 30, 15, 8]
}).set_index(["Company","Department"])
df7
```

```python
df7.groupby(level=0)['Employees'].sum().idxmax()
```
