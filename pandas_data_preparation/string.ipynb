```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'Full_Name': ['Mr. Ahmed Ali', 'Dr. Sara Khan', 'Ms. Fatima Rahman'],
    'Email': ['ahmed.ali@company.com', 'sara.khan@tech.org', 'fatima.r@startup.io'],
    'Phone': ['(998) 90-123-4567', '998-91-234-5678', '998 93 345 6789']
})
df
```

```python
# Remove titles like "Mr.", "Dr.", "Ms." from Full_Name
df['Full_Name'] = df['Full_Name'].str.split('.').str[1]
df
```

```python
# Extract only the username part from the email
df['Email'] = df['Email'].str.split('@').str[0]
df
```

```python
# Convert phone numbers to numbers only
df['Phone'] = df['Phone'].replace(r'\D', '', regex=True)
df
```

```python
df1 = pd.DataFrame({
    'Product_Code': ['ELEC-LAP-001', 'FURN-CHA-002', 'ELEC-PHO-003', 'FURN-TAB-004'],
    'Description': ['  High Performance Laptop  ', 'ERGONOMIC OFFICE CHAIR', '  smartphone 5G  ', 'wooden table'],
    'Category': ['Electronics', 'Furniture', 'Electronics', 'Furniture']
})
df1
```

```python
# Extract the first part from Product_Code (ELEC, FURN
df1['Product_Code'].str.split('-').str[0]
```

```python
# Remove spaces from Description and capitalize the first letter of each word
df1['Description'] = df1['Description'].str.strip()
df1['Description'] = df1['Description'].str.capitalize()
df1
```

```python
# If the length of Category is greater than 10, write "Long", otherwise "Short"
df1['Category'].apply(lambda x: 'Long' if len(x)>10 else 'short')
```

```python
df2 = pd.DataFrame({
    'Customer_ID': ['USR_001_ACTIVE', 'ADM_002_INACTIVE', 'USR_003_ACTIVE', 'MOD_004_ACTIVE'],
    'Username': ['user@example.com', 'admin@system.net', 'customer@shop.uz', 'moderator@forum.org'],
    'Status': ['Active', 'Inactive', 'Active', 'Active']
})
df2
```

```python
# Extract only the number part from Customer_ID
df2['Customer_ID'].str.split('_').str[1]
```

```python
# Check if the username contains the word "admin" or "moderator"
(df2['Username'].str.split('@').str[0] == 'admin') | (df2['Username'].str.split('@').str[0] == 'moderator')
```

```python
# If the username is longer than 20 characters, shorten it (the first 20 characters
df2['Username'].apply(lambda x: x[:20] if len(x)>20 else x)
```
