```python
import numpy as np
import pandas as pd
```

```python
A = pd.DataFrame({
		'Transport':['poyezd', 'avtomobil', 'kema'],
		'Tezligi':['70', '180', '45']
})
B = pd.DataFrame({
		'Transport':['poyezd', 'avtomobil', 'tramvay'],
		'Narxi':['98000', '10000', '65000']
})
A
```

```python
B
```

```python
# kalit ustun bo'yicha DF larni birlashtirish
pd.merge(A, B)
```

```python
# A dataframedagi barcha qiymatlar bilan qo'shish
pd.merge(A, B, how='left')
```

```python
# B dataframedagi barcha qiymatlar bilan qo'shish
pd.merge(A, B, how='right')
```

```python
# ikkala DF dagi barcha elementlarni qo'shish
pd.merge(A, B, how='outer')
```

- yuqorida kalit ustun `Transport` bo'ldi
- ikkala DF da bir xil ustun mavjud bo'lmagan ustunlar bo'yicha birlashtirish

```python
A = A.rename({'Transport':"Transport_turi"}, axis='columns')
A
```

```python
B
```

```python
# bir xil bo'lmagan ustunli DF larni bir ustun bo'yicha birlashtirish
merged = pd.merge(A, B, left_on='Transport_turi', right_on='Transport')
merged
```

```python
# Transport_turi ustunini tashavorish
merged.drop('Transport_turi', axis=1)
```

```python
A = A.set_index('Transport_turi')
A
```

```python
# A ning indekslari bo'yicha B ning Transport ustuni bilan birlashtirish
pd.merge(A, B, left_index=True, right_on='Transport')
```

```python
df_mahsulot = pd.DataFrame({
    'Mahsulot_ID': ['P001', 'P002', 'P003', 'P004'],
    'Nomi': ['Noutbuk', 'Telefon', 'Monitor', 'Printer'],
    'Narx': [1200, 800, 350, 250]
})

# DF 2: Sotuvlar (Kalit: item_id ustuni. DF1 dagi kalit nomiga mos emas)
df_sotuv = pd.DataFrame({
    'Tranzaksiya_ID': [101, 102, 103, 104, 105, 106],
    'item_id': ['P001', 'P003', 'P003', 'P005', 'P002', 'P001'],
    'Sotilgan_Miqdor': [2, 1, 3, 1, 4, 1]
})

# DF 3: Toifalar (Kalit: Toifa_ID indeksi)
df_toifa = pd.DataFrame({
    'Toifa_Nomi': ['Elektronika', 'Aksessuar', 'Maishiy'],
    'Javobgar_Xodim': ['Ali', 'Vali', 'Guli']
}, index=[10, 20, 30])
df_toifa.index.name = 'Toifa_ID'

# DF 4: Mahsulot_Toifasi (Kalit: Mahsulot_ID indeksi)
df_mahsulot_toifa = pd.DataFrame({
    'Toifa_ID': [10, 10, 10, 20]
}, index=['P001', 'P002', 'P003', 'P004'])
df_mahsulot_toifa.index.name = 'Mahsulot_ID'
```
