```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)
```

```python
df1 = pd.DataFrame({
    "product": ["Phone","Laptop","Tablet","Monitor","Keyboard"],
    "price": [1200, None, 600, None, 80],
    "quantity": [5, 3, None, 4, 10]
})
df1
```

```python
# Determine how many blank values ​​are in which columns
df1.isnull().sum()
```

```python
# Check if total revenue is being calculated incorrectly due to empty values
df1['price'] * df1['quantity']
```

```python
df1 = df1.dropna().copy()
df1
```

```python
# recalculate total revenue based on the cleaned dataset
df1['revenue'] = df1['price'] * df1['quantity']
df1
```

```python
df2 = pd.DataFrame({
    "student": ["Ali","Vali","Ali","Hasan","Vali","Zafar"],
    "math": [85, 90, 85, 70, 90, 88]
})
df2
```

```python
# Identify duplicate entries
df2.duplicated()
```

```python
# Remove duplicate rows and determine how many unique students remain
len(df2.drop_duplicates())
```

```python
df3 = pd.DataFrame({
    "gender": ["M","F","M","F","F","M"],
    "score": [78, 88, 90, 67, 95, 80]
})
df3
```

```python
# Convert the values ​​in the Gender column to full word form
df3['gender'] = df3['gender'].replace({'M':'male', 'F':'female'})
df3
```

```python
# Change values ​​below 80 in the Score column to “Low”, otherwise to “High”
df3['score'].map(lambda x: 'low' if x<80 else 'high')
```

```python
df4 = pd.DataFrame({
    "age": [15, 22, 37, 45, 18, 29, 60]
})
df4
```

```python
# Divide the ages into 3 intervals and count how many people are in each interval
bins = [14, 24, 44, 65]
df4['groups'] = pd.cut(df4['age'], bins, labels=['teenagers', 'middle-age', 'adults'])
df4
```

```python
df4['groups'].value_counts()
```

```python
df5 = pd.DataFrame({
    "region": np.random.choice(["Toshkent","Samarqand","Buxoro"], 15),
    "sales": np.random.randint(1000, 5000, 15)
})
df5
```

```python
# Calculate the average sales for each region
df5.groupby('region')['sales'].mean()
```

```python
# Calculate the average sales for each region
df5.groupby('region')['sales'].sum().idxmax()
```
