```python
df_sales = pd.DataFrame({
    ("Electronics", "Phone"): [1200, 1500, 1400],
    ("Electronics", "Laptop"): [2000, 2200, 2100],
    ("Furniture", "Table"): [500, 600, 550],
    ("Furniture", "Chair"): [300, 350, 320],
}, index=["Jan", "Feb", "Mar"])

df_sales
```

```python
df_scores = pd.DataFrame({
    ("Math", "Test1"): [78, 90, 66, 82],
    ("Math", "Test2"): [85, 92, 70, 88],
    ("Physics", "Test1"): [80, 75, 60, 90],
    ("Physics", "Test2"): [88, 78, 65, 94],
}, index=["Ali", "Bekzod", "Dilshod", "Sardor"])

df_scores
```

```python
# test ustunlarini o'rtacha qiymatlari
df_scores.mean()
```

```python
# Har bir fanning 2 ta test bo‘yicha umumiy o‘rtacha natijasini hisoblash
df_scores.mean().groupby(level=0).mean()
```

```python
df_fin = pd.DataFrame({
    ("2022", "Revenue"): [120000, 150000, 130000],
    ("2022", "Profit"): [20000, 25000, 21000],
    ("2023", "Revenue"): [135000, 160000, 140000],
    ("2023", "Profit"): [22000, 28000, 23000],
}, index=["CompanyA", "CompanyB", "CompanyC"])

df_fin
```

```python
# faqatgina revenue ustunlari qiymatlarini olish
reven = df_fin.xs('Revenue', level=1, axis=1)
reven
```

```python
growth = reven['2023'] - reven['2022']
growth
```

```python
arrays1 = [
    ['Phone','Phone','Laptop','Laptop','Tablet','Tablet'],
    ['Sales','Profit','Sales','Profit','Sales','Profit']
]

tuples1 = list(zip(*arrays1))
index1 = pd.MultiIndex.from_tuples(tuples1, names=['Product','Metric'])

df_sales = pd.DataFrame(
    [
        [1200, 300, 800, 200, 500, 100],   # Tashkent
        [900, 250, 700, 180, 400, 90],     # Samarkand
        [700, 200, 600, 150, 350, 80],     # Bukhara
        [1000, 270, 750, 190, 450, 95]     # Andijan
    ],
    index=['Tashkent','Samarkand','Bukhara','Andijan'],
    columns=index1
)

df_sales
```
