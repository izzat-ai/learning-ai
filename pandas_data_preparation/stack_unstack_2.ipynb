```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'Math': [78, 85, 90],
    'Physics': [82, 88, 91],
    'Chemistry': [75, 80, 89]
}, index=['Ali', 'Vali', 'Hasan'])
df1
```

```python
df1_st = df1.stack()
df1_st
```

```python
arrays = [
    ['2022', '2022', '2023', '2023'],
    ['Q1', 'Q2', 'Q1', 'Q2']
]

columns = pd.MultiIndex.from_arrays(arrays, names=['Year', 'Quarter'])

df2 = pd.DataFrame({
    ('2022', 'Q1'): [100, 120],
    ('2022', 'Q2'): [130, 140],
    ('2023', 'Q1'): [150, 160],
    ('2023', 'Q2'): [170, 180]
}, index=['CompanyA', 'CompanyB'])

df2.columns = columns
df2
```

```python
df2.index
```

```python
df2.columns
```

```python
df2.stack(future_stack=True).stack(future_stack=True)
```

```python
df3 = pd.DataFrame({
    'Product': ['Phone', 'Phone', 'Laptop', 'Laptop'],
    'Month': ['Jan', 'Feb', 'Jan', 'Feb'],
    'Sales': [200, 250, 300, 350]
}).set_index(['Product', 'Month'])
df3
```
