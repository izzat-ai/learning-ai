```python
df2 = pd.DataFrame({
    "name": ["Ali", "Vali", "Ali", "Sanjar", "Vali"],
    "age": [20, 30, 20, 25, 30]
})
df2
```

```python
# takrorlangan qatorlarni aniqlang (True/False qaytarsin)
df2.duplicated()
```

```python
# takrorlangan qiymatlarni o'chirish
df2.drop_duplicates()
```

```python
df3 = pd.DataFrame({
    "name": ["Ali", "Ali", "Ali", "Vali", "Vali"],
    "score": [90, 95, 90, 80, 80],
    "group": ["A", "A", "B", "A", "A"]
})
df3
```

```python
# faqat name ustuniga qarab takrorlarni olib tashlash
df3.drop_duplicates(['name'])
```

```python
# 'name', 'score' ustunlari bo'yicha takrorlarni o'chirish
df3.drop_duplicates(['name', 'score'])
```

```python
df4 = pd.DataFrame({
    "id": [1,2,3,3,4,4,4],
    "value": [10,20,30,30,40,50,40]
})
df4
```

```python
# har bir id uchun faqat oxirgi uchragan qatorni qoldirish
df4.drop_duplicates(['id'], keep='last')
```
