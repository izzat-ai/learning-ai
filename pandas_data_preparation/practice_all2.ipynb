This page will be a practice of data preparation. DFs will be created using AI.

```python
import numpy as np
import pandas as pd
```

```python
arrays = [
    ["Toshkent", "Toshkent", "Samarqand", "Samarqand"],
    ["Jan", "Feb", "Jan", "Feb"]
]

index = pd.MultiIndex.from_arrays(arrays, names=("city", "month"))

df_sales = pd.DataFrame(
    {"sales": [1200, 1500, 900, 1100]},
    index=index
)
df_sales
```

```python
# faqat toshkent ma'lumotlarini olish
df_sales.loc['Toshkent']
```

```python
# fevralga oid ma'lumotlar
df_sales.loc[(slice(None), 'Feb'), :]
```

```python
# oddiy indeksli DF qilish
df_sales.reset_index()
```
