```python
import numpy as np
import pandas as pd
```

```python
arrays = [
    ["Toshkent", "Toshkent", "Samarqand", "Samarqand"],
    ["Jan", "Feb", "Jan", "Feb"]
]

index = pd.MultiIndex.from_arrays(arrays, names=("city", "month"))

df_sales = pd.DataFrame(
    {"sales": [1200, 1500, 900, 1100]},
    index=index
)
df_sales
```

```python
# faqat toshkent ma'lumotlarini olish
df_sales.loc['Toshkent']
```

```python
# fevralga oid ma'lumotlar
df_sales.loc[(slice(None), 'Feb'), :]
```

```python
# oddiy indeksli DF qilish
df_sales.reset_index()
```

```python
df_scores = pd.DataFrame({
    "student": ["Ali", "Vali", "Hasan"],
    "math": [85, 78, 92],
    "english": [88, 80, 90]
})
df_scores
```

```python
# jadvalni uzun formatka o'tkizish
df_scores = pd.melt(df_scores ,id_vars=["student"], value_vars=["math", "english"], var_name="subject", value_name="grade")
df_scores
```

```python
# jadvalni eski holatiga qaytarish
df_scores = pd.pivot(df_scores, index="student", columns="subject", values="grade")
df_scores.reset_index(inplace=True)
```

```python
df_scores
```

```python
df_marks = pd.DataFrame({
    ("Math", "Mid"): [78, 85, 90],
    ("Math", "Final"): [88, 92, 95],
    ("English", "Mid"): [75, 80, 85],
    ("English", "Final"): [82, 87, 90]
}, index=["Ali", "Vali", "Hasan"])
df_marks
```

```python
# faqatgina mathga oid baholarni olish
df_marks['Math']
```

```python
# fanlar bo'yicha o'rtacha baholarni aniqlash
df_marks.groupby(level=0).mean()
```

```python
# ustunlarni soddaroq ko'rinishga keltirish
df_marks.stack(future_stack=True)
```

```python
arrays = [
    ["2023", "2023", "2024", "2024"],
    ["Q1", "Q2", "Q1", "Q2"]
]

index = pd.MultiIndex.from_arrays(arrays, names=("year", "quarter"))

df_revenue = pd.DataFrame(
    {"revenue": [5000, 7000, 6000, 8000]},
    index=index
)
df_revenue
```

```python
# multi indekslar o'rnini almashtirish
df_revenue.swaplevel('year', 'quarter')
```

```python
# yillar bo'yicha saralash
df_revenue = df_revenue.sort_index(level='year')
df_revenue
```

```python
# faqat 2024-yil ma'lumotlarini olish
df_revenue.loc['2024']
```

```python
df_customers = pd.DataFrame({
    "customer_id": [1, 2, 3],
    "name": ["Ali", "Vali", "Hasan"]
})

df_orders = pd.DataFrame({
    "order_id": [101, 102, 103, 104],
    "customer_id": [1, 2, 1, 3],
    "amount": [250, 400, 150, 300]
})
df_customers
```

```python
df_orders
```

```python
# ikkala DF larni customer_id ustuni orqali birlashtirish
cus_order = pd.merge(df_customers, df_orders)
cus_order
```

```python
# har bir mijozning jami buyurtma summasini aniqlash
cus_order.groupby('name')['amount'].sum()
```

```python
df_jan = pd.DataFrame({
    "day": [1, 2, 3],
    "sales": [200, 250, 300]
})

df_feb = pd.DataFrame({
    "day": [1, 2, 3],
    "sales": [220, 270, 310]
})
df_jan
```

```python
df_feb
```

```python
# ikkala DF larni birlashtirish
jan_feb = pd.concat([df_jan, df_feb], keys=['january', 'february'])
jan_feb
```

```python
# oylarni ustun qilish
jan_feb = jan_feb.reset_index(level=0).rename(columns={'level_0': 'month'})
jan_feb.reset_index(drop=True)
```

```python
# sales ustuni bo'yicha tartiblash (kamayish)
jan_feb.sort_values(by='sales', ascending=False)
```

```python
arrays = [
    ["Electronics", "Electronics", "Furniture", "Furniture"],
    ["Q1", "Q2", "Q1", "Q2"]
]

index = pd.MultiIndex.from_arrays(arrays, names=("category", "quarter"))

df_profit = pd.DataFrame(
    {"profit": [5000, 7000, 3000, 4500]},
    index=index
)
df_profit
```

```python
df_teachers = pd.DataFrame({
    "teacher_id": [1, 2, 3],
    "name": ["Ahmad", "Dilshod", "Zarina"]
})

df_classes = pd.DataFrame({
    "class_id": [101, 102, 103],
    "teacher_id": [1, 2, 2],
    "students": [25, 30, 28]
})
df_teachers
```

```python
df_classes
```

```python
teach_class = pd.merge(df_teachers, df_classes)
teach_class
```

```python
# har bir oqituvchi nechta oquvchiga dars berayotganini aniqlash
teach_class.groupby('name')['students'].sum()
```

```python
arrays = [
    ["Asia", "Asia", "Europe", "Europe"],
    ["Uzbekistan", "Japan", "France", "Germany"]
]

index = pd.MultiIndex.from_arrays(arrays, names=("continent", "country"))

df_population = pd.DataFrame(
    {"population_m": [36, 125, 67, 83]},
    index=index
)
df_population
```

```python
df_transport = pd.DataFrame({
    "city": ["Toshkent", "Buxoro"],
    "bus": [120, 80],
    "metro": [300, 0],
    "taxi": [200, 150]
})
df_transport
```

```python
# transport turlari alohida ustun boladigan uzun format yaratish
pd.melt(df_transport,
        value_vars=['bus', 'metro', 'taxi'],
        var_name='transports',
        value_name='prices')
```
