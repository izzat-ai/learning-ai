```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'Store': ['Store_A', 'Store_B', 'Store_A', 'Store_C', 'Store_B', 'Store_A', 'Store_C'],
    'Product': ['Laptop', 'Phone', 'Laptop', 'Tablet', 'Phone', 'Phone', 'Tablet'],
    'Quantity': [5, 8, 3, 6, 4, 7, 2],
    'Price': [1000, 500, 1000, 300, 500, 500, 300]
})
df
```

```python
# Total number of products sold per store
df.groupby('Store')['Quantity'].sum()
```

```python
# Average Price per Store
df.groupby('Store')['Price'].mean()
```

```python
# How many different product types are there for each store?
df.groupby('Store')['Product'].unique()
```

```python
df['total_income'] = df['Quantity']*df['Price']
df
```

```python
# Total revenue per Store
df.groupby('Store')['total_income'].sum()
```

```python
df2 = pd.DataFrame({
    'Department': ['IT', 'HR', 'IT', 'Sales', 'HR', 'IT', 'Sales', 'HR'],
    'Employee': ['Ali', 'Vali', 'Sardor', 'Jasur', 'Kamol', 'Nodir', 'Aziz', 'Bobur'],
    'Salary': [5000, 4000, 5500, 4500, 4200, 6000, 4800, 3900],
    'Experience': [3, 5, 7, 4, 6, 9, 5, 4]
})
df2
```

For each Department:
  1. Highest Salary
  2. Lowest Salary
  3. Average Experience
  4. Number of Employees

```python
df2.groupby('Department').agg({
    'Salary': ['max', 'min'],
    'Experience': 'mean',
    'Employee': 'sum'})
```

```python
df3 = pd.DataFrame({
    'Region': ['North', 'South', 'North', 'East', 'South', 'North', 'East', 'West'],
    'Month': ['Jan', 'Jan', 'Feb', 'Jan', 'Feb', 'Jan', 'Feb', 'Jan'],
    'Sales': [15000, 12000, 18000, 14000, 13000, 16000, 15000, 11000],
    'Customers': [50, 45, 60, 48, 47, 55, 52, 42]
})
df3
```

Group by Region and Month and calculate:
  1. Total sales (Sales sum)
  2. Average number of customers (Customers mean)
  3. How many records are in each group (count)

```python
df3.groupby(['Region', 'Month']).agg({
    'Sales':'sum',
    'Customers':'mean',
    'Month':'count',
})
```

```python
df4 = pd.DataFrame({
    'Student': ['Ali', 'Vali', 'Sardor', 'Jasur', 'Kamol'],
    'Math': [85, 90, 78, 92, 88],
    'Physics': [88, 85, 90, 87, 91],
    'Chemistry': [90, 88, 85, 89, 86],
    'Biology': [87, 91, 88, 90, 89]
})
df4
```

For each student:
1. Total score for all subjects
2. Average score
3. Highest score
4. Lowest score

```python
subjects = ['Math', 'Physics', 'Chemistry', 'Biology']
df4.groupby('Student')[subjects].agg(['sum', 'max', 'min', 'mean'])
```
