```python
import numpy as np
import pandas as pd
```

```python
arrays = [
    ['Apple', 'Apple', 'Apple', 'Samsung', 'Samsung', 'Samsung'],
    [2023, 2022, 2024, 2023, 2022, 2024]
]

index = pd.MultiIndex.from_arrays(arrays, names=['Company', 'Year'])

df1 = pd.DataFrame(
    {'Revenue': [80, 75, 90, 70, 65, 85]},
    index=index
)

df1
```

```python
# asosiy indeks sifatida Year ustunini joylash
df1.swaplevel().sort_index()
```

```python
arrays = [
    ['Europe', 'Europe', 'Asia', 'Asia', 'Europe', 'Asia'],
    ['Germany', 'France', 'Japan', 'China', 'France', 'China']
]

index = pd.MultiIndex.from_arrays(arrays, names=['Region', 'Country'])

df2 = pd.DataFrame(
    {'Sales': [200, 180, 300, 350, 190, 360]},
    index=index
)

df2
```

```python
# country asosiy indeks bo'lishi
df2.swaplevel(0, 1).sort_index()
```

```python
arrays = [
    ['USA', 'USA', 'USA', 'Germany', 'Germany', 'Germany'],
    [2024, 2023, 2022, 2024, 2023, 2022],
    ['Q1', 'Q1', 'Q1', 'Q1', 'Q1', 'Q1']
]

index = pd.MultiIndex.from_arrays(arrays, names=['Country', 'Year', 'Quarter'])

df3 = pd.DataFrame(
    {'Profit': [5.2, 4.8, 4.1, 2.1, 1.9, 1.7]},
    index=index
)
df3
```

```python
# year, country, quater
df3.swaplevel(i=1, j=0).sort_index()
```
