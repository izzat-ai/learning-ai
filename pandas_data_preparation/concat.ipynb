```python
df_yanvar = pd.DataFrame({
    'Mahsulot': ['A', 'B', 'C'],
    'Sotuv': [100, 150, 75]
})

# DF 2: Fevral sotuvlari (Bir xil ustunlar, turli qatorlar)
df_fevral = pd.DataFrame({
    'Mahsulot': ['A', 'D', 'E'],
    'Sotuv': [120, 80, 200]
})

# DF 3: Narxlar (Ustunlar farqli, lekin qatorlar indeksi o'xshash)
df_narx = pd.DataFrame({
    'Narx_USD': [10.5, 25.0, 5.0]
}, index=['A', 'B', 'C'])

# DF 4: Ombor ma'lumoti (Ustunlar farqli, lekin qatorlar indeksi o'xshash)
df_ombor = pd.DataFrame({
    'Ombor_soni': [50, 120, 30]
}, index=['A', 'B', 'C'])
```
