```python
df3 = pd.DataFrame({
    "order_date": ["2024-01-01", "2024-02-10", "2024-02-25", "2024-03-05"],
    "amount": [120, 250, 180, 90]
})
df3
```

```python
# order_date ustunini datetime formatiga o'tkizish
df3['order_date'] = pd.to_datetime(df3['order_date'])
df3.info()
```

```python
# 2024-02-01 dan 2024-02-28 gacha buyurtmalarni ajratish
df3[df3['order_date'].between('2024-02-01', '2024-02-28')]
```

```python
df4 = pd.DataFrame({
    "date": ["2024-01-05", "2024-01-10", "2024-02-03", "2024-02-20", "2024-03-01"],
    "sales": [100, 150, 200, 120, 300]
})
df4
```

```python
df4.info()
```

```python
# date ustunini datetime formatiga o'tkizish
df4['date'] = pd.to_datetime(df4['date'])
```

```python
# month ustunini yaratish
df4['month'] = df4['date'].dt.month
df4
```

```python
# month bo'yicha salesning o'rtacha qiymatini hisoblash
df4.groupby('month')['sales'].mean()
```
