```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Savdo_Xodimi': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Ball': [88, 92, np.nan, 75, 80, np.nan, 95, 78],  # Raqamli ustun (mean bilan to'ldirish)
    'Sotuv_soni': [15, np.nan, 18, 20, np.nan, 22, 25, 14],
    'Tashrif': [5, 6, 4, np.nan, 7, 5, np.nan, 6],
    'Shahar': ['Toshkent', np.nan, 'Buxoro', 'Samarqand', np.nan, 'Toshkent', 'Buxoro', np.nan], # Nomli ustun (ffill/bfill bilan to'ldirish)
    'Mijoz_Turi': ['Kichik', 'Katta', np.nan, 'Katta', 'Kichik', 'Katta', 'Kichik', np.nan],
    'Kanal': [1, 2, 1, 3, 2, np.nan, 1, 3] # Diskret raqamli ustun (median/mode bilan to'ldirish)
}
df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
# ustunlardagi nan qiymatlar soni
df.isnull().sum()
```

```python
# barcha nanlarni 0 - standard qiymat bilan to'ldirish
df.fillna(0)
```

```python
# ball ustunidagi nanlarni 11 bilan to'ldirish
df.fillna({'Ball':11})
```

```python
# sotuv_soni ustunidagi nanlarni 5 va tashrif ustunini nanlarini 4 bilan to'ldirish
df.fillna({'Sotuv_soni':5, 'Tashrif':4})
```

```python
# ball ustunidagi nanlarni , shu ustun o'rtacha qiymati bn to'ldirish
ball_mean = np.mean(df['Ball']).round(1)
df.fillna({'Ball':ball_mean})
```

```python
df
```

```python
# ustunlardagi nan qiymatlarni o'zidan oldingi qiymat bn to'ldirish
df.ffill()
```

```python
# yuqoridagini aksi , nanni o'zidan keyingi qiymat bilan to'ldirish
df.bfill()
```

```python
df['Shahar'].value_counts()
```

```python
df[df['Shahar'] == 'Toshkent']
```

```python
# toshkentning barcha nanlarini bir xil ma'lumot bilan to'ldirish
df[df['Shahar'] == 'Toshkent'] = df[df['Shahar'] == 'Toshkent'].ffill()
```

```python
df
```

```python
# Buxoro ma'lumotlariniyam bir xil qilib to'ldirish
df[df['Shahar'] == 'Buxoro'] = df[df['Shahar'] == 'Buxoro'].bfill()
```

```python
df.info()
```

```python
df2 = df.copy()
df2
```

```python
df2.drop(['Savdo_Xodimi', 'Shahar', 'Mijoz_Turi'], axis='columns', inplace=True)
```

```python
df2
```

```python
# barcha ustunlardagi nanlarni o'zining o'rta qiymati bilan to'ldirish
df2.fillna(df2.mean())
```

```python
df3 = pd.DataFrame({
    "age": [25, np.nan, 30, np.nan, 40],
    "salary": [500, 600, np.nan, 800, np.nan]
})
df3
```

```python
# age ustunidagi NaN’larni o‘rtacha qiymat bilan to‘ldirish
df3['age'].fillna(df3['age'].mean().round(1))
```

```python
# salary ustunidagi NaN’larni median bilan to‘ldirish
df3['salary'].fillna(df3['salary'].median())
```

```python
df4 = pd.DataFrame({
    "gender": ["male", "female", np.nan, "male", np.nan, "female"],
    "city": ["Toshkent", np.nan, "Samarqand", "Toshkent", np.nan, "Toshkent"]
})
df4
```

```python
# gender va city ustunlaridagi NaN qiymatlarni eng ko‘p uchragan qiymat bilan to‘ldirish
df4['gender'].value_counts()
```

```python
df4['city'].value_counts()
```

```python
df4['city'].fillna('Toshkent')
```

```python
df4['gender'].fillna('female')
```

```python
df5 = pd.DataFrame({
    "A": [1, np.nan, 3],
    "B": [np.nan, 5, 6],
    "C": [7, 8, np.nan]
})
df5
```

```python
# A-0, B-1, C-maksimal qiymat bilan to'ldirish
fill_df5 = {
    'A':0,
    'B':1,
    'C':df5['C'].max()
}
df5.fillna(fill_df5)
```
