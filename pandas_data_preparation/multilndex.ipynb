```python
n = 200_000

products = ["iPhone", "Samsung S23", "Redmi Note", "MacBook", "Dell XPS"]
categories = ["Smartphone", "Smartphone", "Smartphone", "Laptop", "Laptop"]

df = pd.DataFrame({
    "order_id": np.arange(1, n+1),
    "product": np.random.choice(products, n),
    "category": np.random.choice(categories, n),
    "price": np.random.randint(100, 2000, n),
    "date": pd.date_range("2024-01-01", periods=n, freq="h")
})
df.head()
```

```python
# narxi > 1500 laptoplarni olish
df[(df['price'] > 1500) & (df['category'] == 'Laptop')]
```

```python
# teroq usul
df.loc[(df['price'] > 1500) & (df['category'] == 'Laptop')]
```

```python
# 2024 yil mart oyidagi barcha buyurtmalar
df.loc[(df['date'].dt.year == 2024) & (df['date'].dt.month == 3)]
```

```python
# iPhone yoki MacBook bo‘lgan barcha buyurtmalar
df.loc[(df['product'] == 'MacBook') | (df['product'] == 'iPhone')]
```

```python
data = {
    "City": ["Toshkent","Toshkent","Samarqand","Samarqand","Fargona","Fargona"],
    "Category": ["Laptop","Phone","Laptop","Phone","Laptop","Phone"],
    "Sales": [1200, 900, 700, 500, 400, 300],
    "Quantity": [3,5,4,2,6,1],
    "Month": ["2024-01","2024-01","2024-02","2024-02","2024-01","2024-02"]
}

df2 = pd.DataFrame(data)
df2
```

```python
df2.groupby('City')['Sales'].agg(['mean', 'sum'])
```

```python
# City va Category bo‘yicha o‘rtacha Quantity chiqarish
df2_mult = df2.groupby(['City', 'Category'])['Quantity'].mean()
```

```python
df2_mult.loc['Toshkent', 'Laptop']
```

```python
df2_mult
```

```python
df2.groupby(['City', 'Month'])['Sales'].sum()
```

```python
# Har bir City va Month bo‘yicha jami sotuvni pivot_table orqali hisoblash
pvt_df2 = pd.pivot_table(df2, index=['City', 'Month'], values='Sales', aggfunc='sum')
pvt_df2
```

```python
pvt2_df2 = pd.pivot_table(df2, index=['City', 'Category'], values='Sales', aggfunc=['min', 'max'])
pvt2_df2
```

```python
pvt2_df2.loc['Toshkent', 'Phone']
```
