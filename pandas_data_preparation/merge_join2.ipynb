```python
import numpy as np
import pandas as pd
```

```python
df_users = pd.DataFrame({
    'user_id': [1, 2, 3, 4],
    'name': ['Ali', 'Vali', 'Hasan', 'Olim']
})

df_orders = pd.DataFrame({
    'user_id': [1, 1, 3, 5],
    'amount': [100, 200, 150, 300]
})
df_users
```

```python
df_orders
```

```python
# faqat ikkala jadvalda ham mavjud boâ€˜lgan foydalanuvchilar chiqsin
pd.merge(df_users, df_orders, how='inner')
```

```python
df_salary = pd.DataFrame({
    'Salary': [1200, 1500, 1800, 2000]
}, index=['A', 'B', 'C', 'D'])

df_bonus = pd.DataFrame({
    'Bonus': [200, 300, 400]
}, index=['A', 'C', 'D'])
df_salary
```

```python
df_bonus
```

```python
# indekslar asosida birlashtirish , maoshi bor lekin bonusi yo'qlar saqlanib qolsin
df_salary.join(df_bonus, how='left')
```

```python
df_companies = pd.DataFrame({
    'CompanyID': [101, 102, 103],
    'Company': ['Google', 'Amazon', 'Microsoft']
})

df_finance = pd.DataFrame({
    'CompanyID': [101, 101, 102, 103, 103],
    'Year': [2023, 2024, 2024, 2023, 2024],
    'Revenue': [500, 600, 450, 550, 650]
})
df_companies
```

```python
df_finance
```

```python
new_df = pd.merge(df_companies, df_finance, how='inner')
new_df
```

```python
main_df = new_df.drop('CompanyID', axis=1)
main_df
```
