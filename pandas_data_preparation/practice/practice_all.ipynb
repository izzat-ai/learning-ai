```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)
```

```python
df1 = pd.DataFrame({
    "product": ["Phone","Laptop","Tablet","Monitor","Keyboard"],
    "price": [1200, None, 600, None, 80],
    "quantity": [5, 3, None, 4, 10]
})
df1
```

```python
# Determine how many blank values ​​are in which columns
df1.isnull().sum()
```

```python
# Check if total revenue is being calculated incorrectly due to empty values
df1['price'] * df1['quantity']
```

```python
df1 = df1.dropna().copy()
df1
```

```python
# recalculate total revenue based on the cleaned dataset
df1['revenue'] = df1['price'] * df1['quantity']
df1
```

```python
df3 = pd.DataFrame({
    "gender": ["M","F","M","F","F","M"],
    "score": [78, 88, 90, 67, 95, 80]
})
df3
```

```python
# Convert the values ​​in the Gender column to full word form
df3['gender'] = df3['gender'].replace({'M':'male', 'F':'female'})
df3
```

```python
# Change values ​​below 80 in the Score column to “Low”, otherwise to “High”
df3['score'].map(lambda x: 'low' if x<80 else 'high')
```

```python
df4 = pd.DataFrame({
    "age": [15, 22, 37, 45, 18, 29, 60]
})
df4
```

```python
# Divide the ages into 3 intervals and count how many people are in each interval
bins = [14, 24, 44, 65]
df4['groups'] = pd.cut(df4['age'], bins, labels=['teenagers', 'middle-age', 'adults'])
df4
```

```python
df4['groups'].value_counts()
```

```python
df6 = pd.DataFrame({
    "income": [2000, 3000, 15000, 2500, 4000, 50000, 2700]
})
df6
```

```python
# sort data
df6_sort = df6.sort_values(by='income', ascending=True)
df6_sort
```

- `IQR = Q3 - Q1`
- `Q1` - 25% , `Q3` - 75%

```python
df6_sort.describe()
```

```python
# Outlier detection using the IQR method
q1 = 2600
q3 = 9500
iqr = q3 - q1
up = q3 + (1.5 * iqr)
low = q1 - (1.5 * iqr) # If the answer is a negative number, it is taken as 0
print("upper: ", up)
print('lower: ', low)
```

- There are only 50000 above 19850. Therefore, the outlier in our data is only 50000.

```python
# determining the mean value together with the outlier
df6_sort['income'].mean()
```

```python
# Removing extreme values
df6_sort[df6_sort['income'] > 19850]
```

```python
df6_sort = df6_sort.drop(5)
df6_sort
```

```python
# determining the average value without outliers
df6_sort['income'].mean()
```

```python
df8 = pd.DataFrame({
    "date": ["2025-01-05","2025-02-10","2025-03-15","2025-04-20"],
    "sales": [1000,1500,1200,1800]
})
df8
```

```python
# Convert a date column to time format
df8['date'] = pd.to_datetime(df8['date'])
df8.info()
```

```python
# separate the year as a separate column
df8['year'] = df8['date'].dt.year
df8
```

```python
# Summarize sales by month
df8['month'] = df8['date'].dt.month_name()
df8.groupby('month')['sales'].sum()
```
