```python
import numpy as np
import pandas as pd
```

```python
df_sales = pd.DataFrame({
    'Region': ['East', 'East', 'West', 'West', 'North', 'North'],
    'Product': ['Phone', 'Laptop', 'Phone', 'Laptop', 'Phone', 'Laptop'],
    'Year': [2022, 2022, 2022, 2022, 2022, 2022],
    'Sales': [120, 200, 150, 180, 100, 160]
})
df_sales
```

```python
values = df_sales.groupby('Region')['Sales'].mean()
values
```

```python
df_sales.pivot_table(df_sales, index='Region', columns=['Product', 'Year'])
```

```python
df_students = pd.DataFrame({
    'Class': ['A', 'A', 'A', 'B', 'B', 'B'],
    'Subject': ['Math', 'Physics', 'English', 'Math', 'Physics', 'English'],
    'Gender': ['M', 'F', 'M', 'F', 'M', 'F'],
    'Score': [78, 85, 90, 88, 76, 92]
})
df_students
```

```python
pd.pivot_table(df_students, index=['Class', 'Subject'], columns='Gender', values='Score')
```

```python
df_company = pd.DataFrame({
    'Company': ['A', 'A', 'A', 'B', 'B', 'B'],
    'Quarter': ['Q1', 'Q2', 'Q3', 'Q1', 'Q2', 'Q3'],
    'Year': [2023, 2023, 2023, 2023, 2023, 2023],
    'Revenue': [1200, 1500, 1700, 1100, 1400, 1600],
    'Cost': [800, 900, 1000, 700, 850, 950]
})
df_company
```

```python
df_company['profit'] = df_company['Revenue'] - df_company['Cost']
df_company
```

```python
pd.pivot_table(df_company, index=['Company'], columns=['Quarter'], values='profit')
```
