```python
def vaqtni_aniqla(x):
  x = x.hour
  if 6<=x<=12:
    return "morning"
  elif 12<=x<=18:
    return "afternoon"
  elif 18<=x<=23:
    return "evening"
  else:
    return "night"

s12.apply(vaqtni_aniqla)
```
