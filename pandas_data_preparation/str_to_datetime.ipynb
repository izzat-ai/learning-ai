```python
import numpy as np
import pandas as pd
import datetime as dt
```

```python
data = {
    'Standart_Sana': ['2023-11-20', '2023-12-05', '2024-01-15', '2024-02-28',
                      '2024-03-10', '2024-04-12', '2024-05-01', '2024-06-30'],

    'Maxsus_Sana_1': ['08/25/2023', '09/10/2023', '10/18/2023', '11/05/2023',
                      '12/01/2023', '01/01/2024', '02/14/2024', '03/08/2024'],

    'Maxsus_Sana_2': ['230101', '230315', '230520', '230712',
                      '230901', '231125', 'Boshqa', '240105']
}
df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
s = pd.Series([
    "2004/03/15",
    "1999/12/01",
    "2001/07/20",
    "2003/11/05"
])
s
```

```python
# datetime formatiga o'tkizish
s = pd.to_datetime(s)
s
```

```python
# har bir sanadan yil, oy, kunni ajratish
year = s.dt.year
month = s.dt.month
day = s.dt.day
```

```python
# hafta kunini aniqlash
s.dt.weekday
```

```python
df2 = pd.DataFrame({
    "order_id": [101, 102, 103, 104],
    "order_date": ["10-02-2024", "15-03-2024", "18-03-2024", "01-01-2024"]
})
df2
```
