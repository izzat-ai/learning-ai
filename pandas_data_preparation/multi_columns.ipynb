```python
import numpy as np
import pandas as pd
```

```python
df_finance = pd.DataFrame(
    {
        ('Revenue', 'Q1'): [1000, 1500, 1200],
        ('Revenue', 'Q2'): [1100, 1600, 1300],
        ('Cost', 'Q1'): [600, 900, 700],
        ('Cost', 'Q2'): [650, 950, 750],
    },
    index=['Company_A', 'Company_B', 'Company_C']
)
df_finance
```

```python
profit = df_finance['Revenue'] - df_finance['Cost']
profit
```

```python
profit.sum(axis=1).to_frame(name='Profit')
```

```python
df_scores = pd.DataFrame(
    {
        ('Math', 'Midterm'): [80, 75, 90],
        ('Math', 'Final'): [85, 78, 92],
        ('Physics', 'Midterm'): [70, 88, 84],
        ('Physics', 'Final'): [75, 90, 86],
    },
    index=['Student_1', 'Student_2', 'Student_3']
)
df_scores
```

```python
df_scores.groupby(level=0).mean().mean(axis=0)
```

```python
df_scores_st = df_scores.stack(future_stack=True, level=0)
df_scores_st
```

```python
df_scores_st.groupby(level=1).mean()
```

```python
df_sales = pd.DataFrame(
    {
        ('Laptop', '2023'): [120, 150, 100],
        ('Laptop', '2024'): [140, 170, 130],
        ('Phone', '2023'): [200, 220, 210],
        ('Phone', '2024'): [230, 250, 240],
    },
    index=['Region_A', 'Region_B', 'Region_C']
)
df_sales
```

```python
sales2023 = df_sales.xs('2023', level=1, axis=1).sum(axis=1)
sales2023
```

```python
sales2024 = df_sales.xs('2024', level=1, axis=1).sum(axis=1)
sales2024
```

```python
(sales2024 - sales2023).to_frame(name='growth')
```
