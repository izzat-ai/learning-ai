```python
import numpy as np
import pandas as pd
```

```python
df_mijoz = pd.DataFrame({
    'Mijoz_ID': [101, 102, 103, 104, 105],
    'Ism': ['Ali', 'Vali', 'Guli', 'Sami', 'Nuri'],
    'Shahar': ['Toshkent', 'Samarqand', 'Buxoro', 'Toshkent', 'Xiva']
})

df_xarid = pd.DataFrame({
    'Tranzaksiya_ID': [5001, 5002, 5003, 5004, 5005, 5006, 5007],
    'Mijoz_ID': [101, 103, 103, 105, 106, 101, 109],
    'Summa': [150.0, 45.5, 210.0, 75.0, 50.0, 300.0, 10.0]
})
df_mijoz
```

```python
df_xarid
```

- yuqoridagi ikki DataFramelarni `Mijoz_ID` ustuni bog'lab turibdi

```python
pd.merge(df_mijoz, df_xarid)
```

- `Mijoz_ID` ustunidagi qiymatlar ikkala DF dayam borlarini birlashtirdi

```python
# yangi dflar yaratish
df1 = df_mijoz.set_index('Mijoz_ID')
df2 = df_xarid.set_index('Mijoz_ID')
df1
```

```python
df2
```

- df1 va df2 larni indekslari bo'yicha birlashtirish
1. `df_mijoz`dagi barcha ma'lumotlar qoladi
2. `df_xarid`dagi ma'lumotlar bo'yicha birlashtirish
3. barcha ma'lumotlar bo'yicha birlashtirish

```python
# N1
df1.join(df2, how='left')
```

```python
# N2
df1.join(df2, how='right')
```

```python
# N3 - barcha ma'lumotlarni birlashtirish
df1.join(df2, how='outer')
```

- *dataframelar sun'iy intellekt tomonidan yaratilmoqda*

```python
df_talaba = pd.DataFrame({
    'Talaba_ID': [1, 2, 3, 4, 5],
    'Ism': ['Ali', 'Vali', 'Guli', 'Sami', 'Nuri'],
    'Fakultet': ['IT', 'Iqtisod', 'IT', 'Moliya', 'Iqtisod']
})

df_baholar = pd.DataFrame({
    'Talaba_ID': [1, 3, 5, 6, 7],
    'Fan': ['Python', 'SQL', 'DataViz', 'Python', 'Web'],
    'Baho': [90, 85, 95, 70, 80]
})

df_ustoz = pd.DataFrame({
    'Ism': ['Anvar', 'Diyora', 'Jahongir'],
    'Fan': ['Python', 'SQL', 'DataViz']
}, index=[10, 20, 30])
df_ustoz.index.name = 'O‘qituvchi_ID'
```

```python
df_talaba
```

```python
# talaba va baholar DF larini birlashtirish , barcha ma'lumotlar qo'shilishi kerak
df_talaba.merge(df_baholar, how='left', on='Talaba_ID')
```

```python
df_baholar[['Fan', 'Baho']].merge(df_ustoz['Fan'].reset_index(), on='Fan', how='inner').set_index("O‘qituvchi_ID")
```

```python
df_baholar.merge(df_talaba)
```
