```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(
    {
        'Sales': [1200, 900, 1500, 1100, 800, 1600],
        'Profit': [200, 150, 300, 180, 100, 350]
    },
    index=pd.MultiIndex.from_tuples(
        [
            ('Asia', 'Uzbekistan'),
            ('Asia', 'Kazakhstan'),
            ('Asia', 'Japan'),
            ('Europe', 'Germany'),
            ('Europe', 'France'),
            ('Europe', 'UK'),
        ],
        names=['Region', 'Country']
    )
)
df
```

```python
# har bir region bo'yicha salesning umiymisini aniqlang
reg_sales = df.groupby(level=0)['Sales'].sum()
reg_sales
```

```python
# qaysi regionning umumiy salesi ko'pligini aniqlash
reg_sales.idxmax()
```

```python
df_emp = pd.DataFrame(
    {
        'Employees': [50, 40, 60, 30, 45, 55],
        'Avg_Salary': [700, 650, 900, 550, 600, 800]
    },
    index=pd.MultiIndex.from_tuples(
        [
            ('TechCorp', 'IT'),
            ('TechCorp', 'HR'),
            ('TechCorp', 'Finance'),
            ('MegaSoft', 'IT'),
            ('MegaSoft', 'HR'),
            ('MegaSoft', 'Finance'),
        ],
        names=['Company', 'Department']
    )
)
df_emp
```

```python
# har bir kampaniyada qaysi Department eng ko'p xodimga ega ekanini aniqlash
filt_dfemp = df_emp.groupby(['Company', 'Department'])['Employees'].sum()
filt_dfemp.groupby(level=0).idxmax()
```

```python
df_gdp = pd.DataFrame(
    {
        'GDP': [5.1, 5.3, 5.7, 4.8, 4.9, 5.2, 3.3, 3.4, 3.6]
    },
    index=pd.MultiIndex.from_tuples(
        [
            ('USA', 2021), ('USA', 2022), ('USA', 2023),
            ('Germany', 2021), ('Germany', 2022), ('Germany', 2023),
            ('Japan', 2021), ('Japan', 2022), ('Japan', 2023),
        ],
        names=['Country', 'Year']
    )
)
df_gdp
```

```python
filt_dfgdp = df_gdp.loc[slice(None), [2021, 2022, 2023], :]
filt_dfgdp
```

```python
# har bir davlat bo'yicha GDP o'rtacha qiymatini aniqlash
gdp_mean = filt_dfgdp.groupby(level=0)['GDP'].mean()
gdp_mean
```

```python
# o'rtacha GDP eng yuqori bo'lgan davlatni aniqlash
gdp_mean.idxmax()
```

```python
df = pd.DataFrame({
        "Company": ["TechCorp","TechCorp","TechCorp",
                    "MegaSoft","MegaSoft","MegaSoft",
                    "DataWorld","DataWorld","DataWorld"],

        "Year": [2021, 2022, 2023,
                 2021, 2022, 2023,
                 2021, 2022, 2023],

        "Revenue": [120, 150, 180,
                    90, 110, 140,
                    70, 85, 100],

        "Profit": [20, 25, 30,
                   10, 12, 18,
                   8, 10, 13]})
df
```

```python
# kompaniya va yil bo'yicha irarxik indekslash
df_mult = pd.pivot_table(df, index=['Company', 'Year'])
df_mult
```

```python
# har bir kompaniyaning eng yuqori revenue ko'rsatkichiga ega yilni aniqlash
df_mult.groupby(level='Company')['Revenue'].idxmax()
```

```python
df_mult
```

```python
# har bir kompaniya bo'yicha o'rtacha profit qiymatini aniqlash
mean_profit = df_mult.groupby(level=0)['Profit'].mean()
mean_profit
```

```python
# qaysi kompaniya eng yuqori o'rtacha foyda qiganini aniqlash
mean_profit.idxmax()
```
