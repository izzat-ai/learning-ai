```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'Student': ['Ali', 'Ali', 'Vali', 'Vali'],
    'Subject': ['Math', 'Physics', 'Math', 'Physics'],
    'Score': [85, 90, 78, 88]
})
df1
```

```python
# indeksda - talabalar, qiymatlar - ballar, ustunlar - fanlar
df1.pivot(index='Student', columns='Subject', values='Score')
```

```python
df2 = pd.DataFrame({
    'City': ['Tashkent', 'Tashkent', 'Samarkand', 'Samarkand'],
    'Year': [2023, 2024, 2023, 2024],
    'Population': [3.0, 3.2, 0.8, 0.85]
})
df2
```

```python
# yillar - ustunlarda, aholi-qiymatlarda
df2.pivot(index='City', columns='Year', values='Population')
```

```python
df3 = pd.DataFrame({
    'Product': ['Phone', 'Phone', 'Laptop', 'Laptop'],
    'Quarter': ['Q1', 'Q2', 'Q1', 'Q2'],
    'Revenue': [5000, 6500, 7000, 8000]
})
df3
```

```python
# mahsulotlar - indeksda, revenue-qiymatlarda
df3.pivot(index='Product', columns='Quarter', values='Revenue')
```
