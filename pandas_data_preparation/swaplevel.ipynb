```python
import numpy as np
import pandas as pd
```

- dataframelar sun'iy intellekt tomonidan yaratilmoqda

```python
data = [
    ('AQSH', 2021, 23.32),
    ('AQSH', 2022, 25.46),
    ('AQSH', 2023, 27.37),
    ('Xitoy', 2021, 17.82),
    ('Xitoy', 2022, 18.10),
    ('Xitoy', 2023, 19.80),
    ('Yaponiya', 2021, 4.94),
    ('Yaponiya', 2022, 4.30),
    ('Yaponiya', 2023, 4.20),
]

index_tuples = [(d[0], d[1]) for d in data]

multi_index = pd.MultiIndex.from_tuples(
    index_tuples,
    names=['Davlat', 'Yil']
)

df = pd.DataFrame(
    [d[2] for d in data],
    index=multi_index,
    columns=['YaIM (trln $)']
)
df
```

```python
# irarxik indekslar o'rnini almashtirish
df.swaplevel('Davlat', 'Yil')
```

```python
# indekslar bo'yicha o'sish yoki kamayish tartibida tartiblash
df.sort_index(level=1, ascending=False)
```

```python
multi_index = pd.MultiIndex.from_tuples(
    [('AQSH', 2021), ('AQSH', 2022), ('Xitoy', 2021), ('Xitoy', 2022)],
    names=['Davlat', 'Yil']
)

df_swap = pd.DataFrame(
    [23.32, 25.46, 17.82, 18.10],
    index=multi_index,
    columns=['YaIM (trln $)']
)
df_swap
```

```python
# yil va davlat indekslarini o'rnini almashtirish
new_df = df_swap.swaplevel('Davlat', 'Yil')
new_df
```

```python
# 0-darajadagi indeks bo'yicha tartiblash
new_df = new_df.sort_index(level=0)
new_df
```
