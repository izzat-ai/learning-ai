```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Rating': ['A', 'B', 'C', 'A', 'B', 'C'],

    'Narx': [100.50, 250.00, 75.90, 300.00, 150.25, 99.99],

    'Status': ['Yuqori', 'Past', 'Yuqori', "O'rta", 'Past', 'Yuqori'],

    'Sotuvlar_Soni': [150, 45, 210, 88, 30, 160],

    'Hudud': ['Shimol', 'Janub', "G'arb", 'Sharq', 'Shimol', 'Janub']
}
df = pd.DataFrame(data)
df
```

```python
# status qiymatlarini map yordamida sonlarga o'zgartirish
# yuqori - 1
# o'rta - 0.5
# past - 0
df['Status'] = df['Status'].map({'Yuqori':1, "O'rta":0.5, 'Past':0})
```

```python
df
```

```python
# sotuvlar_soni qiymatlarini 2 ga ko'paytirish
df['Sotuvlar_Soni'].map(lambda x: x*2)
```

```python
# status ustuni qiymatlarini kvadratlarini hisoblash
df['Status'].apply(lambda x: x**2)
```

```python
# sotuvlar_soni umumiy yig'indisini hisoblash
df[['Sotuvlar_Soni']].apply(lambda col: col.sum())
```

```python
df2 = pd.DataFrame({
    "a": [1, 2, 3],
    "b": [10, 20, 30]
})
df2
```

```python
# har bir qatorni yig'indisini hisoblash
df2.apply(lambda row: row.sum(), axis=1)
```

```python
df4 = pd.DataFrame({
    "first": ["Ali", "Vali", "Hasan"],
    "last": ["Karimov", "Toshpulatov", "Rahimov"]
})
df4
```

```python
# full_name ustunini yaratish
df4['full_name'] = df4.apply(lambda row: row['first']+' '+row['last'], axis=1)
df4
```

```python
df5 = pd.DataFrame({
    "product": ["A", "B", "C", "D"],
    "price": [100, 250, 80, 400],
})
df5
```

```python
# narxi 200 dan katta bo‘lsa “High”, kichik bo‘lsa “Low” , level ustunini yaratish

# solishtiradigan funksiya yaratish
def tekshir(x):
  if x>200:
    return "High"
  else:
    return "Low"

df5['level'] = df5['price'].apply(tekshir)
df5
```
