```python
import numpy as np
import pandas as pd
from datetime import date, timedelta
```

```python
start_date = date(2024, 1, 1)

# DataFrame uchun ma'lumotlar
data = {
    'Ism': [f'Xodim_{i}' for i in range(1, 11)],
    'Yoshi': [25, 27, 22, 30, 24, 28, 23, 31, 26, 29],
    'Guruh': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'A', 'C'],
    'Baho': [4.5, 3.8, 4.9, 3.5, 4.1, 4.7, 3.9, 4.2, 4.6, 4.0],
    'Sana': [start_date + timedelta(days=i * 10) for i in range(10)]
}
df = pd.DataFrame(data)
df
```

```python
df.sample(frac=1, random_state=42)
```

```python
randlist = np.random.permutation(len(df))
randlist
```

```python
df.take(randlist)
```
