```python
import numpy as np
import pandas as pd
```

```python
df_jan = pd.DataFrame({
    'Product': ['Phone', 'Laptop', 'Tablet'],
    'Sales': [120, 80, 60]
})

df_feb = pd.DataFrame({
    'Product': ['Phone', 'Laptop', 'Tablet'],
    'Sales': [150, 90, 70]
})
df_jan
```

```python
df_feb
```

```python
# ikkala DF larni pastga qarab qo'shish , indekslar tartiblangan bo'lsin
pd.concat([df_jan, df_feb], ignore_index=True)
```

```python
df_info = pd.DataFrame({
    'Name': ['Ali', 'Vali', 'Hasan'],
    'Score': [85, 90, 78]
})

df_extra = pd.DataFrame({
    'Age': [20, 21, 22],
    'City': ['Tashkent', 'Samarkand', 'Bukhara']
})
df_info
```

```python
df_extra
```

```python
# ikkala DF ni yonma yon birlashtirish
pd.concat([df_info, df_extra], axis=1)
```
