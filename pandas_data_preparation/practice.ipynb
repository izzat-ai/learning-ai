```python
import numpy as np
import pandas as pd
import datetime as dt
```

```python
df1 = pd.DataFrame({
    "name": ["Ali", "Vali", "Hasan", None],
    "age": [23, None, 19, 21],
    "salary": [3000, 3500, None, 4000]
})
df1
```

```python
# age ustunidagi nanni o'rtacha yosh bilan to'ldirish
df1['age'].fillna(df1['age'].mean())
```

```python
# salary ustunidagi NaN qiymatni median bilan to‘ldirish
df1['salary'].fillna(df1['salary'].median())
```

```python
# name ustunidagi NaN qiymatni "NoName" bilan to‘ldirish
df1['name'].fillna('NoName')
```

```python
# yangi DF ga barcha nanlarni 0 bilan to'ldirish
df1_2 = df1.fillna(0)
```

```python
df10 = pd.DataFrame({
    "time": pd.to_datetime(["2024-04-01 05:00","2024-04-01 13:00","2024-04-01 20:30"])
})
df10
```

```python
print(type(df10['time'].dt.hour))
```

```python
def vatqni_top(x):
  x = x.hour
  if 6<=x<=12:
    return "morning"
  elif 12<=x<=18:
    return "afternoon"
  elif 18<=x<=24:
    return "evening"
  else:
    return "night"

df10['time'].apply(vatqni_top)
```
