```python
import numpy as np
import pandas as pd
import datetime as dt
```

```python
df1 = pd.DataFrame({
    "name": ["Ali", "Vali", "Hasan", None],
    "age": [23, None, 19, 21],
    "salary": [3000, 3500, None, 4000]
})
df1
```

```python
# age ustunidagi nanni o'rtacha yosh bilan to'ldirish
df1['age'].fillna(df1['age'].mean())
```

```python
# salary ustunidagi NaN qiymatni median bilan to‘ldirish
df1['salary'].fillna(df1['salary'].median())
```

```python
# name ustunidagi NaN qiymatni "NoName" bilan to‘ldirish
df1['name'].fillna('NoName')
```

```python
# yangi DF ga barcha nanlarni 0 bilan to'ldirish
df1_2 = df1.fillna(0)
```

```python
df1_2
```

```python
df2 = pd.DataFrame({
    "product": ["A", "A", "B", "C", "C", "C"],
    "price": [100, 100, 150, 200, 200, 250]
})
df2
```

```python
# qaysi qatorlar takrorlangan :
df2[df2.duplicated()]
```

```python
# Takroriy qatorlarni to‘liq o‘chirib yuborish
df2.drop_duplicates()
```

```python
# Faqat birinchi uchraganini qoldirib, boshqalarni o‘chirib yuborish
df2.drop_duplicates(keep='first')
```

```python
# Faqat oxirgi uchraganini qoldirib, boshqalarni o‘chirib yuborish
df2.drop_duplicates(keep='last')
```
