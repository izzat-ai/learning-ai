```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Region': ['Shimol', 'Janub', 'Shimol', 'G\'arb', 'Janub', 'Sharq', 'Shimol', 'G\'arb', 'Sharq', 'Janub'],

    'Jins': ['Erkak', 'Ayol', 'Ayol', 'Erkak', 'Ayol', 'Erkak', 'Erkak', 'Ayol', 'Ayol', 'Erkak'],

    'Sotuv_Summasi': [1500, 2100, 1200, 3000, 1800, 500, 2500, 4000, 1000, 2200],

    'Mijozlar_Soni': [15, 20, 10, 35, 18, 5, 22, 40, 8, 25],

    'Komissiya': [150, 210, np.nan, 300, 180, 50, 250, np.nan, 100, 220]
}
df = pd.DataFrame(data)
df
```

```python
# jins bo'yicha mijozlar_sonini aniqlash
df.groupby('Jins')['Mijozlar_Soni'].sum()
```

```python
# mintaqa bo'yicha mijozlar sonini aniqlash
df.groupby('Region')['Mijozlar_Soni'].sum()
```

```python
# jins bo'yicha sotuvlar sonini min, max, mean va umumiysini aniqlash
df.groupby('Jins')['Sotuv_Summasi'].agg(['min', 'max', 'mean', 'sum', 'count'])
```

```python
# mintaqa va jins bo'yicha mijozlar sonini aniqlash
df.groupby(['Region', 'Jins'])['Mijozlar_Soni'].sum()
```

```python
# mintaqa bo'yicha o'rtacha sotuv summasi
df.groupby('Region')['Sotuv_Summasi'].mean()
```
