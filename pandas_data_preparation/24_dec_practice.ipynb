```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "Name": ["Ali", "Vali", "Hasan", "Husan"],
    "Age": [20, np.nan, 22, np.nan],
    "Score": [85, 90, np.nan, 88]
})
df1
```

```python
df1['Age'] = df1['Age'].fillna(df1['Age'].mean())
df1
```

```python
df1['Score'] = df1['Score'].fillna(df1['Score'].mean())
df1
```

```python
df7 = pd.DataFrame({
    "Age": [7, 15, 23, 35, 62]
})
df7
```

```python
df8 = pd.DataFrame({
    "Department": ["IT", "IT", "HR", "HR", "Sales"],
    "Salary": [1200, 1500, 900, 1000, 1100]
})
df8
```

```python
# har bir bo'lim bo'yicha o'rtacha qiymatni aniqlash
df8.groupby('Department')['Salary'].mean()
```

```python
# har bir bo'limda nechta xodim borligini aniqlash
df8.value_counts('Department')
```

```python
# har bir bo'lim bo'yicha eng yuqori maosh
df8.groupby('Department')['Salary'].max()
```
