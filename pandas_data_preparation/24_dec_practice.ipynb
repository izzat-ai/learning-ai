```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "Name": ["Ali", "Vali", "Hasan", "Husan"],
    "Age": [20, np.nan, 22, np.nan],
    "Score": [85, 90, np.nan, 88]
})
df1
```

```python
df1['Age'] = df1['Age'].fillna(df1['Age'].mean())
df1
```

```python
df1['Score'] = df1['Score'].fillna(df1['Score'].mean())
df1
```

```python
df3 = pd.DataFrame({
    "Product": ["Phone", "Laptop", "Phone", "Tablet", "Phone"],
    "Price": [500, 1200, 500, 800, 500]
})
df3
```

```python
# takrorlangan qatorlarni tashavorish
df3.drop_duplicates()
```

```python
# product ustuni bo'yicha takrorlangan qatorlarni tashavorish
df3.drop_duplicates(subset='Product')
```
