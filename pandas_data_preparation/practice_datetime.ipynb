```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'Employee': ['John Smith', 'Alice Johnson', 'Bob Williams'],
    'Join_Date': ['2020-03-15', '2019-07-22', '2021-11-08'],
    'Salary': [5000, 6500, 4800]
})
df
```

```python
df.info()
```

```python
df['Join_Date'] = pd.to_datetime(df['Join_Date'])
df.info()
```

```python
import datetime as dt
```

```python
# Create separate columns by separating the year, month, and day from the Join_Date column
df['Year'] = df['Join_Date'].dt.year
df['Month'] = df['Join_Date'].dt.month
df['Day'] = df['Join_Date'].dt.day
df
```

```python
df1 = pd.DataFrame({
    'Transaction_Date': ['2024-01-05', '2024-01-06', '2024-01-07', '2024-01-13', '2024-01-14'],
    'Amount': [1500, 2000, 1800, 2200, 1900],
    'Customer': ['C001', 'C002', 'C001', 'C003', 'C002']
})
df1
```

```python
df1.info()
```

```python
df1['Transaction_Date'] = pd.to_datetime(df1['Transaction_Date'])
df1.info()
```

```python
# Determine which day of the week each transaction occurred on
df1['Transaction_Day'] = df1['Transaction_Date'].dt.day_name()
df1
```

```python
df2 = pd.DataFrame({
    'Event': ['Conference 2024', 'Workshop', 'Meeting'],
    'Start': ['2024-06-10', '2024-07-15', '2024-08-20'],
    'End': ['2024-06-15', '2024-07-18', '2024-08-22']
})
df2
```

```python
df2.info()
```

```python
df2['Start'] = pd.to_datetime(df2['Start'])
df2['End'] = pd.to_datetime(df2['End'])
df2.info()
```

```python
# Calculate how many days each event will last
df2['Continue'] = np.abs((df2['Start'] - df2['End']).dt.days)
df2
```

```python
# Specify the start date in the format "10-Jun-2024"
df2['Start'] = df2['Start'].dt.strftime('%d-%b-%Y')
```
