```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Yoshi': [25, 30, 45, 52, 21, 38, 60, 29],

    'Oylik_Daromad': [3500, 7800, 4100, 9500, 2900, 5600, 12000, 6200],

    'Ish_Staji': [3, 7, 15, 20, 1, 10, 25, 5],

    'Kompaniya_Reytingi': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B'],

    'Qarz_Baho': [0.65, 0.45, 0.80, 0.20, 0.90, 0.35, 0.15, 0.50],

    'Manzil_Kodi': [1001, 1002, 1001, 1003, 1002, 1003, 1001, 1002]
}
df = pd.DataFrame(data)
df
```

```python
df.rename(columns=str.lower, inplace=True)
df
```

```python
df['yoshi'].value_counts()
```

```python
# yoshi bo'yicha guruhlash
yosh_bins = [20, 45, 60]
df['guruh_yosh'] = pd.cut(df['yoshi'], yosh_bins, labels=['yosh', 'katta'])
df
```

```python
df['guruh_yosh'].value_counts()
```

```python
df['guruh_yosh'] = pd.qcut(df['yoshi'], q=3, precision=0)
df['guruh_yosh'].value_counts()
```
