```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
import random
```

```python
# ushbu dataset sun'iy intellektdan olindi

data = {
    'Transaction_ID': [1001, 1002, 1003, 1004, 1005, 1006, 1007, 1008, 1009, 1010, 1011, 1012, 1013, 1014, 1015],
    'Mijoz_ID': ['C101', 'C102', 'C103', 'C101', 'C104', 'C105', 'C102', 'C106', 'C107', 'C108', 'C105', 'C103', 'C109', 'C110', 'C101'],
    'Sana': ['2025-01-10', '2025/01/15', '2025-01-10', '2025-02-01', '2025-02-05', '2025-02-10', '2025-01-15', '2025-03-01', '2025-03-05', '2025-03-10', '2025-02-10', '2025-01-10', '2025-03-15', '2025-04-01', '2025-02-01'],
    'Kategoriya': ['Elektronika', 'Kiyim', 'Elektronika', 'Aksessuar', 'Kiyim', 'Elektronika', 'Kiyim', 'Aksessuar', 'Kiyim', 'Elektronika', 'Kiyim', 'Elektronika', 'Aksessuar', 'Kiyim', 'Elektronika'],
    'Summa': [150.00, 45.50, 150.00, 12.00, 60.00, 250.00, 45.50, 18.00, 80.00, 150.00, 60.00, 150.00, 15.00, np.nan, 150.00],
    'Reyting': [4, 5, 4, 3, 5, 2, 5, 3, 4, 4, 5, 4, 3, 5, 4]
}

df = pd.DataFrame(data)

df.loc[15] = ['1003', 'C103', '2025-01-10', 'Elektronika', 150.00, 4]
df.loc[16] = [1016, 'C111', 'Xato_sana', 'Kiyim', 55.00, 4]
df.loc[17] = [1017, 'C102', '2025-04-05', 'kiyim', 45.50, 5]
```

- `sana` ustuniga ishlov berish

```python
df['Sana'].value_counts()
```

```python
df[df['Sana'] == '2025/01/15'].index
```

```python
df[df['Sana'] == 'Xato_sana'].index
```

```python
df.loc[1, 'Sana'] = '2025-01-15'
df.loc[16, 'Sana'] = np.nan
```

```python
df['Sana'].value_counts()
```

```python
df['Sana'] = pd.to_datetime(df['Sana'])
df.info()
```

- `Summa` ustuniga ishlov berish

```python
df['Summa'].isnull().sum()
```

```python
df['Summa'].mean()
```

```python
df['Summa'].fillna(df['Summa'].mean(), inplace=True)
```

```python
df['Summa']
```

- `Kategoriya` ustuniga ishlov berish

```python
df['Kategoriya'].value_counts()
```

```python
df['Kategoriya'] = df['Kategoriya'].str.lower()
```

```python
df['Kategoriya'].value_counts()
```

```python
df[df[['Mijoz_ID', 'Kategoriya']].duplicated()]
```

```python
# mijoz idsi va kategoriyasi bir xil bo'lgan ma'lumotlarni o'chirish
df = df.drop_duplicates(subset=['Mijoz_ID', 'Kategoriya'])
```

- `kategoriya` bo'yicha guruhlash

```python
df.groupby('Kategoriya')[['Summa', 'Reyting']].agg(['sum', 'mean'])
```

- har bir mijozning qilgan haridlarini umumiy summasini aniqlash

```python
umum_sum = df.groupby('Mijoz_ID')['Summa'].sum()
umum_sum
```

```python
umum_sum[umum_sum>200.00]
```
