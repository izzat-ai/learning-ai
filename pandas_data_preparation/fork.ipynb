Ushbu bo'limda, melb_data.csv datasetidagi barcha NaN (null) qiymatlar tahlil qilish uchun asoslangan va mantiqan to'g'ri qiymatlar bilan to'ldirildi. Barcha o'zgarishlar `inplace=False` rejimida amalga oshirildi .

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/izzat-ai/learning-ai/refs/heads/main/pandas_data_preparation/melb_data.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

- `date` ustuni matn ko'rinishida ekan
- `bedroom2, bathroom, car` ustunlari floatda ekan (int bo'lishi kerak)
- `yearbuilt` ham int bo'lishi kerak
- `Propertycount` int bo'lishi kerak

```python
df['Date'].sample(10)
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
df.head()
```

```python
df2 = df.copy()
```

```python
df['Bedroom2'] = df['Bedroom2'].astype(int)
df['Bathroom'] = df['Bathroom'].astype(int)
df['Car'] = df['Car'].astype(int)
df.info()
```

```python
df.isnull().sum()
```

```python
df.fillna(0)
```

```python
df['BuildingArea'].mean()
```

```python
int(df['YearBuilt'].mean())
```

```python
df.fillna({'BuildingArea':df['BuildingArea'].mean(), 'YearBuilt':df['YearBuilt'].mean().astype(int)})
```

```python
df['Regionname'].value_counts()
```

```python
df.groupby('Regionname')['YearBuilt'].mean().astype(int)
```

```python
df['YearBuilt'].isnull().sum()
```

```python
reg_year = df.groupby('Regionname')['YearBuilt'].transform('mean').astype(int)
reg_year
```

```python
df['YearBuilt'].fillna(reg_year)
```

```python
df['BuildingArea'].isnull().sum()
```

```python
numeric_df = df.select_dtypes(include=['number'])
numeric_df.info()
```

```python
numeric_df.corrwith(numeric_df['BuildingArea'])
```

```python
land_bultare = df.groupby('Landsize')['BuildingArea'].transform('mean')
df['BuildingArea'].fillna(land_bultare)
```

```python
df['Rooms'].isnull().sum()
```

```python
df['Rooms'].value_counts()
```

```python
df['Rooms'].mean()
```

```python
df.groupby(['Rooms', 'Type'])['BuildingArea'].mean()
```

```python
df[df['Rooms'] == 10]
```

```python
h_area = df[df['Type'] == 'h']['BuildingArea'].mean()
h_area
```

```python
df[(df['Rooms'] == 10) & (df['Type'] == 'h')].fillna(h_area)
```

```python
df['CouncilArea'].ffill()
```
