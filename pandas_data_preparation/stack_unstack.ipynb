```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(
    {
        'xodimlar': [25, 30, 12, 15, 40, 38]
    },
    index=pd.MultiIndex.from_tuples(
        [
            ('Google', 'AI'),
            ('Google', 'Marketing'),
            ('Amazon', 'AI'),
            ('Amazon', 'Marketing'),
            ('Meta', 'AI'),
            ('Meta', 'Marketing'),
        ],
        names=['kompaniya', 'bo‘lim']
    )
)
df
```

```python
df2 = df.unstack()
df2
```

```python
df2.columns
```

```python
df2.index
```

```python
# AI ustunidagi qiymatlarni olish
df2[('xodimlar', 'AI')]
```

```python
# asosiy ustun xodimlarni tashlab yuborish
df2 = df2.droplevel(level=0, axis=1)
df2
```

```python
df2.columns
```

```python
df2.index
```

```python
# indekslar nomini bo'limni o'chirish
df2 = df2.rename_axis(columns=None)
df2
```

```python
df2.stack()
```

```python
data = {
    ('xodimlar', 'AI'): [12, 25, 40],
    ('xodimlar', 'Marketing'): [15, 30, 38],
}
index = ['Amazon', 'Google', 'Meta']
columns = pd.MultiIndex.from_tuples(data.keys(), names=['asosiy_ustun', 'bo‘lim'])

df5 = pd.DataFrame(data, index=index, columns=columns)
df5.index.name = 'kompaniya'
```

```python
df5
```

```python
df5.unstack()
```

```python
# bo'lim ustunini ichki indeks qilish
df5.stack(level=1, future_stack=True)
```
