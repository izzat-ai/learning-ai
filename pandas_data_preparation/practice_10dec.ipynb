```python
df_g = pd.DataFrame({
    "Continent": ["Asia", "Asia", "Europe", "Europe"],
    "Country": ["Korea", "India", "Spain", "Italy"],
    "GDP": [1.6, 3.4, 1.4, 1.9]
}).set_index(["Continent", "Country"])
df_g
```

```python
df_g.swaplevel('Continent', 'Country').sort_index(level=0)
```

```python
df_i1 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Ali", "Vali", "Soli"]
})

df_i2 = pd.DataFrame({
    "ID": [1, 2, 4],
    "Salary": [900, 1100, 1000]
})
df_i1
```

```python
df_k1 = pd.DataFrame({"Name": ["A", "B"], "Score": [80, 90]})
df_k2 = pd.DataFrame({"Name": ["C", "D"], "Score": [85, 88]})
df_k1
```

```python
df_l1 = pd.DataFrame({"Q1": [10, 20]}, index=["Laptop", "Phone"])
df_l2 = pd.DataFrame({"Q2": [15, 25]}, index=["Laptop", "Phone"])
df_l1
```

```python
df_m = pd.DataFrame({
    "Year": [2022, 2022, 2023, 2023],
    "Product": ["Phone", "Laptop", "Phone", "Laptop"],
    "Sales": [500, 600, 550, 650]
})
df_m
```

```python
df_m.pivot(index='Year', columns='Product', values='Sales')
```
