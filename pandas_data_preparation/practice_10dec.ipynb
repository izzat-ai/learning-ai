```python
import numpy as np
import pandas as pd
```

```python
df_a = pd.DataFrame({
    "Country": ["USA", "USA", "Canada", "Canada"],
    "Year": [2022, 2023, 2022, 2023],
    "GDP": [25.3, 26.1, 2.1, 2.3]
}).set_index(["Country", "Year"])
df_a
```

```python
# gdp qiymatlarini ustunga aylantirish
df_a.unstack()
```

```python
df_b = pd.DataFrame({
    "City": ["London", "London", "Paris", "Paris"],
    "Type": ["Temperature", "Rain", "Temperature", "Rain"],
    "Value": [14, 120, 17, 80]
}).set_index(["City", "Type"])
df_b
```

```python
df_c = pd.DataFrame({
    "Region": ["Asia", "Asia", "Europe", "Europe"],
    "Country": ["Japan", "China", "Germany", "France"],
    "Population": [125, 1440, 83, 65]
})
df_c
```

```python
df_c = df_c.set_index(['Region', 'Country'])
df_c
```

```python
# faqatgina Asia mamlakatlarini olish
df_c.loc['Asia']
```

```python
df_d = pd.DataFrame({
    "Category": ["Fruit", "Fruit", "Drink", "Drink"],
    "Item": ["Apple", "Banana", "Cola", "Juice"],
    "Price": [3, 2, 5, 4]
})
df_d
```
