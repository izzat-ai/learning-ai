```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Country': ['USA', 'USA', 'USA', 'Germany', 'Germany', 'Germany', 'Japan', 'Japan'],
    'Year': [2022, 2023, 2024, 2022, 2023, 2024, 2022, 2023],
    'GDP': [23000, 24000, 25500, 4200, 4400, 4600, 5100, 5300],
    'Inflation': [8.0, 4.5, 3.2, 7.1, 6.0, 5.2, 2.8, 3.1]
}

df = pd.DataFrame(data)
df = df.set_index(['Country', 'Year'])
df
```

```python
df.xs(2023, level='Year')
```

```python
df.groupby(level=0)['GDP'].idxmax()
```

```python
data2 = {
    'Country': ['USA', 'USA', 'Germany', 'Germany', 'Japan', 'Japan'],
    'Year': [2022, 2023, 2022, 2023, 2022, 2023],
    'GDP': [23000, 24000, 4200, 4400, 5100, 5200]
}

df2 = pd.DataFrame(data)
df2
```

```python
# irarxik indeks yaratish
df2 = df2.set_index(['Country', 'Year'])
df2
```

```python
# germaniyaning barcha yillardagi ma'lumotlarini olish
df2.loc['Germany']
```

```python
# Germaniyaning 2023-yildagi ma'lumotini olish
df2.loc['Germany', 2023]
```

```python
# barcha davlatlarni faqatgina 2023-yildagi ma'lumotlarini olish
df2.loc[(slice(None), 2023), :]
```

```python
# 1-darajali indeks bo'yicha guruhlash
# har bir davlatning o'rtacha GDP sini aniqlash
df2.groupby(level='Country')['GDP'].mean()
```

```python
# irarxik indeksni - oddiy indeks qilish (to'liq holda)
df2.reset_index()
```

```python
data3 = {
    'Company': ['A', 'A', 'A', 'B', 'B', 'B'],
    'Year': [2022, 2023, 2024, 2022, 2023, 2024],
    'Revenue': [120, 150, 180, 100, 130, 160]
}
df3 = pd.DataFrame(data3)
df3
```

```python
# irarxik indeks qilish
df3 = df3.set_index(['Company', 'Year'])
df3
```

- Har bir kompaniya boyicha 2022  2024 oraligida daromad osish miqdorini hisoblang

```python
df3_un = df3.unstack()
df3_un
```

```python
df3_22_24 = df3_un.loc[:, (slice(None), [2022, 2024])]
df3_22_24
```

```python
# o'sishni hisoblab yangi ustunga joylash
#df3['gro']
df3_22_24['growth'] = df3_22_24['Revenue', 2024] - df3_22_24['Revenue', 2022]
df3_22_24
```

```python
df3_22_24['Revenue', 2024] - df3_22_24['Revenue', 2022]
```

```python
data4 = {
    'Country': ['USA', 'USA', 'USA', 'Germany', 'Germany', 'Germany'],
    'Year': [2021, 2022, 2023, 2021, 2022, 2023],
    'GDP': [21000, 22000, 23000, 4000, 4200, 4500]
}
df4 = pd.DataFrame(data4)
df4
```

```python
# har bir davlat uchun o'rtacha GDP ni hisoblash
df4.groupby(level='Country')['GDP'].mean()
```
