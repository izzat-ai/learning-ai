```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    "Region": ["Toshkent","Toshkent","Farg'ona","Farg'ona"],
    "Year": [2023, 2024, 2023, 2024],
    "Sales": [120, 150, 90, 110]
})
df
```

- uzun datasetdan , keng dataset yasash

```python
df_piv = df.pivot(index='Year', columns='Region', values='Sales')
df_piv
```

```python
df_piv.loc[2024]
```

- regionni indeks qilib , valuesni Sales qilamiz va ustunlarni year qilish

```python
df.pivot(index='Region', columns='Year', values='Sales')
```

```python
df1 = pd.DataFrame({
    "Country": ["Uzbekistan","Uzbekistan","Kazakhstan","Kazakhstan","Kyrgyzstan","Kyrgyzstan"],
    "Year": [2022, 2023, 2022, 2023, 2022, 2023],
    "GDP": [88, 95, 180, 190, 12, 14]
})
df1
```

```python
# Har bir davlatning yil bo‘yicha GDP ko‘rsatkichlari yonma-yon ustunlarda chiqadigan tartibli jadvalni yaratish
df1.pivot(index='Country', columns='Year', values='GDP')
```

```python
df2 = pd.DataFrame({
    "Product": ["Phone","Phone","Phone","Laptop","Laptop","Tablet","Tablet"],
    "Month": ["Jan","Feb","Mar","Jan","Feb","Jan","Mar"],
    "Sales": [120, 135, 150, 200, 210, 70, 90]
})
df2
```

```python
df2.pivot(index='Month', columns='Product', values='Sales')
```

```python
df3 = pd.DataFrame({
    "Company": ["Artel","Artel","Artel","Uzum","Uzum","Uzum","Nestle","Nestle"],
    "Quarter": ["Q1","Q2","Q3","Q1","Q2","Q3","Q1","Q2"],
    "Revenue": [50, 55, 60, 40, 42, 47, 90, 92]
})
df3
```
