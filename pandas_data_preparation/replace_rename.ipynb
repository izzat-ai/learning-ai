```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Holat': ['Yaxshi', 'Yomon', 'O\'rtacha', 'Yaxshi', 'Yomon', 'Yomon', 'Yaxshi'],

    'Sotuv_Xato_Kodi': [101, 204, 301, 101, 204, 404, 301],

    'Status_raqam': [5, 4, 3, 5, 4, 3, 5],

    'Joylashuv_Kod': ['SHM', 'JNB', 'GRB', 'SHM', 'JNB', 'SHM', 'GRB'],

    'Sharh': ['OK', 'Yomon', 'OK', 'OK', 'Yomon', 'Juda Yomon', 'OK']
}
df = pd.DataFrame(data)
df
```

```python
# holat ustunidagi o'rtacha qiymatini o'zgartirish
df.replace("O'rtacha", 'Yaxshi')
```

```python
# sotuv_xato_kodidagi 404 ni 301 qilish
df.replace(404, 301)
```

```python
# sharhdagi juda yomon qiymatini o'zgartirish
df.replace('Juda Yomon', 'Yomon')
```

```python
# sotuv_xato_kodi ustunidagi barcha qiymatlarni o'zgartirish
df.replace({101:100, 204:200, 301:300, 404:100})
```

```python
# ustun nomlarini kichik harflarda yozish
df.rename(columns=str.lower)
```

```python
# 3-va 5-indekslarni matnliga o'zgartirish + ustun nomlari kichik harflar
df.rename(index={3:'uchinchi', 5:'beshinchi'}, columns=str.lower)
```

```python
# sotuv_xato_kodi ustun nomini xato_kod ga o'zgartirish
df.rename(columns={'Sotuv_Xato_Kodi':"xato_kod"})
```

```python
# bir vaqtda 1-4-6-indekslarni matnli qilib yozish + Status_raqam va Joylashuv_kod nomlarini o'zgartirish
df.rename(index={1:'birinchi', 4:"to'rtinchi", 6:'oltinchi'}, columns={'Status_raqam':'status_num', 'Joylashuv_Kod':'place_code'})
```
