```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'Student': ['Ali', 'Vali', 'Hasan', 'Husan', 'Zoir', 'Dilshod', 'Bekzod', 'Kamol'],
    'Math': np.random.randint(50, 100, 8),
    'Physics': np.random.randint(40, 95, 8),
    'Chemistry': np.random.randint(45, 100, 8),
    'English': np.random.randint(30, 90, 8)
    })
df
```

```python
# English ustuni bo'yicha reytinglash | eng yuqori ball olgan o'quvchi birinchi
df['English'].rank(ascending=False)
```

```python
# eng past ball olgan o'quvchi 1-reytingda qilib belgilash
df['English'].rank(method='min')
```

```python
# o'rtacha qiymat bo'yicha reytinglash
df['English'].rank(method='average')
```

- `method=` ning 4 xil usuli bor . Hammasi , DF ning biror qiymati 2 va undan ortiq takrorlanib kelgan hollarda ishlatiladi
- `average` - bir xil qiymatlar bo'lsa o'rta reyting beriladi
- `min` - eng kichik reyting beriladi
- `max` - eng katta reyting beriladi
- `dense` - bitta o'rinni oladi va keyingi o'rin 1 taga oshadi
- `first` - teng qiymatlar qator tartibiga qarab belgilanadi

```python
data = {
    'Student': ['Ali', 'Vali', 'Hasan', 'Husan', 'Zoir', 'Dilshod', 'Bekzod', 'Kamol'],
    'Algebra': [88, 58, 93, 63, 75, 71, 69, 79],
    'Geometriya': [73, 92, 46, 51, 55, 48, 46, 48],
    'Informatika': [68, 76, 78, 65, 50, 51, 45, 39]
}

df = pd.DataFrame(data)
df
```

```python
# Algebra ustunidagi ballar bo'yicha reytinglash | eng yuqori bal 1-o'rin bo'ladi
df['algebra_reyting'] = df['Algebra'].rank(ascending=False)
df
```

```python
# Informatika ballarini reytinglash , bir xil balliklarni bitta o'ringa undan keyingisini keyingi o'ringa qo'yish
df['informatika_dense'] = df['Informatika'].rank(ascending=False, method='dense')
df
```
