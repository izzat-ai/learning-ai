```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    "Name": ["Ali", "Vali", "Hasan", "Gani", "Soli"],
    "Age": [23, 30, 27, 19, 35],
    "Salary": [400, 500, 450, 300, 700],
    "City": ["Toshkent", "Samarqand", "Toshkent", "Farg'ona", "Nukus"]
})
df
```

```python
# yoshi 25 dan kattalar
df[df['Age']>25]
```

```python
df2 = pd.DataFrame({
    "Product": ["Laptop", "Phone", "Tablet", "Camera", "Monitor"],
    "Price": [900, 700, 300, 1200, 500],
    "Stock": [5, 0, 12, 3, 7]
})
df2
```

```python
# Narxi 500 dan katta VA stokda mavjud bo‘lgan mahsulotlarni ajratish
df2[(df2['Price']>500) & (df2['Stock']>0)]
```

```python
df3 = pd.DataFrame({
    "Student": ["Ali", "Vali", "Hasan", "Gani", "Aziz"],
    "Grade": ["A", "C", "B", "A", "D"]
})
df3
```

```python
# Bahosi A yoki B bo‘lgan talabalarni aniqlash
df3[(df3['Grade'] == 'A') | (df3['Grade'] == 'B')]['Student']
```

```python
df4 = pd.DataFrame({
    "City": ["Toshkent", "Farg'ona", "Buxoro", "Samarqand", "Nukus"],
    "Population": [3000000, 1100000, 270000, 1000000, 330000]
})
df4
```

```python
# Farg‘ona, Toshkent va Samarqand shaharlarini ajratish
uch_shahar = ["Farg'ona", 'Toshkent', 'Samarqand']
df4[df4['City'].isin(uch_shahar)]
```

```python
df5 = pd.DataFrame({
    "Movie": ["Avatar", "Avengers", "Titanic", "Toy Story", "Top Gun"],
    "Genre": ["Action", "Action", "Drama", "Kids", "Action"]
})
df5
```

```python
# Nomi "A" harfi bilan boshlanadigan filmlarni filtrlash
df5[df5['Movie'].str.startswith("A")]
```

```python
df6 = pd.DataFrame({
    "Car": ["BMW", "Toyota", "Kia", "Honda", "Chevrolet"],
    "MaxSpeed": [240, 180, 200, 220, 190]
})
df6
```

```python
# MaxSpeed 190 dan 230 gacha bo‘lgan mashinalarni tanlab olish
df6[df6['MaxSpeed'].between(190, 230)]
```

```python
df7 = pd.DataFrame({
    "Name": ["Ibrohim", "Jasur", "Lola", "Malika", "Anvar"],
    "Age": [21, 29, 24, 32, 27],
    "Salary": [500, 700, 450, 800, 600]
})
df7
```

```python
# Yoshi 25 dan katta bo‘lganlarning faqat Name va Salary ustunlarini qaytarish
df7[df7['Age']>25][['Age', 'Salary']]
```
