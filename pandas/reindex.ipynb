```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Talaba_ID': np.arange(101, 112).tolist(), # 101 dan 111 gacha ID
    'Yosh': np.random.randint(18, 25, size=11).tolist(), # 18-24 yosh
    'Jins': ['Erkak', 'Ayol'] * 5 + ['Erkak'], # Jins (takrorlangan)
    'Asosiy_fan': np.random.choice(['Dasturlash', 'Iqtisod', 'Dizayn', 'Tibbiyot'], size=11).tolist(),
    '1-oralig_bahosi': np.random.randint(60, 100, size=11).tolist(),
    '2-oralig_bahosi': np.random.randint(65, 98, size=11).tolist(),
    'Davomat_foizi': np.random.randint(80, 100, size=11).tolist()
}

df = pd.DataFrame(data)
df
```

```python
df_rows5 = df.head()
df_rows5
```

```python
# indekslarni o'zgartirish
df_rows5.index = ['c', 'e', 'b', 'a', 'd']
```

```python
df_rows5
```

```python
# indekslarni tartiblash
idxs = ['c', 'e', 'b', 'a', 'd']
idxs.sort()
df_rows5.reindex(idxs)
```

```python
# kerakli indekslarni olib , keraksizlarini tashavorish
df_rows5.reindex(['a', 'c', 'e'])
```

```python
# indekslarni ko'proq qilish
new_idx = ['a', 'b', 'c', 'd', 'e', 'f', 'g']
df_nans = df_rows5.reindex(new_idx)
```

```python
df_nans
```

```python
# kerakli ustunlarni olib qoganini tashavorish
df_cols = df_rows5.reindex(columns=['Asosiy_fan', 'Talaba_ID', 'Yosh'])
```

```python
df_cols
```

```python
# df_cols indekslarini tartiblash
df_cols.index = df_cols.index.sort_values()
```

```python
df_cols
```

```python
# df dan tasodifiy 5 ta qatorni olish
df.sample(5)
```

```python
# kerakli qatorlar va ustunlarni opqolib qoganini tashavorish
df.loc[[2, 3, 4], ['Asosiy_fan', 'Yosh', 'Jins']]
```

```python
# df yaratish
data3 = {
    'index':['A', 'B', 'C', 'D'],
    'ism':['Ali', 'Valijon', 'Noila', 'Sardor'],
    'maosh':[5000, 6500, 4800, 7200]
}
df3 = pd.DataFrame(data3)
df3
```

```python
# index ustunini indeks qilish
df3.set_index(df3.columns[0], inplace=True)
df3
```

```python
# indekslar tartibini o'zgartirish
df3_new_idx = ['D', 'C', 'A', 'B']
new_df3 = df3.reindex(index=df3_new_idx)
```

```python
new_df3
```

```python
# df3 ni indekslarini kengaytirish
df3_new_idx2 = ['D', 'E', 'A', 'F', 'B']
new_df3_2 = df3.reindex(index=df3_new_idx2)
new_df3_2
```

```python
# ombor_df yaratish
ombor_data = {
    'index':['M1', 'M2', 'M3'],
    'mahsulot':['olma', 'banan', 'sabzi'],
    'zaxira_soni':[150, 200, 100]
}
df_ombor = pd.DataFrame(ombor_data)
df_ombor
```

```python
df4 = pd.DataFrame({
    'Product': ['Phone', 'Laptop', 'Tablet'],
    'Price': [500, 1200, 700]
}, index=[101, 103, 105])
df4
```

```python
s = pd.Series(
    [80, 85, 90],
    index=['Math', 'Physics', 'Chemistry'])
s
```
