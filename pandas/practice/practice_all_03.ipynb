```python
df16 = pd.DataFrame({
    'Department': ['IT', 'HR', 'IT', 'Sales', 'HR', 'IT'],
    'Employee': ['Ali', 'Vali', 'Sardor', 'Jasur', 'Kamol', 'Nodir'],
    'Salary': [5000, 4000, 5500, 4500, 4200, 6000]
})
df16
```

```python
# Group by department and calculate the average salary in each department
df16.groupby('Department')['Salary'].mean()
```

```python
# Group by Department and find the number of employees in each department
df16.groupby('Department')['Employee'].count()
```

```python
# Group by department and find maximum and minimum salaries
df16.groupby('Department')['Salary'].agg(['max', 'min'])
```

```python
df17 = pd.DataFrame({
    'Product': ['Laptop', 'Phone', 'Laptop', 'Tablet', 'Phone', 'Laptop'],
    'Store': ['A', 'A', 'B', 'A', 'B', 'A'],
    'Sales': [5, 8, 3, 6, 9, 4],
    'Revenue': [5000, 4000, 3000, 1800, 4500, 4000]
})
df17
```

```python
# Group by Product and Store and calculate total sales (Sales)
df17.groupby(['Product', 'Store'])['Sales'].sum()
```

```python
# Group by Product and find the average revenue (Revenue)
df17.groupby('Product')['Revenue'].mean()
```

```python
# Group by Store and calculate total revenue, average sales, and maximum revenue at the same time
df17.groupby('Store').agg({
    'Revenue':['sum', 'max'],
    'Sales':'mean'
})
```
