```python
df_sales = pd.DataFrame({
    "order_id": [1, 2, 3, 4, 5, 6],
    "region": ["Toshkent", "Samarqand", "Toshkent", "Buxoro", "Toshkent", "Andijon"],
    "sales": [1200, 850, 1500, 700, 1800, 950]
})
df_sales
```

```python
# faqatgina toshkentga oid ma'lumotlarni olish
df_sales[df_sales['region'] == 'Toshkent']
```

```python
# sotuv summasi 1000 dan yuqorilarini olish
df_sales[df_sales['sales']>1000]
```

```python
# Toshkentdagi buyurtmalarni 1000 dan yuqorilarini olish
df_sales[(df_sales['region'] == 'Toshkent') & (df_sales['sales']>1000)]
```

```python
df_products = pd.DataFrame({
    "product": ["Laptop", "Phone", "Tablet", "Monitor"],
    "price": [1200, 800, 600, 300],
    "quantity": [5, 10, 7, 12]
})
df_products
```

```python
# har bir mahsulot uchun umumiy daromadni hisoblash
df_products['revenues'] = df_products['price'] * df_products['quantity']
df_products
```

```python
# eng ko'p daromad keltirgan mahsulotni aniqlash
product_max = df_products['revenues'].idxmax()
product_max
```

```python
df_products.loc[product_max]
```

```python
df_employees = pd.DataFrame({
    "emp_id": [1, 2, 3, 4, 5],
    "name": ["Aziz", "Dilshod", "Kamola", "Malika", "Jasmin"],
    "salary": [4000, 5200, 4800, 6100, 4500]
})
df_employees
```

```python
df_orders = pd.DataFrame({
    "category": ["Electronics", "Clothing", "Electronics", "Clothing", "Food", "Food"],
    "order_amount": [120, 80, 200, 150, 60, 90]
})
df_orders
```

```python
# har bir kategoriya bo'yicha umumiy buyurtma summasini hisoblash
orders_group = df_orders.groupby('category')['order_amount'].sum()
orders_group
```

```python
# kategoriyalar bo'yicha o'rtacha buyurtma qiymatini aniqlash
df_orders.groupby('category')['order_amount'].mean()
```

```python
# eng ko'p daromad keltirgan kategoriyani aniqlash
df_orders.groupby('category')['order_amount'].sum().idxmax()
```

```python
df_scores = pd.DataFrame({
    "student": ["A", "B", "C", "D", "E"],
    "score": [85, None, 78, None, 92]
})
df_scores
```

```python
# qaysi qatorlarda nan borligini aniqlash
df_scores.isnull()
```

```python
# score ustunidagi nanlarni o'rta qiymat bn to'ldirish
df_scores['score'] = df_scores['score'].fillna(df_scores['score'].mean())
```

```python
df_scores
```
