```python
import numpy as np
import pandas as pd
```

```python
indekslar_df1 = ['A01', 'A02', 'A03', 'A04']
indekslar_df2 = ['A02', 'A04', 'A05', 'A06'] # A02 va A04 umumiy

# Ustunlarning ba'zilari umumiy, ba'zilari farqli
ustunlar_df1 = ['Ism', 'Maosh', 'Bo\'lim']
ustunlar_df2 = ['Maosh', 'Bonus', 'Bo\'lim'] # Maosh va Bo'lim umumiy

# 2. DF1 (Asosiy Ma'lumotlar)
data1 = {
    'Ism': ['Ali', 'Vali', 'Noila', 'Sardor'],
    'Maosh': [5000, 7500, 4800, 6200],
    'Bo\'lim': ['IT', 'HR', 'IT', 'Marketing']
}
df1 = pd.DataFrame(data1, index=indekslar_df1)

# 3. DF2 (Qo'shimcha Ma'lumotlar)
data2 = {
    'Maosh': [7500, 6200, 5500, 8000],  # Umumiy 'Maosh' ustuni
    'Bonus': [500, 800, 300, 1000],
    'Bo\'lim': ['HR', 'Marketing', 'Moliya', 'IT'] # Umumiy 'Bo'lim' ustuni
}
df2 = pd.DataFrame(data2, index=indekslar_df2)
```

```python
df1
```

```python
df2
```

```python
# df1 va df2 dan Maosh ustunini Series qilib olish
ser1 = df1['Maosh']
ser2 = df2['Maosh']
```

```python
ser1
```

```python
ser2
```

```python
# ser1 va ser2 ni indekslari bo'yicha qo'shish
ser1+ser2
```

```python
# df1 va df2 larni qator va ustunlari bo'yicha qp'shish
df1+df2
```

```python
# df lardan sonli ustunlarni ajratib olish
df1_2 = df1[['Maosh']]
df2_2 = df2[['Maosh']]
```

```python
df1_2+df2_2
```

```python
df1_2.add(df2_2)
```

```python
df1_2.add(df2_2, fill_value=0) # NaN larni o'rniga 0 qo'yib qo'shvoradi
```

```python
# df2_2 dan df1_2 ni ayirish
df2_2.sub(df1_2, fill_value=0)
```
