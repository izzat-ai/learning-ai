```python
import numpy as np
import pandas as pd
```

```python
indekslar_df1 = ['A01', 'A02', 'A03', 'A04']
indekslar_df2 = ['A02', 'A04', 'A05', 'A06'] # A02 va A04 umumiy

# Ustunlarning ba'zilari umumiy, ba'zilari farqli
ustunlar_df1 = ['Ism', 'Maosh', 'Bo\'lim']
ustunlar_df2 = ['Maosh', 'Bonus', 'Bo\'lim'] # Maosh va Bo'lim umumiy

# 2. DF1 (Asosiy Ma'lumotlar)
data1 = {
    'Ism': ['Ali', 'Vali', 'Noila', 'Sardor'],
    'Maosh': [5000, 7500, 4800, 6200],
    'Bo\'lim': ['IT', 'HR', 'IT', 'Marketing']
}
df1 = pd.DataFrame(data1, index=indekslar_df1)

# 3. DF2 (Qo'shimcha Ma'lumotlar)
data2 = {
    'Maosh': [7500, 6200, 5500, 8000],  # Umumiy 'Maosh' ustuni
    'Bonus': [500, 800, 300, 1000],
    'Bo\'lim': ['HR', 'Marketing', 'Moliya', 'IT'] # Umumiy 'Bo'lim' ustuni
}
df2 = pd.DataFrame(data2, index=indekslar_df2)
```

```python
data_mahsulot = {
    'Sotish_Narxi': [250, 120, 80],
    'Sotilgan_soni': [10, 25, 40]
}
indekslar = ['A', 'B', 'C']
mahsulot_df = pd.DataFrame(data_mahsulot, index=indekslar)

# 2. xarajat_df (Tannarx va Marketing)
data_xarajat = {
    'Tan_Narxi': [150, 60, 40],
    'Marketing_Xarajati': [50, 30, 15]
}
xarajat_df = pd.DataFrame(data_xarajat, index=indekslar)
```

```python
mahsulot_df
```

```python
xarajat_df
```

```python
# mahsulot_df ga yangi ustun qo'shish
mahsulot_df['jami_daromad'] = mahsulot_df['Sotilgan_soni']*mahsulot_df['Sotish_Narxi']
mahsulot_df
```

```python
# sotish_narxi dan tan_narx ni ayirish
foyda_ser = mahsulot_df['Sotish_Narxi'] - xarajat_df['Tan_Narxi']
foyda_ser
```

```python
# sotilgan_soni va foydani ko'paytirish
(foyda_ser * mahsulot_df['Sotilgan_soni']) - xarajat_df['Marketing_Xarajati']
```

```python
# sof_foyda yangi ustuniga yuqoridagi qiymatlarni berish
mahsulot_df['sof_foyda'] = (mahsulot_df['Sotish_Narxi']-xarajat_df['Tan_Narxi'])*mahsulot_df['Sotilgan_soni']-xarajat_df['Marketing_Xarajati']
mahsulot_df
```

**2-topshiriq**

```python
data_eski = {'Soni': [100, 150, 50]}
indekslar_eski = ['A', 'B', 'C']
ombor_eski = pd.DataFrame(data_eski, index=indekslar_eski)

# 2. ombor_yangi (Joriy oydagi zaxira)
data_yangi = {'Soni': [180, 60, 90]}
# Indekslar o'zgargan: B va C o'rnini almashtirgan, A o'rniga D kelgan.
indekslar_yangi = ['B', 'C', 'D']
ombor_yangi = pd.DataFrame(data_yangi, index=indekslar_yangi)
```

```python
ombor_eski
```

```python
ombor_yangi
```

```python
# ombor_yangidan ombor_eskini ayirib farq nomli DF yaratish
farq_df = ombor_yangi.sub(ombor_eski, fill_value=0)
farq_df
```

- `ombor_yangi` da mavjud bo'lmagan ombor_eski ning indekslariga 0 qo'yib ayiradi
