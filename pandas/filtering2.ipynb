```python
import pandas as pd
```

```python
data = {
    'ID': ['D001', 'D002', 'D003', 'D004', 'D005', 'D006', 'D007'],
    'Ism': ['Ali', 'Valijon', 'Noila', 'Hasan', 'Kamol', 'Dilshod', 'Zohid'],
    'Lavozim': ['Backend', 'Frontend', 'Backend', 'DevOps', 'Frontend', 'Backend', 'DataSci'],
    'Tajriba_yil': [5, 2, 8, 4, 1, 6, 3],
    'Maosh (USD)': [80000, 45000, 120000, 70000, 40000, 95000, 75000],
    'Shahar': ['Toshkent', 'Samarqand', 'Toshkent', 'Buxoro', 'Toshkent', 'Samarqand', 'Toshkent']
}

# DataFrame yaratish
dasturchi_df = pd.DataFrame(data)
```

```python
dasturchi_df
```

```python
# Tajribasi 5 yildan ko'p bo'lgan dasturchilarni tanlash
katta_tajriba = dasturchi_df[dasturchi_df['Tajriba_yil']>5]
katta_tajriba
```

```python
# Lavozimi Backend bo'lgan VA maoshi $90000$ dan yuqori bo'lgan dasturchilarni tanlash
filt1 = (dasturchi_df['Lavozim'] == 'Backend') & (dasturchi_df['Maosh (USD)'] > 90000)
yuqori_maosh = dasturchi_df[filt1]
yuqori_maosh
```

```python
# Yoki Toshkent shahrida ishlaydigan YOKI Lavozimi Frontend bo'lgan dasturchilarni tanlash
filt2 = (dasturchi_df['Shahar'] == 'Toshkent') | (dasturchi_df['Lavozim'] == 'Frontend')
poy_yoki_front = dasturchi_df[filt2]
poy_yoki_front
```

```python
# Ismi A harfi bilan boshlanadigan dasturchilarni tanlash
filt3 = dasturchi_df['Ism'].str.startswith('A')
ism_a = dasturchi_df[filt3]
ism_a
```
