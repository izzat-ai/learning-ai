**1-topshiriq**

```python
data = {
    'Mijoz': ['Alisher', 'Valijon', 'Noila', 'Sardor', 'Zaynab'],
    'Summa': [500, 1200, 850, 300, 2500],
    'Status': ['Tasdiqlandi', 'Rad etildi', 'Tasdiqlandi', 'Tasdiqlandi', 'Kutishda']
}

# Maxsus indekslar (Trans_ID)
tranzaksiya_idlar = ['T101', 'T102', 'T103', 'T104', 'T105']

tranzaksiya_df = pd.DataFrame(data, index=tranzaksiya_idlar)
tranzaksiya_df
```

```python
# T102 va T104 qatorlarini olish
tranzaksiya_df.loc[['T102', 'T104']]
```

```python
# Summa va Status ustunlarini olish
tranzaksiya_df.loc[:, ['Summa', 'Status']]
```

```python
# 0 dan 3 gacha bo'lgan indekslar ma'lumotlarini olish
tranzaksiya_df.iloc[0:3]
```

```python
# Mijoz va Summa ustun ma'lumotlarini indeks tartib raqami bo'yicha olish
tranzaksiya_df.iloc[:, [0, 1]]
```

```python
tranzaksiya_df
```

```python
# T105 ning Statusini "Tasdiqlandi" qilish
tranzaksiya_df.at['T105', 'Status'] = 'Tasdiqlandi'
tranzaksiya_df
```

```python
# T105 indeksi tartib raqami bo'yicha Summasini 3_000 ga o'zgartirish
tranzaksiya_df.iat[-1, 1] = 3000
tranzaksiya_df
```

**2-topshiriq**

```python
data2 = {
    'Ism': ['Alisher', 'Bobur', 'Dilfuza', 'Elmurod', 'Gavhar', 'Nodir'],
    'Sport': ['Futbol', 'Basketbol', 'Voleybol', 'Futbol', 'Voleybol', 'Basketbol'],
    'Yosh': [25, 30, 22, 28, 24, 32],
    'Maosh': [80000, 120000, 55000, 95000, 60000, 130000]
}

# Maxsus indekslar (Kiyim raqamlari)
kiyim_raqamlari = [7, 10, 5, 11, 9, 3]

sportchi_df = pd.DataFrame(data2, index=kiyim_raqamlari)

sportchi_df
```

```python
# Futbolchilarni maoshini 10% ga oshirish
sportchi_df.loc[sportchi_df['Sport'] == 'Futbol', 'Maosh'] = sportchi_df['Maosh']*1.10
```

```python
sportchi_df
```

```python
# 1-indeksdan 5-indeksgacha Ism va Yosh ustunlari qiymatlarini olish
sportchi_df.iloc[[1, 2, 3, 4, 5], [0, 2]]
```

```python
# indeksi 5 bo'lgan qatorning Yoshini 23 ga yangilash
sportchi_df.at[3, 'Yosh'] = 23
sportchi_df
```

```python
# tartib raqami 3 bo'lgan qatorning Maoshini (tartib raqami 4) 100000 ga yangilash
sportchi_df.iat[3, 3] = 100_000
sportchi_df
```
