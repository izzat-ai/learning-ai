```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Ism': ['Lola', 'Alisher', 'Komila', 'Rustam', 'Zaynab', 'Farhod'],
    'Lavozim': ['Hamshira', 'Shifokor', 'Hamshira', 'Shifokor', 'Teknisiy', 'Hamshira'],
    'Maosh': [4000, 9500, 4500, 11000, 3500, 4200],
    'Ish_staji': [3, 12, 5, 15, 2, 7],
    'Bo\'lim': ['Terapiya', 'Jarrohlik', 'Terapiya', 'Jarrohlik', 'Labaratoriya', 'Jarrohlik']
}

df = pd.DataFrame(data)
df
```

```python
# 2-indeksdagi ma'lumotlarni olish (Komilaning ma'lumotlarini olish)
df.loc[[2]]
```

- bu usul DataFrame qilib oladi
- `loc[2]` - series qilib oladi

```python
# Rustam va Zaynabning indeksi bo'yicha ma'lumotlarini olish
df.loc[[3, 4]]
```

```python
# Lola , Alisher va Komilaning ma'lumotlarini olish | indeksi orqali
df.loc[[0, 1, 2]]
```

```python
# Rustam va Farhodni ish_stajini olish
df.loc[[3, 5], ['Ish_staji']]
```

```python
# Komila va Zaynabning ish staji va bo'limini olish
df.loc[[2, 4], ['Ish_staji', "Bo'lim"]]
```

```python
df2 = pd.DataFrame(
    data={
        'Qiymat_A': np.random.randint(10, 100, size=8),
        'Qiymat_B': np.random.randint(50, 150, size=8),
        'Holat': np.array(['Yaxshi', 'Yomon'] * 4),
        'Sana': pd.to_datetime(['2025-10-01', '2025-10-05', '2025-10-10', '2025-10-15',
                                '2025-10-20', '2025-10-25', '2025-10-30', '2025-11-05'])
    },
    index=['X01', 'X02', 'X03', 'X04', 'X05', 'X06', 'X07', 'X08']
)
df2
```
