```python
data2 = {
    'Ism': ['Ali', 'Valijon', 'Noila', 'Sardor', 'Zaynab'],
    'Familiya': ['Olimov', 'Qodirov', 'Karimova', 'Vohidov', 'Ergasheva'],
    'Baho': [5, 4, 5, 3, 4],
    'Fan': ['Matematika', 'Fizika', 'Adabiyot', 'Fizika', 'Matematika'],
    'Guruh_ID': ['G1', 'G2', 'G1', 'G3', 'G2']
}

df2 = pd.DataFrame(data2)
df2
```

```python
# familiya va guruh ustunlarini o'chirish
df_tahrirlangan = df2.drop(['Familiya', 'Guruh_ID'], axis='columns')
```
