```python
haroratlar = [-2, 1, 8, 16, 22, 28, 32, 30, 25, 17, 9, 3]
oylar = ['Yan', 'Fev', 'Mar', 'Apr', 'May', 'Iyun', 'Iyul', 'Avgust', 'Sen', 'Okt', 'Noy', 'Dek']

ser2 = pd.Series(haroratlar, index=oylar)
ser2
```

```python
# aprel oyini haroratini olish
ser2['Apr']
```

```python
# iyul va noyabr oylarini haroratlarini olish
ser2[['Iyul', 'Noy']]
```

```python
# martdan avgustgacha bo'lgan oylar haroratini olish
ser2['Mar':'Avgust']
```

```python
# 20 gardusdan yuqori bo'lgan oylarni olish
ser2[ser2>20]
```

```python
narxlar = [150.50, 151.20, 149.80, 152.00, 153.50, 150.00, 154.10, 155.00]
sanalar = pd.to_datetime([
    '2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04',
    '2024-01-05', '2024-01-06', '2024-01-07', '2024-01-08'
])

ser3 = pd.Series(narxlar, index=sanalar)
ser3
```

```python
# 2024-01-04 sanasidagi narxni olish
ser3.loc['2024-01-04']
```

```python
# eng birinchi va eng ohirgi qiymatlarni olish
ser3[[0, -1]]
```
