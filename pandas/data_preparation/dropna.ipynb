```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Sotuvchi': ['Ahmad', 'Bekzod', 'Dilora', 'Farhod', 'Guli', 'Hasan', 'Kamol', 'Lola'],
    'Yanvar': [100, 150, np.nan, 200, 120, 180, 250, 90],
    'Fevral': [np.nan, 160, 110, 210, 130, np.nan, 260, 95],
    'Mart': [110, 170, 120, np.nan, 140, 190, 270, 100],
    'Apr_Bonusi': [500, np.nan, 550, 600, np.nan, 500, 700, np.nan],
    'Manzil': ['Toshkent', np.nan, 'Buxoro', 'Samarqand', np.nan, 'Toshkent', 'Buxoro', np.nan],
    'Baholash': [5, 4, 5, 3, 4, 5, 4, 3]
}
df = pd.DataFrame(data)
df
```

```python
# ustunlar bo'yicha nan qiymatlar sonini aniqlash
df.isnull().sum()
```

```python
# nan mavjud bo'lgan barcha qatorlarni tashavorish
df_dropna = df.dropna()
df_dropna
```

```python
# nan qatorlar tashalganidan keyin qolgan qatorlar sonini aniqlash
len(df_dropna)
```

```python
# nan qiymatga ega bo'lgan barcha ustunlarni tashavorish
df.dropna(axis=1)
```

```python
df.isnull().sum()
```

```python
# toza qiymatlari 6 tadan kam bo'lgan qatorlarni tashavorish
df.dropna(thresh=6)
```

```python
# toza qiymatlari 7 tadan kam bo'lgan ustunlarni tashavorish
df.dropna(thresh=7, axis=1)
```

```python
# ustunlarda necha foiz nanlar borligini aniqlash
df.isnull().sum()/len(df)*100
```

```python
df
```

```python
# manzil ustunidagi NaN larni tashavorish
df.dropna(subset=['Manzil'])
```

```python
# yanvar va manzil ustunlaridagi nanlarni tashavorish
df.dropna(subset=['Yanvar', 'Manzil'])
```
