```python
import numpy as np
import pandas as pd
```

**create a dataframe**

```python
# create dataframe with dictionary
data = {
    'Ism': ['Ali', 'Vali', 'Hasan', 'Husan'],
    'Yosh': [25, 30, 28, 22],
    'Shahar': ['Toshkent', 'Samarqand', 'Buxoro', 'Farg‘ona']
}
df = pd.DataFrame(data)
print(df)
```

```python
# get the first 5 rows of data from a dataframe
df.head()
```

```python
df.head(3)
```

```python
# determine the size of a dataframe
df.shape
```

```python
# create df with list
lst = [
    ['Ali', 25, 'Toshkent'],
    ['Vali', 30, 'Samarqand'],
    ['Hasan', 28, 'Buxoro'],
    ['Husan', 22, 'Farg‘ona']
]
df2 = pd.DataFrame(lst)
df2
```

```python
# change column order
df = pd.DataFrame(data, columns=['Shahar', 'Ism', 'Yosh'])
print(df)
```

```python
# assign values ​​to df indices
df2 = pd.DataFrame(lst, index=['a', 'b', 'c', 'd'])
print(df2)
```

```python
# Convert a column of df to a series
print(df.Shahar)
print(type(df.Shahar))
```

```python
# second method
print(df['Shahar'])
print(type(df['Shahar']))
```

```python
# change all values ​​in a column to a single number
df2[1] = 19
print(df2)
```

```python
# Modify some elements of a dataframe using Series
ser = pd.Series(data=["Qo'qon", "Yaypan", "Nursux", "Oqmasjid"], index=['a', 'c', 'd', 'b'])
df2[2] = ser
print(df2)
```

```python
# changing all values ​​of a column in a DF using list
lst2 = [51, 62, 22, 22]
df2[1] = lst2
print(df2)
```

```python
# add a new column to DF
df2['shahar'] = ['Far\'gona', 'Tuman', 'Farg\'ona', 'Tuman']
df2
```

```python
# delete a column
del df2[1]
```
