```python
df10 = pd.DataFrame({
    'Salary': [1000, 1200, 1100],
    'Bonus': [200, 300, 250]
})
df10
```

```python
# yangi ustun yaratish va umumiy daromadni hisoblash
df10['Income'] = df10['Bonus'] + df10['Salary']
df10
```

```python
df11 = pd.DataFrame({
    'Name': ['Ali', 'Vali', 'Soli'],
    'Score': [70, 85, 90]
})
df11
```

```python
def bahola(x):
  if x>85:
    return 5
  elif x>70:
    return 4
  else:
    return 3

# yangi ustun yaratib , baholash
df11['new_score'] = df11['Score'].apply(bahola)
df11
```

```python
# ismlarni kichik harflarda yozish
df11['Name'] = df11['Name'].str.lower()
df11
```

```python
df15 = pd.DataFrame({
    'Region': ['East', 'West', 'East', 'West'],
    'Sales': [100, 200, 150, 180]
})
df15
```
