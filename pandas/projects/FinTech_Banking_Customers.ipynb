- Ushbu sahifada FinTech va Bank ma'lumotlarini tahlil qildim . Maqsad : **mijozlarning demografik ko'rsatkichlari va tranzaksiyalarini tahlil qilib, ularning xarajat segmentlarini aniqlash**

```python
import numpy as np
import pandas as pd
```

```python
# 1-Jadval: Mijozlar bazasi
users_data = {
    'User_ID': [1, 2, 3, 4, 5],
    'Age': [28, 34, np.nan, 45, 22],
    'City': ['Tashkent', 'Samarkand', 'Bukhara', 'Tashkent', 'Samarkand'],
    'Segment': ['Silver', 'Gold', 'Silver', 'Platinum', 'Silver']
}
df_users = pd.DataFrame(users_data)

# 2-Jadval: Tranzaksiyalar tarixi
tx_data = {
    'Tx_ID': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    'User_ID': [1, 2, 1, 3, 4, 2, 99], # 99-ID li foydalanuvchi bizning bazada yo'q!
    'Amount': [150.0, 500.0, 45.0, 1200.0, 300.0, np.nan, 80.0], # 1006-tx da summa yo'q
    'Category': ['Electronics', 'Travel', 'Grocery', 'Electronics', 'Clothes', 'Travel', 'Grocery'],
    'Date': ['2026-01-15', '2026-01-16', '2026-02-10', '2026-02-11', '2026-03-01', '2026-03-02', '2026-03-15']
}
df_tx = pd.DataFrame(tx_data)
```

### Ma'lumotlarni o'rganish

```python
# mijozlar yoshidagi NaN qiymatlarni o'rtacha yosh bilan to'ldirish
df_users['Age'] = df_users['Age'].fillna(df_users['Age'].mean())
df_users
```

```python
# mijozlar yoshlarini butun son ko'rinishida saqlash
df_users['Age'] = df_users['Age'].astype(int)
df_users
```

```python
df_tx
```

```python
# tranzaksiyadagi summalarni kategoriyasiga qarab o'rtacha qiymat bilan to'ldirish
df_tx['Amount'] = df_tx.groupby('Category')['Amount'].transform(lambda x: x.fillna(x.mean()))
df_tx
```

```python
# Date ma'lumotlarini datetime formatiga o'tkizish
df_tx['Date'] = pd.to_datetime(df_tx['Date'])
df_tx
```

```python
# tranzaksiya qaysi oyda bo'lganini aniqlash
df_tx['Month'] = df_tx['Date'].dt.month_name()
df_tx
```

### Ma'lumotlarni birlashtirish

```python
# User_ID ustuni bo'yicha mijozlar va tranzaksiyalar ma'lumotlarini birlashtirish
df_bank = pd.merge(df_users, df_tx, on='User_ID', how='inner')
df_bank
```

```python
# viloyatlar bo'yicha daromadli hududlarni aniqlash | harajat va tranzaksiyalar soni orqali aniqlash
df_bank.groupby('City')['Amount'].agg(total_amount='sum', transactions='count').reset_index()
```

```python
# kim qayerga qancha pul sarflayotganini aniqlash pivot table yordamida
segment_cat = pd.pivot_table(df_bank, values='Amount', index='Segment', columns='Category', aggfunc='sum', fill_value=0)
segment_cat
```

```python
# 30 yoshdan yuqori va 200$ ko'p harajat qilgan mijozlarni aniqlash
df_bank[(df_bank['Age'] > 30) & (df_bank['Amount'] > 200)]
```
