```python
import numpy as np
import pandas as pd
```

```python
# Store sales data tables for one day
raw_data = {
    'Customer_ID': [101, 102, 103, 104, 105, 106],
    'Name': ['Anvar', 'Malika', None, 'Shaxzod', 'Zilola', 'Anvar'],
    'Product': ['Yengil Oyoq kiyim', 'Sumka', 'Telefon', 'Ko\'ylak', 'Sumka', 'Yengil Oyoq kiyim'],
    'Price': [45.0, np.nan, 800.0, 30.0, 65.0, 45.0],
    'Quantity': [2, 1, 1, 3, 2, 1],
    'Status': ['Yetkazildi', 'Bekor qilindi', 'Yetkazildi', 'Yetkazildi', 'Kutilmoqda', 'Yetkazildi']
}

df = pd.DataFrame(raw_data)
print(df)
```

```python
bag_price = df.loc[4, 'Price']
df.loc[1, 'Price'] = bag_price
```

```python
# delete duplicate information
df = df.drop_duplicates()
df
```

```python
# determine how much money each customer spent in total and save the data to a new table
df['Total_spent'] = np.where(df['Status'] == 'Bekor qilindi', 0.0, df['Price'] * df['Quantity'])
df
```

```python
# Customers whose products are shipped and whose purchase amount is over 50
df[(df['Status'] == 'Yetkazildi') & (df['Total_spent'] > 50)]
```
