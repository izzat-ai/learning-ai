```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    "Name": ["Ali", "Vali", "Hasan", "Husan", "Gulbahor"],
    "Age": [23, 30, 27, 25, 28],
    "City": ["Toshkent", "Samarqand", "Farg'ona", "Namangan", "Buxoro"],
    "Score": [85, 90, 78, 88, 92]
})
df
```

```python
# bir ustun nomini o'zgartirish
df.rename(columns={"Name":'name'}, inplace=True)
```

```python
df
```

```python
# indeks nomini yoki qiymatini o'zgartirish
df.rename(index={0:"nol"})
```

```python
# ustunni nomi yordamida o'chirish
df.drop(columns='Score')
```

```python
# indeksni nomi yordamida o'chirish
df.drop(index=3)
```

##### qayta indekslash `reindex`

```python
# series bilan qayta indekslash
s = pd.Series([10, 20, 30], index=["a", "b", "c"])
s
```

```python
s.reindex(['a', 'c', 'd'])
```

```python
# df bilan qayta indekslash | yo'q qiymatga NaN qo'yish
df2 = pd.DataFrame({
    "Name": ["Ali", "Vali", "Hasan"],
    "Age": [23, 30, 27]
}, index=[1, 2, 3])
df2
```

```python
df2.reindex([1, 3, 4])
```

```python
df.index = [4, 3, 2, 1, 0]
df
```

```python
# reindex bilan indekslarni tartiblash
df = df.reindex(sorted(df.index))
df
```

```python
# yangi indeksni qiymatlarini NaN o'rniga N qiymat bilan to'ldirish
df2.reindex([1, 3, 4], fill_value=0)
```

```python
df
```

```python
df.reindex(columns=['name', 'age', 'City', 'score'], fill_value=0)
```

```python
# ustunlarni tartiblash
df.reindex(columns=['City', 'name', 'Age', 'Score'])
```

```python
new_idx = [3, 'a', 2, 'c', 0]
df.reindex(index=new_idx, fill_value=0)
```

*1-topshiriq | `reindex`*

```python
ser2 = pd.Series([120, 150, 90], index=["yanvar", "fevral", "mart"])
ser2
```

```python
# ser2 ni "yanvar", "mart", "aprel", "may" indekslari bilan qayta indekslash
ser2.reindex(["yanvar", "mart", "aprel", "may"], fill_value=0)
```

```python
df3 = pd.DataFrame({"Name": ["Ali", "Vali", "Hasan", "Gani"], "Age": [23, 30, 27, 31]}, index=[3, 1, 4, 2])
df3
```

```python
# df3 indekslarini o'sish tartibida qayta indekslash
df3 = df3.reindex(sorted(df3.index))
df3
```

```python
df4 = pd.DataFrame({
    "age": [20, 25, 30],
    "salary": [300, 400, 500]
})
df4
```

```python
# df4 ustunlarini "salary", "age", "city" qilib qayta indekslash
df4.reindex(columns=["salary", "age", "city"])
```
