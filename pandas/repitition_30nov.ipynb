##### qayta indekslash `reindex`

```python
df.index = [4, 3, 2, 1, 0]
df
```

##### filtrlash

```python
df11 = pd.DataFrame({
    "Name": ["Ali", "Vali", "Hasan", "Gani", "Soli"],
    "Age": [23, 30, 27, 19, 35],
    "Salary": [400, 500, 450, 300, 700],
    "City": ["Toshkent", "Samarqand", "Toshkent", "Farg'ona", "Nukus"]
})
df11
```

```python
# yoshi 25 dan katta bo'lgan ma'lumotlarni olish
df11[df11['Age']>25]
```

```python
# Maoshi 500 ga teng bo‘lganlar
df11[df11.Salary == 500]
```

```python
# Toshkentda yashaydiganlar
df11[df11.City == 'Toshkent']
```

```python
# Yoshi 25 dan katta VA maoshi 400 dan yuqori
df11[(df11.Age>25) & (df11.Salary>400)]
```

```python
# Toshkentda yoki Samarqandda yashaydiganlar
df11[(df11.City == 'Toshkent') | (df11.City == 'Samarqand')]
```

```python
# Yoshi 20–30 orasida bo‘lganlar
df11[(df11['Age']>=20) & (df11['Age']<=30)]
```

```python
# Toshkent yoki Farg‘onada yashaydiganlar
tosh_far = ['Toshkent', "Farg'ona"]
df11[df11['City'].isin(tosh_far)]
```

```python
# Yoshi 23 yoki 35 bo‘lganlar
df11[df11['Age'].isin([23, 35])]
```
