```python
df = pd.DataFrame({
    "A": [12, 45, 33, 8, 27, 19],
    "B": [5.5, 7.8, 3.3, 9.1, 6.4, 8.0],
    "C": [100, 120, 95, 130, 110, 90],
    "D": [-3, 14, -7, 20, 0, 5]
})
df
```

```python
data = {
    'Yanvar': [1200, 900, 1500, 800],
    'Fevral': [1500, 1100, 1400, 950],
    'Mart': [1800, 1000, 1900, 1100],
    'Aprel': [1400, 1300, 1700, 900]
}

# Maxsus indekslar (Davlatlar)
davlatlar = ['AQSh', 'Germaniya', 'Yaponiya', 'Braziliya']

sotuvlar_df = pd.DataFrame(data, index=davlatlar)
sotuvlar_df
```

```python
# Har bir oy bo'yicha jami sotuvlar yig'indisini aniqlash
jami_oylik = sotuvlar_df.sum()
jami_oylik
```

```python
# Har bir davlat uchun o'rtacha oylik sotuvni aniqlash
mean_davlat = sotuvlar_df.mean(axis=1)
mean_davlat
```

```python
sotuvlar_df
```

```python
# Har bir oyda sotuvlar hajmi eng yuqori bo'lgan davlatning nomini aniqlash
max_davlat = sotuvlar_df.idxmax()
max_davlat
```

```python
# Yaponiya uchun sotuvlar hajmi eng past bo'lgan oyning nomini aniqlash
sotuvlar_df2 = sotuvlar_df.T
sotuvlar_df2['Yaponiya'].idxmin()
```

```python
sotuvlar_df.idxmin(axis='columns')
```

```python
sotuvlar_df.loc['Yaponiya'].idxmin()
```

```python
# Mart oyida sotuvlar hajmi eng yuqori bo'lgan davlatning tartib raqamini
sotuvlar_df['Mart'].argmax()
```

```python
data2 = {
    '1-chorak': [45, 52, 60, 40],
    '2-chorak': [55, 48, 62, 50],
    '3-chorak': [60, 50, 65, 45],
    '4-chorak': [48, 58, 55, 42]
}

# Maxsus indekslar (Xodim ID'lari)
xodim_idlar = ['E001', 'E002', 'E003', 'E004']

# DataFrame yaratish
df2 = pd.DataFrame(data2, index=xodim_idlar)
df2
```

```python
# barcha xodimlarni har bir chorakdagi o'rtacha balini aniqlash
orta_chorak = df2.mean(axis=1)
orta_chorak
```

```python
# Har bir xodim uchun uning eng yuqori va eng past chorak natijalari o'rtasidagi farqni aniqlash
farq_ser = df2.max(axis=1) - df2.min(axis=1)
farq_ser
```

```python
# eng past farqda ishlagan xodim idsini aniqlash
barqaror_xodim = farq_ser.idxmin()
barqaror_xodim
```

```python
# butun DF ichidan eng yuqori elementni tartib raqamini aniqlash
df2.values.argmax()
```
