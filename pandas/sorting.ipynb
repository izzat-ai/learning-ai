```python
import numpy as np
import pandas as pd
```

```python
indexlar = ['x3', 'a1', 'k9', 'b2', 'm7', 'z0', 'q4', 'd8']

# 7 ta ustun nomi â€” tartibsiz aralash
ustunlar = ['J', 'A', 'X', 'D', 'C', 'M', 'B']

df = pd.DataFrame(np.random.randn(8, 7) * 10, index=indexlar, columns=ustunlar)
df
```

```python
# ustunlarni tartiblash
df.sort_index(axis=1, inplace=True)
```

```python
# indekslarni tartiblash
df.sort_index(inplace=True)
```

```python
df
```

```python
# ustunlarni teskari qilib tartiblash
df.sort_index(ascending=False, axis=1)
```

```python
# indekslarni teskari tartiblash
df.sort_index(ascending=False)
```

```python
# C ustuni qiymatlari bo'yicha tartiblash
df.sort_values(by='C')
```

```python
# C ustunidagi qiymatlarni teskari tartiblash
df.sort_values(by='C', ascending=False)
```

```python
# D va M ustunlari qiymatlarini tartiblash
df.sort_values(by=['M', 'D'])
```

```python
# D va M ustunlarini teskari tartiblash
df.sort_values(by=['D', 'M'], ascending=False)
```

```python
data = {
    'Mintaqa': ['Shimol', 'G\'arb', 'Sharq', 'Janub', 'Shimol'],
    'Xodimlar_soni': [45, 30, 55, 20, 45],
    'Savdo_hajmi': [120000, 85000, 150000, 60000, 130000]
}

# Maxsus indekslar (Filial_Nomi)
filial_nomlari = ['Markaz', 'G\'arb', 'Sharq', 'Janub', 'Shimoliy_2']

# DataFrame yaratish
df2 = pd.DataFrame(data, index=filial_nomlari)
df2
```

```python
# DataFrame'ni uning nomlangan indekslari bo'yicha teskari tartibda saralang
df_index_teskari = df2.sort_index(ascending=False)
df_index_teskari
```

```python
# Avval Xodimlar_soni bo'yicha o'suvchi tartibda, so'ngra (xodimlar soni bir xil bo'lsa) Savdo_hajmi bo'yicha kamayuvchi tartibda saralang
# natijani df_qiymat_tartibi ga saqlash
df_qiymat_tartibi = df_index_teskari.sort_values(by='Xodimlar_soni')
df_qiymat_tartibi
```

```python
df_qiymat_tartibi = df_index_teskari.sort_values(by='Savdo_hajmi')
df_qiymat_tartibi
```

```python
data2 = {
    'Zaxira_soni': [450, 120, 80, 120, 250],
    'Narx_USD': [120, 50, 250, 80, 180],
    'Kategoriya': ['Elektronika', 'Aksessuar', 'Elektronika', 'Aksessuar', 'Elektronika'],
    'Eng_ko\'p_sotuv_oy': ['Aprel', 'Mart', 'Yanvar', 'May', 'Aprel']
}

# Maxsus indekslar (Partiya ID'lari)
partiya_idlar = ['ZK-201', 'ZK-105', 'ZK-310', 'ZK-102', 'ZK-400']

# DataFrame yaratish
mahsulot_inventar = pd.DataFrame(data2, index=partiya_idlar)
mahsulot_inventar
```

```python
# DataFrame'ni ustun nomlari bo'yicha teskari alifbo tartibida saralash
df_ustun_teskari = mahsulot_inventar.sort_index(axis=1, ascending=False)
df_ustun_teskari
```

```python
# Avval Kategoriya bo'yicha o'suvchi tartibda, So'ngra, Zaxira_soni bo'yicha o'suvchi tartibda, Narx_USD bo'yicha kamayuvchi tartibda saralash
# natijani df_murakkab_tartibga saqlash
df_murakkab_tartib = mahsulot_inventar.sort_values(by=['Kategoriya', 'Zaxira_soni', 'Narx_USD'], ascending=[True, True, False])
df_murakkab_tartib
```

```python
# Narx_USD ustunini series qilib olish
narx_ser = mahsulot_inventar['Narx_USD']
narx_ser
```

```python
# series qiymatlarini o'suvchi tartibda saqlash
narx_ser = narx_ser.sort_values()
narx_ser
```
