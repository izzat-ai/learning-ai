```python
import numpy as np
import pandas as pd
```

```python
n = 100
age = np.random.randint(18, 60, n)
salary = age * 120 + np.random.normal(0, 300, n)  # Age bilan kuchli bog‘liq

# Kuchli manfiy korrelyatsiya
hours_worked = np.random.randint(4, 10, n)
free_time = 15 - hours_worked + np.random.normal(0, 0.5, n)

# O‘rtacha korrelyatsiya
experience = age * 0.4 + np.random.normal(0, 3, n)

# Bog‘liq bo‘lmagan ustun (noise)
random_score = np.random.randint(0, 100, n)

df = pd.DataFrame({
    "Age": age,
    "Salary": salary,
    "HoursWorked": hours_worked,
    "FreeTime": free_time,
    "Experience": experience,
    "RandomScore": random_score
})
df
```

```python
# yosh va ish_haqqi orasidagi bog'liqni tekshirish
df['Age'].corr(df['Salary'])
```

```python
# ish_haqqi va ish vaqti bog'liqligini aniqlash
df['Salary'].corr(df['HoursWorked'])
```

```python
# ish haqqi va tajribani bog'liqligini aniqlash
df['Salary'].corr(df['Experience'])
```

```python
# bo'sh vaqt va ish haqqi bog'liqligi qanchalik
df['FreeTime'].corr(df['Salary'])
```

```python
# bo'sh vaqt va yoshni bog'liqligi
df['FreeTime'].corr(df['Age'])
```

```python
# ish vaqti va tajriba
df['HoursWorked'].corr(df['Experience'])
```

```python
# ish haqqini barcha ustunlar bilan bog'liqligini aniqlash
df.corrwith(df['Salary'])
```

```python
data = {
    'Reklama_Xarajati': [1000, 1500, 800, 2000, 1200],
    'Ijtimoiy_Tarmoq': [500, 700, 450, 1100, 600],
    'Veb_Trafik': [1200, 1800, 900, 2500, 1500],
    'Sotuvlar': [50, 75, 40, 90, 60]
}

# Maxsus indekslar (Kampaniya ID'lari)
kampaniya_idlar = ['K1', 'K2', 'K3', 'K4', 'K5']

# DataFrame yaratish
kampaniya_df = pd.DataFrame(data, index=kampaniya_idlar)
kampaniya_df
```

```python
# barcha ustunlar korrelyasiyasini aniqlash
kampaniya_df.corr()
```

```python
# reklama_harajati va sotuvlarni bog'liqligini aniqlash
kampaniya_df['Reklama_Xarajati'].corr(kampaniya_df['Sotuvlar'])
```
