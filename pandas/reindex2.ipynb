```python
import numpy as np
import pandas as pd
```

```python
sales = pd.Series(
    [120, 300, 180],
    index=['January', 'March', 'June']
)
sales
```

```python
# January, February, March, April, May, June oylar tartibida indekslash
months = ['January', 'February', 'March', 'April', 'May', 'June']
sales.reindex(months)
```

```python
df = pd.DataFrame({
    'Product': ['Phone', 'Laptop', 'Tablet'],
    'Sales': [500, 800, 300]
}, index=[2, 4, 7])
df
```

```python
# 0, 1, 2, 3, 4, 5 indekslardan tashkil topgan DF hosil qilish
df.reindex([0, 1, 2, 3, 4, 5])
```

```python
df_temp = pd.DataFrame({
    'Temperature': [15, 18, 20, 22]
}, index=['Mon', 'Wed', 'Fri', 'Sun'])
df_temp
```

```python
weekdays = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
df_temp.reindex(weekdays)
```
