```python
import numpy as np
import pandas as pd
```

1. Dataset sun'iy intellektdan olinadi
2. Dataset : Xususiy klinikaning 3 yillik bemorlar va shifokorlar tahlili bazasi

```python
data = {
    'Filial': ['Toshkent', 'Toshkent', 'Toshkent', 'Toshkent', 'Samarqand', 'Samarqand', 'Samarqand', 'Samarqand'],
    'Yonalish': ['Kardiologiya', 'Kardiologiya', 'Pediatriya', 'Pediatriya', 'Kardiologiya', 'Kardiologiya', 'Pediatriya', 'Pediatriya'],
    'Yil': [2024, 2025, 2024, 2025, 2024, 2025, 2024, 2025],
    'Bemorlar_Soni': [120, 150, 300, 280, 80, 95, 210, 240],
    'Umumiy_Tushum': [45000000, 60000000, 35000000, 42000000, 28000000, 35000000, 22000000, 29000000]
}

df_clinic = pd.DataFrame(data)
df_clinic
```

```python
# Filial va Yonalish ustunlarini irarxik indeks qilish
df_irar = pd.pivot_table(df_clinic,
               index=['Filial', 'Yonalish'])
df_irar
```

```python
# irarxik jadvalni indekslarini tartiblash
df_irar = df_irar.sort_index()
df_irar
```

```python
# Toshkentning ma'lumotlarini olish
df_irar.loc['Toshkent']
```

```python
# barcha filialdagi Pediatriyalar ma'lumotlarini olish
df_irar.xs('Pediatriya', level=1)
```

```python
# Yonalish indeksini ustun qilish
df_irar.reset_index(level=1)
```

```python
# irarxik indekslarni yo'q qilib , jadvalni eski holatiga qaytarish
df_irar.reset_index()
```

#####**Yangi dataset bilan ishlash**

```python
data2 = {
    'Shifokor': ['Anvarov', 'Anvarov', 'Karimova', 'Karimova', 'Anvarov', 'Anvarov', 'Karimova', 'Karimova'],
    'Yonalish': ['Kardiologiya', 'Kardiologiya', 'Pediatriya', 'Pediatriya', 'Kardiologiya', 'Kardiologiya', 'Pediatriya', 'Pediatriya'],
    'Filial': ['Toshkent', 'Samarqand', 'Toshkent', 'Samarqand', 'Toshkent', 'Samarqand', 'Toshkent', 'Samarqand'],
    'Yil': [2024, 2024, 2024, 2024, 2025, 2025, 2025, 2025],
    'Xizmatlar_Soni': [150, 90, 310, 220, 180, 110, 290, 260],
    'Reyting': [4.8, 4.2, 4.9, 4.5, 4.9, 4.4, 5.0, 4.7]
}

df_raw = pd.DataFrame(data2)
df_raw
```

```python
# shifokor, yonalishi va filialini irarxik indeks qilib ma'lumotlarni ko'rish
df_3idx = df_raw.set_index(['Shifokor', 'Yonalish', 'Filial'])
df_3idx
```

```python
# faqat indekslarni ko'rish
df_3idx.index
```

```python
# shifokor va yonalish indekslar o'rnini almashtirish
df_3idx = df_3idx.swaplevel('Shifokor', 'Yonalish').sort_index()
df_3idx
```

```python
# faqatgina Samarqand filiali ma'lumotlarini olish
df_samarqand = df_3idx.xs('Samarqand', level=2)
df_samarqand
```

```python
# barcha irarxik indekslarni eski holga keltirish
df_3idx.reset_index()
```
