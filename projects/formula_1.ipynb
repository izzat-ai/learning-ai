```python
# check
def check_health(df, name):
    print(f"--- {name} data report ---")
    print(f"null values:\n{df.isnull().sum()}\n")
    print(f"number of duplicates: {df.duplicated().sum()}")
    print("-" * 30)

check_health(cons, "Constructors")
check_health(cons_standing, "Standings")
check_health(races, "Races")
```

- `const` dataframe - No empty (null) values ​​and no duplicates
- `cons_standing` dataframe - The position column contains 212 null values . In F1 history, some constructors may have been disqualified or not finished the season
- `races` dataframe - 731 empty values ​​(null) . The exact hours and minutes of older races (e.g., from the 1950s-60s) may not be stored in the historical database
